```python
def count_words(text):
    """Подсчет слов в строке с учетом лишних пробелов"""
    if not text or text.isspace():  # Проверка на пустую строку или строку из пробелов
        return 0
    
    # Убираем пробелы в начале и конце
    text = text.strip()
    
    # Заменяем множественные пробелы на одиночные
    while '  ' in text:
        text = text.replace('  ', ' ')
    
    # Подсчитываем слова
    return text.count(' ') + 1

# Тестирование
test_strings = [
    "  Дана   строка  с   лишними   пробелами  ",
    "Привет мир",
    "   ",
    "",
    "Слово",
    "  Много   пробелов   здесь   "
]

for s in test_strings:
    print(f"'{s}' -> {count_words(s)} слов")
```

```python
def counting_words_1(text):
    return len(text.split())

# Тестирование
test_strings = [
    "  Дана   строка  с   лишними   пробелами  ",
    "Привет мир",
    "   ",
    "",
    "Слово",
    "  Много   пробелов   здесь   "
]

for s in test_strings:
    print(f"'{s}' -> {counting_words_1(s)} слов")
```

```python
text = 'asdfghlkjhgf'
first = text.find('h')
second = text.rfind('h')
result = text[:first] + text[second:first:-1] + text[second:]
print(result)
```

```python
def reverse_between_h_1(s):
    """
    Разворачивает последовательность символов между первым и последним вхождением 'h'.
    """
    first_h = s.find('h')
    last_h = s.rfind('h')
    return s[:first_h + 1] + s[first_h + 1:last_h][::-1] + s[last_h:]


# Тестирование в компактном формате
test_strings = [
    "hAhBhCh",
    "hello world h",
    "hh",
    "h123h",
    "h   h",
    "hHh",
    "start h middle h end",
    "h!@#$%h",
    "abcdefghijklmnopqrstuvwxyzh",
    "h h h h"
]

print("=" * 60)
print("ТЕСТИРОВАНИЕ reverse_between_h")
print("=" * 60)

for i, s in enumerate(test_strings, 1):
    result = reverse_between_h_1(s)
    print(f"Тест {i:2d}: '{s:40}' -> '{result}'")

print("=" * 60)
print(f"Всего тестов: {len(test_strings)}")
print("=" * 60)
```

## Задача 4

Начальник кадровой службы хочет узнать, сколько мужчин-однофамильцев работает в организации.\
У него есть список фамилий, и на основании этого списка нужно вычислить количество фамилий, которые совпадают с другими.

```python
# Тестирование
test_cases = [
    ["Иванов", "Петров", "Сидоров", "Иванов", "Петров", "Кузнецов"],  # Иванов(2), Петров(2) -> 4
    ["Иванов", "Иванов", "Иванов", "Петров", "Сидоров"],               # Иванов(3) -> 3
    ["Иванов", "Петров", "Сидоров"],                                   # Нет однофамильцев -> 0
    ["Иванов", "Иванов", "Петров", "Петров", "Сидоров", "Сидоров"],    # Все по 2 -> 6
    ["Иванов"],                                                         # Один человек -> 0
    [],                                                                 # Пустой список -> 0
    ["Петров", "Петров", "Петров", "Петров", "Иванов"],                # Петров(4) -> 4
    ["Иванов", "Петров", "Иванов", "Сидоров", "Петров", "Кузнецов"],   # Иванов(2), Петров(2) -> 4
    ["Абрамов", "Абрамов", "Абрамов"],                                 # Абрамов(3) -> 3
    ["Смирнов", "Смирнов", "Кузнецов", "Попов", "Попов", "Васильев"]   # Смирнов(2), Попов(2) -> 4
]
```

```python
def count_namesake_males(surnames):
    """
    Подсчитывает количество фамилий, которые совпадают с другими (однофамильцев).
    
    Args:
        surnames: список фамилий
    
    Returns:
        Количество фамилий, имеющих хотя бы одного однофамильца
    """
    # Находим фамилии, которые встречаются больше одного раза
    from collections import Counter
    surname_counts = Counter(surnames)
    
    # Суммируем количество всех фамилий, которые встречаются > 1 раза
    return sum(count for count in surname_counts.values() if count > 1)

for i, surnames in enumerate(test_cases, 1):
    result = count_namesake_males(surnames)
    
    # Формируем строку с фамилиями для красивого вывода
    surnames_str = str(surnames)
    if len(surnames_str) > 50:
        surnames_str = surnames_str[:47] + "..."
    
    print(f"Тест {i:2d}: {surnames_str:55} -> {result:2d} однофамильцев")
```

```python
# Альтернативное решение без Counter
def count_namesake_males_alt(surnames):
    """
    Подсчитывает количество фамилий, которые совпадают с другими.
    Альтернативное решение без использования Counter.
    """
    unique_surnames = set(surnames)
    total_namesakes = 0
    
    for surname in unique_surnames:
        count = surnames.count(surname)
        if count > 1:
            total_namesakes += count
    
    return total_namesakes

for i, surnames in enumerate(test_cases, 1):
    result = count_namesake_males_alt(surnames)
    
    # Формируем строку с фамилиями для красивого вывода
    surnames_str = str(surnames)
    if len(surnames_str) > 50:
        surnames_str = surnames_str[:47] + "..."
    
    print(f"Тест {i:2d}: {surnames_str:55} -> {result:2d} однофамильцев")
```

```python
# Компактное решение в одну строку
def count_namesake_males_oneliner(surnames):
    """
    Однострочное решение для подсчета однофамильцев.
    """
    return sum(surnames.count(s) for s in set(surnames) if surnames.count(s) > 1)

for i, surnames in enumerate(test_cases, 1):
    result = count_namesake_males_oneliner(surnames)
    
    # Формируем строку с фамилиями для красивого вывода
    surnames_str = str(surnames)
    if len(surnames_str) > 50:
        surnames_str = surnames_str[:47] + "..."
    
    print(f"Тест {i:2d}: {surnames_str:55} -> {result:2d} однофамильцев")
```

```python
# Тестирование
test_cases = [
    ["яблоко", "груша", "банан", "киви", "ананас"],                    # Разная длина
    ["кот", "пёс", "слон", "тигр", "лев"],                              # Одинаковая длина
    ["aaa", "bb", "c", "dd", "eee"],                                    # Разная длина
    ["", "a", "ab", "abc", "abcd"],                                     # Пустая строка
    ["hello", "world", "python", "code", "test"],                       # Одинаковая длина
    ["zzz", "aaa", "bbb", "ccc"],                                       # Одинаковая длина, разные буквы
    ["Анна", "Борис", "Виктор", "Алексей", "Дмитрий"],                  # Русские имена
    ["12", "1", "123", "1234", "12345"],                                # Цифры как строки
    ["  ", " ", "   ", "    ", ""],                                     # Пробелы
    ["кот", "кот", "кот", "пёс", "пёс"]                                 # Дубликаты
]
```

```python
def sort_strings(strings):
    """
    Сортирует строки по возрастанию длины, а строки одинаковой длины
    сортирует по возрастанию в алфавитном порядке.
    
    Args:
        strings: список строк
    
    Returns:
        Отсортированный список строк
    """
    return sorted(strings, key=lambda s: (len(s), s))


for i, strings in enumerate(test_cases, 1):
    sorted_result = sort_strings(strings)
    
    print(f"\nТест {i}:")
    print(f"  Исходный список: {strings}")
    print(f"  Отсортированный: {sorted_result}")
    print(f"  Порядок сортировки:")
    
    # Показываем ключи сортировки для каждого элемента
    print("  Ключи сортировки (длина, строка):")
    for s in sorted_result:
        print(f"    '{s}' -> ({len(s)}, '{s}')")
```

```python
# Компактное решение в одну строку
sort_strings_oneliner = lambda strings: sorted(strings, key=lambda s: (len(s), s))

for i, strings in enumerate(test_cases, 1):
    sorted_result = sort_strings_oneliner(strings)
    
    print(f"\nТест {i}:")
    print(f"  Исходный список: {strings}")
    print(f"  Отсортированный: {sorted_result}")
    print(f"  Порядок сортировки:")
    
    # Показываем ключи сортировки для каждого элемента
    print("  Ключи сортировки (длина, строка):")
    for s in sorted_result:
        print(f"    '{s}' -> ({len(s)}, '{s}')")
```
