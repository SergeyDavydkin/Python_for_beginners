```python
# создадим Series объект
import pandas as pd
s = pd.Series([1,2,3,4], index=['a', 'b', 'c', 'd'])
print(s)
print('=='*30)

x = pd.Series(np.random.randint(-5, 6, 10),
              index=list('abcdefghij'))
print(x)
print('=='*30)

print('вывод индексов')
print(s.index)
print(x.index)
```

```python
# Важно отметить, что элементы индекса могут иметь одинаковые значения:
y = pd.Series(np.random.randint(-5, 6, 10),
              index=['a']*5 + ['b']*5)
y
```

```python
# Для создания `Series` необязательно указывать индексы, 
# `pandas` автоматически проставит числовой индекс. 
# Также `Series` можно создавать, передавая не только список с данными, 
# но и словарь, ключи которого станут индексами в созданном `Series`.
d = {'Moscow': 1000, 'London': 300, 'Prague': None}
cities = pd.Series(d)
print(cities)

# Если вместе со словарем указан еще и индекс, то из данных будут извлечены 
# только те элементы, ключи которых присутствуют в `index`:
data = {i: np.random.randint(10) for i in 'abcdefghij'}
z = pd.Series(data)
z = pd.Series(data, index=list('facdxyz'))
print(z)
```

```python
# Для ключей, которых нет в `data`, но есть в `index` 
# создаются значения `NaN`, которые обозначают отсутствующие элементы.
w = pd.Series(777, index=list('abcdefghij'))
w
```

```python
q = pd.Series(np.random.randint(-5, 6, 10),
              index=list('abcdefghij'))
print(q)
print('=='*30)
print('=='*30)

print('Извлечение разреженного среза (среза с заданным шагом 2):')
print(q[::2])
print('=='*30)
```

## Часть 1: Структура данных Series

### Задание 1.1. Создание Series из списка Python

```python
import pandas as pd
import numpy as np

# Создание Series из списка
s1 = pd.Series([10, 20, 30, 40, 50])
print("Series из списка:")
print(s1)
print(f"Тип: {type(s1)}")
print(f"Индексы: {s1.index}")
print(f"Значения: {s1.values}")
```

### Задание 1.2. Создание Series из ndarray массива numpy

```python
# Создание Series из массива numpy
arr = np.array([1.5, 2.5, 3.5, 4.5, 5.5])
s2 = pd.Series(arr)
print("\nSeries из numpy массива:")
print(s2)
```

### Задание 1.3. Создание Series из словаря (dict)

```python
# Создание Series из словаря (ключи становятся индексами)
data_dict = {'a': 100, 'b': 200, 'c': 300, 'd': 400}
s3 = pd.Series(data_dict)
print("\nSeries из словаря:")
print(s3)
```

### Задание 1.4. Создание Series с использованием константы

```python
# Создание Series с константным значением
s4 = pd.Series(5, index=range(1, 6))
print("\nSeries с константой 5 (индексы 1-5):")
print(s4)

# Создание Series с заданными индексами и константой
s5 = pd.Series(3.14, index=['x', 'y', 'z'])
print("\nSeries с константой 3.14 (индексы x,y,z):")
print(s5)
```

### Задание 1.5. Работа с элементами Series — Обращение по индексу и метке

```python
# Создаем Series для работы
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
print("\nИсходный Series:")
print(s)

# ПРАВИЛЬНЫЙ способ: обращение по числовому индексу (позиции) через .iloc[]
print(f"\nЭлемент по позиции 0 (iloc[0]): {s.iloc[0]}")
print(f"Элемент по позиции 2 (iloc[2]): {s.iloc[2]}")

# Обращение по метке (label) через .loc[]
print(f"Элемент по метке 'a' (loc['a']): {s.loc['a']}")
print(f"Элемент по метке 'd' (loc['d']): {s.loc['d']}")

# Несколько элементов по позициям через .iloc[]
print(f"\nНесколько элементов по позициям [0, 2, 4]:\n{s.iloc[[0, 2, 4]]}")

# Несколько элементов по меткам через .loc[]
print(f"Несколько элементов по меткам ['a', 'c', 'e']:\n{s.loc[['a', 'c', 'e']]}")
```

### Задание 1.6. Получение slice'ов (срезов)

```python
# Срезы в Series
s = pd.Series([10, 20, 30, 40, 50, 60, 70], index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
print("\nИсходный Series:")
print(s)

# Срез по позициям (не включает правую границу)
print("\nСрез по позициям [2:5]:")
print(s[2:5])

# Срез по меткам (включает правую границу!)
print("\nСрез по меткам ['c':'e']:")
print(s['c':'e'])

# Срез с шагом
print("\nСрез с шагом 2 [::2]:")
print(s[::2])

# Обратный срез
print("\nОбратный срез [::-1]:")
print(s[::-1])
```

### Задание 1.7. Условные выражения (работа как с векторами)

```python
# Условные выражения в Series
s = pd.Series([15, 25, 35, 45, 55], index=['a', 'b', 'c', 'd', 'e'])
print("\nИсходный Series:")
print(s)

# Сравнение (возвращает Series с boolean)
print("\ns > 30:")
print(s > 30)

# Фильтрация с использованием условия
print("\nЭлементы > 30:")
print(s[s > 30])

# Сложение с константой (векторизация)
print("\ns + 10:")
print(s + 10)

# Умножение на константу
print("\ns * 2:")
print(s * 2)

# Операции между Series
s2 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print("\ns + s2:")
print(s + s2)

print("\ns * s2:")
print(s * s2)

# Сложные условия
print("\nЭлементы > 20 и < 50:")
print(s[(s > 20) & (s < 50)])

print("\nЭлементы <= 25 или >= 45:")
print(s[(s <= 25) | (s >= 45)])
```

## Часть 3: Работа с элементами DataFrame

### Задание 3.1. Выбор столбца

```python
# Создаем DataFrame для работы
df = pd.DataFrame({
    'Имя': ['Анна', 'Борис', 'Виктор', 'Галина', 'Дмитрий'],
    'Возраст': [25, 30, 28, 24, 27],
    'Зарплата': [50000, 60000, 55000, 48000, 52000],
    'Город': ['Москва', 'СПб', 'Казань', 'Москва', 'СПб']
})
print("Исходный DataFrame:")
print(df)

# Выбор одного столбца (возвращает Series)
print("\nВыбор столбца 'Имя':")
print(df['Имя'])
print(f"Тип: {type(df['Имя'])}")

# Альтернативный способ (через точку)
print("\nВыбор столбца 'Возраст' через точку:")
print(df.Возраст)

# Выбор нескольких столбцов (возвращает DataFrame)
print("\nВыбор нескольких столбцов ['Имя', 'Зарплата']:")
print(df[['Имя', 'Зарплата']])
```

### Задание 3.2. Выбор строки по метке (loc)

```python
# Устанавливаем индексы
df_with_index = df.set_index('Имя')
print("DataFrame с индексами по именам:")
print(df_with_index)

# Выбор строки по метке
print("\nСтрока с меткой 'Виктор':")
print(df_with_index.loc['Виктор'])

# Выбор нескольких строк по меткам
print("\nСтроки с метками ['Анна', 'Галина']:")
print(df_with_index.loc[['Анна', 'Галина']])

# Выбор строк и столбцов одновременно
print("\nСтрока 'Борис', столбцы 'Возраст' и 'Зарплата':")
print(df_with_index.loc['Борис', ['Возраст', 'Зарплата']])
```

### Задание 3.3. Выбор строки по индексу (iloc)

```python
# Выбор строки по числовому индексу (позиции)
print("\nСтрока с индексом 2 (третья строка):")
print(df.iloc[2])

# Выбор нескольких строк по позициям
print("\nСтроки с индексами 0 и 3:")
print(df.iloc[[0, 3]])

# Срез по строкам (с 1 по 3, не включая 4)
print("\nСтроки с 1 по 3 (индексы 1,2,3):")
print(df.iloc[1:4])

# Выбор строк и столбцов по позициям
print("\nСтроки 0-2, столбцы 1-3:")
print(df.iloc[0:3, 1:4])
```

### Задание 3.4. Слайс по строкам

```python
# Различные способы слайсинга строк
print("Исходный DataFrame:")
print(df)

# Слайс с использованием числовых индексов
print("\nСлайс df[1:4]:")
print(df[1:4])

# Слайс с шагом
print("\nСлайс df[::2] (каждая вторая строка):")
print(df[::2])

# Слайс с использованием loc (по меткам индекса)
df_indexed = df.set_index('Имя')
print("\nDataFrame с индексом по именам:")
print(df_indexed)
print("\nСлайс по меткам ['Борис':'Галина']:")
print(df_indexed.loc['Борис':'Галина'])
```

### Задание 3.5. Выбор строки по условию

```python
# Фильтрация строк по условию
print("Исходный DataFrame:")
print(df)

# Условие: возраст > 26
print("\nСотрудники с возрастом > 26:")
print(df[df['Возраст'] > 26])

# Условие: зарплата >= 55000
print("\nСотрудники с зарплатой >= 55000:")
print(df[df['Зарплата'] >= 55000])

# Составное условие (И)
print("\nСотрудники из Москвы с зарплатой > 50000:")
print(df[(df['Город'] == 'Москва') & (df['Зарплата'] > 50000)])

# Составное условие (ИЛИ)
print("\nСотрудники из Москвы ИЛИ с зарплатой > 55000:")
print(df[(df['Город'] == 'Москва') | (df['Зарплата'] > 55000)])

# Использование .query()
print("\nИспользование query(): возраст между 25 и 28:")
print(df.query('25 <= Возраст <= 28'))

# Условие с isin()
cities = ['Москва', 'Казань']
print(f"\nСотрудники из городов {cities}:")
print(df[df['Город'].isin(cities)])
```

### Задание 3.6. Запись датафрейма в файл

```python
# Запись DataFrame в различные форматы

# 1. Запись в CSV
df.to_csv('employees.csv', index=False, encoding='utf-8-sig')
print("\nDataFrame записан в 'employees.csv'")

# 2. Запись в CSV с индексами
df.to_csv('employees_with_index.csv', encoding='utf-8-sig')
print("DataFrame записан в 'employees_with_index.csv' (с индексами)")

# 3. Запись в Excel (требуется openpyxl или xlsxwriter)
try:
    df.to_excel('employees.xlsx', index=False, sheet_name='Сотрудники')
    print("DataFrame записан в 'employees.xlsx'")
except Exception as e:
    print(f"Excel не поддерживается: {e}")

# 4. Запись в JSON
df.to_json('employees.json', orient='records', force_ascii=False, indent=2)
print("DataFrame записан в 'employees.json'")

# 5. Запись в HTML
df.to_html('employees.html', index=False)
print("DataFrame записан в 'employees.html'")

# 6. Запись в Markdown
try:
    print("\nDataFrame в Markdown формате:")
    print(df.to_markdown())
except:
    print("Markdown требует установки tabulate: pip install tabulate")

# 7. Запись в Clipboard (для вставки в Excel/Google Sheets)
df.to_clipboard(index=False, excel=True)
print("\nDataFrame скопирован в буфер обмена (можно вставить в Excel)")

# Проверка созданных файлов
import os
print("\nСозданные файлы:")
for file in ['employees.csv', 'employees_with_index.csv', 'employees.json', 'employees.html']:
    if os.path.exists(file):
        size = os.path.getsize(file)
        print(f"  {file} ({size} байт)")
```
