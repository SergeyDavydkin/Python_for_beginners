```python
import tkinter as tk
import tkinter.ttk as ttk
from tkinter import filedialog as fd
from PIL import ImageTk, Image
import numpy as np
import cv2

class App(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Tkinter PhotoImage Demo')
        self.createleftframe()
        self.createrightframe()

    def createleftframe(self):
        self.leftframe = tk.Frame(self, width=800, height=600, bg='#8D5')
        self.pimage = self.createimage()
        self.img_label = tk.Label(self.leftframe, image=self.pimage, bg='#8D5')
        self.img_label.pack(fill='both', expand=True)
        self.leftframe.pack(side='left', fill='both', expand=True)

    def createimage(self):
        self.image = Image.open('data/cat2.jpg')
        self.image = self.image.resize((800, 600))
        return ImageTk.PhotoImage(self.image)

    def createrightframe(self):
        self.rightframe = tk.Frame(self, width=200, height=600, bg='#1A5')
        
        self.openbutton = ttk.Button(self.rightframe, text=u'Открыть', command=self.insertimage)
        self.openbutton.pack(side='top', expand=False)
        
        self.greybutton = ttk.Button(self.rightframe, text=u'В серый', command=self.grayimage)
        self.greybutton.pack(side='top', expand=False)
        
        blue_label = ttk.Label(self.rightframe, text='Шкала синего')
        blue_label.pack(side=tk.TOP, padx=10, pady=5)
        
        self.scale = ttk.Scale(self.rightframe, from_=0, to=255, command=self.onscaleblue)
        self.scale.set(0)  # Устанавливаем начальное значение
        self.scale.pack(side=tk.TOP, padx=10)
        
        self.rightframe.pack(side='right', fill='y')

    def onscaleblue(self, val):
        v = int(float(val))
        cv2_img = np.array(self.image)
        cv2_img = cv2.cvtColor(cv2_img, cv2.COLOR_RGB2BGR)
        
        # Безопасное добавление синего канала
        cv2_img_int = cv2_img.astype(np.int32)
        cv2_img_int[:, :, 0] = (cv2_img_int[:, :, 0] + v) % 256
        cv2_img = cv2_img_int.astype(np.uint8)
        
        self.image = Image.fromarray(cv2_img)
        self.pimage = ImageTk.PhotoImage(self.image)
        self.img_label.configure(image=self.pimage)
    
    # Альтернативное решение (проще)
    # Если не хотите возиться с типами данных, используйте cv2.add():
    # def onscaleblue(self, val):
    #     v = int(float(val))
    #     cv2_img = np.array(self.image)
    #     cv2_img = cv2.cvtColor(cv2_img, cv2.COLOR_RGB2BGR)
        
    #     # cv2.add автоматически обрезает значения до 0-255
    #     cv2_img[:, :, 0] = cv2.add(cv2_img[:, :, 0], v)
        
    #     self.image = Image.fromarray(cv2_img)
    #     self.pimage = ImageTk.PhotoImage(self.image)
    #     self.img_label.configure(image=self.pimage)

    def insertimage(self):
        filename = fd.askopenfilename()
        if filename:  # Проверка, что файл выбран
            self.image = Image.open(filename)
            self.image = self.image.resize((800, 600))
            self.pimage = ImageTk.PhotoImage(self.image)
            self.img_label.configure(image=self.pimage)
            self.scale.set(0)  # Сбрасываем ползунок

    def grayimage(self):
        if self.greybutton['text'] == u'В серый':
            cv2_img = np.array(self.image)
            cv2_img = cv2.cvtColor(cv2_img, cv2.COLOR_RGB2GRAY)
            # Для grayscale нужно преобразовать обратно в RGB для совместимости
            cv2_img = cv2.cvtColor(cv2_img, cv2.COLOR_GRAY2RGB)
            self.grayimage_pil = Image.fromarray(cv2_img)
            self.pimage = ImageTk.PhotoImage(self.grayimage_pil)
            self.img_label.configure(image=self.pimage)
            self.greybutton['text'] = u'В цветной'
        else:
            self.pimage = ImageTk.PhotoImage(self.image)
            self.img_label.configure(image=self.pimage)
            self.greybutton['text'] = u'В серый'

form = App()
form.mainloop()
```

```python
import tkinter as tk
import sys

def check_solution():
    """
    Проверяет правильность выполнения задания
    Запустите эту функцию после написания кода
    """
    
    print("\n" + "="*50)
    print("🔍 ПРОВЕРКА ВАШЕЙ РАБОТЫ")
    print("="*50 + "\n")
    
    # Создаем окно с вашим кодом
    root = tk.Tk()
    
    # === ЗДЕСЬ ДОЛЖЕН БЫТЬ ВАШ КОД ===
    # Вставьте сюда свой код создания визитки
    # --------------------------------------
    
    root.title("Моя Визитка")
    root.geometry("400x300")
    
    label_name = tk.Label(root, text="Имя: Анна Смирнова")
    label_age = tk.Label(root, text="Возраст: 20 лет")
    label_city = tk.Label(root, text="Город: Санкт-Петербург")
    
    label_name.pack(pady=5)
    label_age.pack(pady=5)
    label_city.pack(pady=5)
    
    # --------------------------------------
    
    # Принудительно обновляем окно для получения реальных размеров
    root.update_idletasks()
    
    # Собираем все метки в список
    all_widgets = root.winfo_children()
    labels = [w for w in all_widgets if isinstance(w, tk.Label)]
    
    # Переменная для подсчета ошибок
    errors = []
    score = 0
    max_score = 6
    
    print("📊 РЕЗУЛЬТАТЫ ПРОВЕРКИ:\n")
    
    # Проверка 1: Заголовок окна
    if root.title() == "Моя Визитка":
        score += 1
        print("  ✅ [1/6] Заголовок окна: 'Моя Визитка'")
    else:
        errors.append(f"Заголовок окна: '{root.title()}' (должно быть 'Моя Визитка')")
        print(f"  ❌ [1/6] Ошибка: {errors[-1]}")
    
    # Проверка 2: Размер окна
    width = root.winfo_width()
    height = root.winfo_height()
    if width == 400 and height == 300:
        score += 1
        print("  ✅ [2/6] Размер окна: 400x300 пикселей")
    else:
        errors.append(f"Размер окна: {width}x{height} (должно быть 400x300)")
        print(f"  ❌ [2/6] Ошибка: {errors[-1]}")
    
    # Проверка 3: Количество меток
    if len(labels) == 3:
        score += 1
        print("  ✅ [3/6] Создано 3 текстовые метки")
    else:
        errors.append(f"Количество меток: {len(labels)} (должно быть 3)")
        print(f"  ❌ [3/6] Ошибка: {errors[-1]}")
    
    # Проверка 4: Текст в метках (содержит ключевые слова)
    required_keywords = ["Имя:", "Возраст:", "Город:"]
    found_keywords = []
    
    for label in labels:
        text = label.cget("text")
        for keyword in required_keywords:
            if keyword in text and keyword not in found_keywords:
                found_keywords.append(keyword)
    
    if len(found_keywords) == 3:
        score += 1
        print("  ✅ [4/6] Все три категории присутствуют (Имя, Возраст, Город)")
    else:
        missing = [k for k in required_keywords if k not in found_keywords]
        errors.append(f"Отсутствуют поля: {', '.join(missing)}")
        print(f"  ❌ [4/6] Ошибка: {errors[-1]}")
    
    # Проверка 5: Использование pack()
    labels_with_pack = 0
    for label in labels:
        try:
            if label.pack_info():
                labels_with_pack += 1
        except:
            pass
    
    if labels_with_pack == 3:
        score += 1
        print("  ✅ [5/6] Все метки размещены с помощью pack()")
    else:
        errors.append(f"Размещено через pack(): {labels_with_pack} из 3 меток")
        print(f"  ❌ [5/6] Ошибка: {errors[-1]}")
    
    # Проверка 6: Отступ pady=5
    labels_with_pady = 0
    for label in labels:
        pack_info = label.pack_info()
        if pack_info and pack_info.get('pady') == 5:
            labels_with_pady += 1
    
    if labels_with_pady == 3:
        score += 1
        print("  ✅ [6/6] У всех меток есть отступ pady=5")
    else:
        errors.append(f"Отступ pady=5 имеют {labels_with_pady} из 3 меток")
        print(f"  ❌ [6/6] Ошибка: {errors[-1]}")
    
    # ИТОГОВЫЙ РЕЗУЛЬТАТ
    print("\n" + "="*50)
    print(f"📈 РЕЗУЛЬТАТ: {score}/{max_score}")
    
    if score == max_score:
        print("🎉 ОТЛИЧНО! Все требования выполнены правильно!")
        print("   Вы успешно освоили создание окна и размещение виджетов!")
    elif score >= 4:
        print("👍 ХОРОШО! Почти всё правильно, но есть небольшие недочеты.")
        print("   Посмотрите на ошибки выше и исправьте их.")
    else:
        print("⚠️ НУЖНО ДОРАБОТАТЬ! Проверьте ошибки и исправьте.")
        print("   Внимательно перечитайте инструкцию.")
    
    print("="*50 + "\n")
    
    # Если есть ошибки, показываем их список
    if errors:
        print("📝 СПИСОК ОШИБОК ДЛЯ ИСПРАВЛЕНИЯ:")
        for i, error in enumerate(errors, 1):
            print(f"  {i}. {error}")
        print("\n💡 Подсказка: Сравните свой код с примером решения.\n")
    else:
        print("✨ Программа будет закрыта через 3 секунды...")
    
    # Закрываем окно через некоторое время
    root.after(3000 if score == max_score else 5000, root.destroy)
    root.mainloop()
    
    return score == max_score

# Запуск проверки
if __name__ == "__main__":
    print("\n👋 Добро пожаловать в проверочную систему!")
    print("Сейчас будет создано окно с вашей визиткой...")
    input("Нажмите Enter, чтобы начать проверку...")
    
    success = check_solution()
    
    if success:
        print("🏆 Поздравляю! Задание принято!")
    else:
        print("🔄 Исправьте ошибки и запустите проверку снова.")
```

```python
import tkinter as tk

def create_and_check():
    """Создает визитку и проверяет все параметры"""
    
    # СОЗДАЕМ ВИЗИТКУ
    root = tk.Tk()
    root.title("Моя Визитка")
    root.geometry("400x300")
    
    label_name = tk.Label(root, text="Имя: Иван Петров")
    label_age = tk.Label(root, text="Возраст: 25 лет")
    label_city = tk.Label(root, text="Город: Москва")
    
    label_name.pack(pady=5)
    label_age.pack(pady=5)
    label_city.pack(pady=5)
    
    # ПРОВЕРЯЕМ (исправленная версия)
    errors = []
    
    # Проверка 1: Заголовок окна
    if root.title() != "Моя Визитка":
        errors.append(f"❌ Заголовок окна: '{root.title()}' (должно быть 'Моя Визитка')")
    else:
        print("✓ Заголовок окна верный")
    
    # Проверка 2: Размер окна (исправлено!)
    root.update_idletasks()  # Важно! Применяем все настройки к окну
    window_width = root.winfo_width()
    window_height = root.winfo_height()
    
    if window_width != 400 or window_height != 300:
        errors.append(f"❌ Размер окна: {window_width}x{window_height} (должно быть 400x300)")
    else:
        print("✓ Размер окна верный (400x300)")
    
    # Проверка 3: Количество меток
    labels = [label_name, label_age, label_city]
    if len(labels) != 3:
        errors.append(f"❌ Количество меток: {len(labels)} (должно быть 3)")
    else:
        print("✓ Создано 3 метки")
    
    # Проверка 4: Текст меток
    expected_texts = ["Имя: Иван Петров", "Возраст: 25 лет", "Город: Москва"]
    for i, (label, expected) in enumerate(zip(labels, expected_texts), 1):
        if label.cget("text") != expected:
            errors.append(f"❌ Метка {i}: '{label.cget('text')}' (должно быть '{expected}')")
        else:
            print(f"✓ Метка {i}: '{expected}'")
    
    # Проверка 5: Отступы
    for i, label in enumerate(labels, 1):
        pack_info = label.pack_info()
        if pack_info.get('pady') != 5:
            errors.append(f"❌ Метка {i}: отступ pady={pack_info.get('pady')} (должно быть 5)")
        else:
            print(f"✓ Метка {i}: отступ pady=5")
    
    # Проверка 6: Использование pack()
    for i, label in enumerate(labels, 1):
        if not label.pack_info():
            errors.append(f"❌ Метка {i}: не используется pack()")
        else:
            print(f"✓ Метка {i}: используется pack()")
    
    # РЕЗУЛЬТАТ
    print("\n" + "="*40)
    if errors:
        print("❌ НАЙДЕНЫ ОШИБКИ:")
        for error in errors:
            print(f"  {error}")
    else:
        print("✅ ВСЕ ТЕСТЫ ПРОЙДЕНЫ УСПЕШНО!")
    
    print("="*40)
    
    # Закрываем окно через 3 секунды
    root.after(3000, root.destroy)
    root.mainloop()
    
    return len(errors) == 0

# Запускаем проверку
if __name__ == "__main__":
    success = create_and_check()
    print(f"\nРезультат: {'✅ ПРОЙДЕН' if success else '❌ НЕ ПРОЙДЕН'}")
```

```python
import tkinter as tk

class Calculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Калькулятор")
        self.root.geometry("500x400")
        self.root.configure(bg="#f0f0f0")
        
        self.create_widgets()
    
    def create_widgets(self):
        # Заголовок
        title = tk.Label(self.root, text="КАЛЬКУЛЯТОР", 
                        font=("Arial", 18, "bold"),
                        bg="#f0f0f0", fg="#333")
        title.grid(row=0, column=0, columnspan=4, pady=10)
        
        # Поля ввода
        tk.Label(self.root, text="Первое число:", font=("Arial", 12),
                bg="#f0f0f0").grid(row=1, column=0, padx=10, pady=10, sticky="e")
        self.entry1 = tk.Entry(self.root, font=("Arial", 14), width=15,
                               justify="right", relief="ridge")
        self.entry1.grid(row=1, column=1, padx=10, pady=10)
        
        tk.Label(self.root, text="Второе число:", font=("Arial", 12),
                bg="#f0f0f0").grid(row=2, column=0, padx=10, pady=10, sticky="e")
        self.entry2 = tk.Entry(self.root, font=("Arial", 14), width=15,
                               justify="right", relief="ridge")
        self.entry2.grid(row=2, column=1, padx=10, pady=10)
        
        # Кнопки операций
        buttons = [
            ("+", 3, 0, "#4CAF50"),
            ("-", 3, 1, "#2196F3"),
            ("×", 3, 2, "#FF9800"),
            ("÷", 3, 3, "#f44336")
        ]
        
        for text, row, col, color in buttons:
            btn = tk.Button(self.root, text=text, font=("Arial", 16, "bold"),
                          bg=color, fg="white", width=5, height=1,
                          command=lambda op=text: self.calculate(op))
            btn.grid(row=row, column=col, padx=5, pady=10)
        
        # Кнопка очистки
        clear_btn = tk.Button(self.root, text="Очистить", font=("Arial", 12),
                             bg="#999", fg="white", command=self.clear)
        clear_btn.grid(row=4, column=0, columnspan=4, pady=5, sticky="we", padx=50)
        
        # Метка результата
        self.result_label = tk.Label(self.root, text="Результат: ---", 
                                     font=("Arial", 14, "bold"),
                                     bg="white", fg="#333",
                                     relief="sunken", height=2)
        self.result_label.grid(row=5, column=0, columnspan=4, padx=10, pady=20, sticky="we")
        
        # История вычислений
        self.history_text = tk.Text(self.root, height=5, width=40, font=("Arial", 10))
        self.history_text.grid(row=6, column=0, columnspan=4, padx=10, pady=5)
        
        tk.Label(self.root, text="История:", font=("Arial", 10),
                bg="#f0f0f0").grid(row=7, column=0, columnspan=4)
    
    def calculate(self, operation):
        try:
            num1 = float(self.entry1.get())
            num2 = float(self.entry2.get())
            
            if operation == '+':
                result = num1 + num2
                op_symbol = '+'
            elif operation == '-':
                result = num1 - num2
                op_symbol = '-'
            elif operation == '×':
                result = num1 * num2
                op_symbol = '*'
            elif operation == '÷':
                if num2 == 0:
                    raise ZeroDivisionError
                result = num1 / num2
                op_symbol = '/'
            
            # Форматируем результат
            if result == int(result):
                result = int(result)
            else:
                result = round(result, 6)
            
            self.result_label.config(text=f"Результат: {result}", fg="green")
            
            # Добавляем в историю
            history_entry = f"{num1} {op_symbol} {num2} = {result}\n"
            self.history_text.insert("1.0", history_entry)
            
            # Ограничиваем историю 10 записями
            if len(self.history_text.get("1.0", "end-1c").split('\n')) > 10:
                self.history_text.delete("end-2l", "end-1c")
            
        except ValueError:
            self.result_label.config(text="Ошибка: введите числа!", fg="red")
        except ZeroDivisionError:
            self.result_label.config(text="Ошибка: деление на ноль!", fg="red")
        except Exception as e:
            self.result_label.config(text=f"Ошибка: {str(e)}", fg="red")
    
    def clear(self):
        """Очищает поля ввода и результат"""
        self.entry1.delete(0, tk.END)
        self.entry2.delete(0, tk.END)
        self.result_label.config(text="Результат: ---", fg="#333")

if __name__ == "__main__":
    root = tk.Tk()
    app = Calculator(root)
    root.mainloop()
```

```python
import tkinter as tk

class ClickCounter:
    def __init__(self, root):
        self.root = root
        self.root.title("Счетчик кликов")
        self.root.geometry("350x300")
        self.root.configure(bg="#f0f0f0")
        
        self.counter = 0
        self.step = 1  # Шаг изменения
        
        self.create_widgets()
    
    def create_widgets(self):
        # Заголовок
        title = tk.Label(self.root, text="СЧЕТЧИК КЛИКОВ", 
                        font=("Arial", 16, "bold"),
                        bg="#f0f0f0", fg="#333")
        title.pack(pady=10)
        
        # Основной счетчик
        self.label_counter = tk.Label(self.root, text=str(self.counter), 
                                     font=("Arial", 64, "bold"),
                                     bg="white", fg="#2196F3",
                                     width=8, relief="ridge")
        self.label_counter.pack(pady=20)
        
        # Фрейм для кнопок
        button_frame = tk.Frame(self.root, bg="#f0f0f0")
        button_frame.pack(pady=10)
        
        # Кнопка увеличения
        self.button_plus = tk.Button(button_frame, text="+1", 
                                    font=("Arial", 18, "bold"),
                                    bg="#4CAF50", fg="white",
                                    width=6, command=self.increment)
        self.button_plus.pack(side=tk.LEFT, padx=10)
        
        # Кнопка уменьшения
        self.button_minus = tk.Button(button_frame, text="-1", 
                                     font=("Arial", 18, "bold"),
                                     bg="#f44336", fg="white",
                                     width=6, command=self.decrement)
        self.button_minus.pack(side=tk.LEFT, padx=10)
        
        # Начальное состояние
        self.button_minus.config(state=tk.DISABLED)
        
        # Дополнительные кнопки шага
        step_frame = tk.Frame(self.root, bg="#f0f0f0")
        step_frame.pack(pady=10)
        
        tk.Label(step_frame, text="Шаг:", font=("Arial", 12),
                bg="#f0f0f0").pack(side=tk.LEFT, padx=5)
        
        self.step_var = tk.StringVar(value="1")
        step_spinbox = tk.Spinbox(step_frame, from_=1, to=10, 
                                 textvariable=self.step_var,
                                 width=5, font=("Arial", 12),
                                 command=self.update_step)
        step_spinbox.pack(side=tk.LEFT, padx=5)
        
        # Кнопка сброса
        reset_button = tk.Button(self.root, text="Сброс", 
                                font=("Arial", 12),
                                bg="#FF9800", fg="white",
                                command=self.reset)
        reset_button.pack(pady=10)
        
        # Информационная метка
        self.info_label = tk.Label(self.root, text="Счетчик не может быть отрицательным",
                                  font=("Arial", 10), fg="gray",
                                  bg="#f0f0f0")
        self.info_label.pack(pady=5)
    
    def update_step(self):
        """Обновляет шаг изменения"""
        self.step = int(self.step_var.get())
    
    def increment(self):
        self.counter += self.step
        self.label_counter.config(text=str(self.counter))
        
        # Анимируем изменение цвета
        self.label_counter.config(fg="#4CAF50")
        self.root.after(200, lambda: self.label_counter.config(fg="#2196F3"))
        
        # Активируем кнопку "-1"
        self.button_minus.config(state=tk.NORMAL)
    
    def decrement(self):
        if self.counter - self.step >= 0:
            self.counter -= self.step
            self.label_counter.config(text=str(self.counter))
            
            # Анимируем изменение цвета
            self.label_counter.config(fg="#f44336")
            self.root.after(200, lambda: self.label_counter.config(fg="#2196F3"))
            
            # Если достигли 0, отключаем кнопку "-1"
            if self.counter == 0:
                self.button_minus.config(state=tk.DISABLED)
        else:
            # Визуальное предупреждение
            self.info_label.config(text="Нельзя уменьшить! Счетчик не может быть отрицательным.",
                                  fg="red")
            self.root.after(1500, lambda: self.info_label.config(
                text="Счетчик не может быть отрицательным", fg="gray"))
    
    def reset(self):
        self.counter = 0
        self.label_counter.config(text=str(self.counter))
        self.button_minus.config(state=tk.DISABLED)
        
        # Мигаем счетчиком
        self.label_counter.config(fg="#FF9800")
        self.root.after(500, lambda: self.label_counter.config(fg="#2196F3"))

if __name__ == "__main__":
    root = tk.Tk()
    app = ClickCounter(root)
    root.mainloop()
```

```python
import tkinter as tk
import sys

def check_solution():
    """Автоматическая проверка задания 5: Счетчик кликов"""
    
    print("\n" + "="*60)
    print("🔍 ПРОВЕРКА ЗАДАНИЯ 5: Счетчик кликов")
    print("="*60 + "\n")
    
    # === ВСТАВЬТЕ ВАШ КОД ЗДЕСЬ ===
    # --------------------------------------
    
    root = tk.Tk()
    root.title("Счетчик кликов")
    root.geometry("300x200")
    
    counter = 0
    
    label_counter = tk.Label(root, text=str(counter), font=("Arial", 48))
    
    def increment():
        global counter
        counter += 1
        label_counter.config(text=str(counter))
        button_minus.config(state=tk.NORMAL)
    
    def decrement():
        global counter
        if counter > 0:
            counter -= 1
            label_counter.config(text=str(counter))
            if counter == 0:
                button_minus.config(state=tk.DISABLED)
    
    button_plus = tk.Button(root, text="+1", command=increment)
    button_minus = tk.Button(root, text="-1", command=decrement)
    
    # ★ ВАЖНО: начальное состояние кнопки "-1" - отключена ★
    button_minus.config(state=tk.DISABLED)
    
    label_counter.pack(pady=20)
    button_plus.pack(pady=5)
    button_minus.pack(pady=5)
    
    # Принудительно обновляем окно для проверки размеров
    root.update_idletasks()
    
    # --------------------------------------
    
    errors = []
    score = 0
    max_score = 7
    
    print("📊 РЕЗУЛЬТАТЫ ПРОВЕРКИ:\n")
    
    # Проверка 1: Заголовок окна
    if "счетчик" in root.title().lower():
        score += 1
        print("  ✅ [1/7] Заголовок окна содержит 'счетчик'")
    else:
        errors.append(f"Заголовок: '{root.title()}' (должен содержать 'счетчик')")
        print(f"  ❌ [1/7] Ошибка: {errors[-1]}")
    
    # Проверка 2: Наличие метки
    labels = [w for w in root.winfo_children() if isinstance(w, tk.Label)]
    if labels:
        score += 1
        print("  ✅ [2/7] Метка для счетчика присутствует")
    else:
        errors.append("Нет метки для отображения числа")
        print(f"  ❌ [2/7] Ошибка: {errors[-1]}")
    
    # Проверка 3: Наличие двух кнопок
    buttons = [w for w in root.winfo_children() if isinstance(w, tk.Button)]
    if len(buttons) >= 2:
        score += 1
        print("  ✅ [3/7] Две кнопки присутствуют")
    else:
        errors.append(f"Найдено {len(buttons)} кнопок (нужно 2)")
        print(f"  ❌ [3/7] Ошибка: {errors[-1]}")
    
    # Проверка 4: Кнопки имеют правильные тексты
    button_texts = [b.cget("text") for b in buttons]
    has_plus = any("+1" in text or "1+" in text for text in button_texts)
    has_minus = any("-1" in text or "1-" in text for text in button_texts)
    
    if has_plus and has_minus:
        score += 1
        print("  ✅ [4/7] Кнопки '+1' и '-1' присутствуют")
    else:
        if not has_plus:
            errors.append("Нет кнопки '+1'")
        if not has_minus:
            errors.append("Нет кнопки '-1'")
        print(f"  ❌ [4/7] Ошибка: {', '.join(errors[-2:]) if len(errors) >= 2 else errors[-1]}")
    
    # Проверка 5: Работа кнопки "+1" (увеличивает ровно на 1)
    try:
        if buttons and labels:
            # Получаем начальное значение
            initial_text = labels[0].cget("text")
            try:
                initial_value = int(initial_text) if str(initial_text).isdigit() else 0
            except:
                initial_value = 0
            
            # Находим кнопку "+1"
            plus_button = None
            for btn in buttons:
                btn_text = btn.cget("text")
                if "+1" in btn_text or "1+" in btn_text:
                    plus_button = btn
                    break
            
            if plus_button:
                # Эмулируем нажатие
                plus_button.invoke()
                root.update()
                
                new_text = labels[0].cget("text")
                try:
                    new_value = int(new_text) if str(new_text).isdigit() else 0
                    
                    if new_value == initial_value + 1:
                        score += 1
                        print("  ✅ [5/7] Кнопка '+1' увеличивает счетчик на 1")
                    else:
                        errors.append(f"После +1: {initial_value} -> {new_value} (должно быть +1)")
                        print(f"  ❌ [5/7] Ошибка: {errors[-1]}")
                except:
                    errors.append(f"Метка содержит не число: '{new_text}'")
                    print(f"  ❌ [5/7] Ошибка: {errors[-1]}")
            else:
                errors.append("Кнопка '+1' не найдена")
                print(f"  ❌ [5/7] Ошибка: {errors[-1]}")
    except Exception as e:
        errors.append(f"Ошибка при проверке '+1': {e}")
        print(f"  ❌ [5/7] Ошибка: {errors[-1]}")
    
    # Проверка 6: Кнопка "-1" неактивна при счетчике = 0
    try:
        if buttons:
            # Находим кнопку "-1"
            minus_button = None
            for btn in buttons:
                btn_text = btn.cget("text")
                if "-1" in btn_text or "1-" in btn_text:
                    minus_button = btn
                    break
            
            if minus_button:
                # Сначала убеждаемся, что счетчик = 0
                # Для этого несколько раз нажимаем "-1" если нужно
                current_label = labels[0] if labels else None
                if current_label:
                    # Сбрасываем до 0 (максимум 10 нажатий)
                    for _ in range(10):
                        current_text = current_label.cget("text")
                        try:
                            if int(current_text) <= 0:
                                break
                        except:
                            break
                        if minus_button.cget("state") == tk.NORMAL:
                            minus_button.invoke()
                            root.update()
                
                # Проверяем состояние кнопки
                if minus_button.cget("state") == tk.DISABLED:
                    score += 1
                    print("  ✅ [6/7] Кнопка '-1' неактивна (DISABLED) при счетчике = 0")
                else:
                    errors.append("Кнопка '-1' активна при счетчике = 0 (должна быть отключена)")
                    print(f"  ❌ [6/7] Ошибка: {errors[-1]}")
            else:
                errors.append("Кнопка '-1' не найдена")
                print(f"  ❌ [6/7] Ошибка: {errors[-1]}")
    except Exception as e:
        errors.append(f"Ошибка при проверке состояния кнопки: {e}")
        print(f"  ❌ [6/7] Ошибка: {errors[-1]}")
    
    # Проверка 7: Кнопка "-1" становится активной после увеличения
    try:
        if buttons and labels:
            # Находим кнопки
            plus_button = None
            minus_button = None
            
            for btn in buttons:
                btn_text = btn.cget("text")
                if "+1" in btn_text or "1+" in btn_text:
                    plus_button = btn
                if "-1" in btn_text or "1-" in btn_text:
                    minus_button = btn
            
            if plus_button and minus_button:
                # Сначала убеждаемся, что счетчик = 0 и кнопка отключена
                # Нажимаем "+1"
                plus_button.invoke()
                root.update()
                
                # Проверяем, стала ли кнопка "-1" активной
                if minus_button.cget("state") == tk.NORMAL:
                    score += 1
                    print("  ✅ [7/7] Кнопка '-1' активируется после увеличения счетчика")
                else:
                    errors.append("Кнопка '-1' остается неактивной после увеличения счетчика")
                    print(f"  ❌ [7/7] Ошибка: {errors[-1]}")
            else:
                errors.append("Кнопки '+1' или '-1' не найдены")
                print(f"  ❌ [7/7] Ошибка: {errors[-1]}")
    except Exception as e:
        errors.append(f"Ошибка при проверке активации кнопки: {e}")
        print(f"  ❌ [7/7] Ошибка: {errors[-1]}")
    
    # ИТОГОВЫЙ РЕЗУЛЬТАТ
    print("\n" + "="*60)
    print(f"📈 РЕЗУЛЬТАТ: {score}/{max_score}")
    
    if score == max_score:
        print("🎉 ОТЛИЧНО! Все тесты пройдены!")
        print("   Вы научились использовать .config() и менять состояние кнопок!")
    elif score >= 5:
        print("👍 ХОРОШО! Почти всё правильно, но есть небольшие недочеты.")
    elif score >= 3:
        print("⚠️ НУЖНО ДОРАБОТАТЬ! Проверьте ошибки выше.")
    else:
        print("❌ ТРЕБУЕТСЯ ПЕРЕДЕЛКА! Внимательно изучите инструкцию.")
    
    print("="*60 + "\n")
    
    if errors:
        print("📝 СПИСОК ОШИБОК ДЛЯ ИСПРАВЛЕНИЯ:")
        for i, error in enumerate(errors, 1):
            print(f"  {i}. {error}")
        print("\n💡 ПОДСКАЗКИ:")
        print("  • Используйте .config(state=tk.DISABLED) и .config(state=tk.NORMAL)")
        print("  • Проверяйте counter > 0 перед уменьшением")
        print("  • Не забывайте обновлять текст метки через .config(text=str(counter))")
        print("  • Для изменения состояния кнопки: button.config(state=tk.DISABLED)")
        print("  • ★ ВАЖНО: При запуске кнопка '-1' должна быть DISABLED ★\n")
    else:
        print("✨ Программа работает правильно! Окно закроется через 3 секунды...")
        root.after(3000, root.destroy)
    
    root.mainloop()
    return score == max_score

if __name__ == "__main__":
    print("\n👋 Добро пожаловать в проверочную систему!")
    print("Сейчас будет проверен ваш счетчик кликов...")
    print("Убедитесь, что ваш код вставлен в функцию check_solution()")
    input("\nНажмите Enter, чтобы начать проверку...")
    
    success = check_solution()
    
    if success:
        print("\n🏆 ПОЗДРАВЛЯЮ! Счетчик работает идеально!")
        print("   Вы успешно освоили:")
        print("   • Изменение свойств виджетов через .config()")
        print("   • Управление состоянием кнопок (NORMAL/DISABLED)")
        print("   • Работу с глобальными переменными")
        print("   • Обработку событий кнопок")
    else:
        print("\n🔄 Исправьте ошибки и запустите проверку снова.")
        print("   Внимательно прочитайте подсказки выше.")
```

```python
import tkinter as tk

class LanguageSelector:
    def __init__(self, root):
        self.root = root
        self.root.title("Выбор языка программирования")
        self.root.geometry("500x400")
        self.root.configure(bg="#f0f0f0")
        
        # Словарь с примерами кода
        self.code_examples = {
            "Python": {
                "code": 'print("Hello World")',
                "color": "#3776AB",
                "description": "Простой и элегантный синтаксис"
            },
            "Java": {
                "code": 'System.out.println("Hello World");',
                "color": "#007396",
                "description": "Строгая типизация, JVM"
            },
            "C++": {
                "code": 'std::cout << "Hello World" << std::endl;',
                "color": "#00599C",
                "description": "Высокая производительность"
            }
        }
        
        self.create_widgets()
    
    def create_widgets(self):
        # Заголовок
        title = tk.Label(self.root, text="💻 Выбор языка программирования 💻",
                        font=("Arial", 16, "bold"),
                        bg="#f0f0f0", fg="#333")
        title.pack(pady=15)
        
        # Переменная для хранения выбора
        self.selected_lang = tk.StringVar()
        self.selected_lang.set("Python")
        
        # Рамка для радиокнопок
        frame_radios = tk.Frame(self.root, bg="#f0f0f0")
        frame_radios.pack(pady=10)
        
        # Создаем радиокнопки
        self.radios = {}
        for lang in ["Python", "Java", "C++"]:
            radio = tk.Radiobutton(frame_radios, 
                                   text=lang,
                                   variable=self.selected_lang,
                                   value=lang,
                                   font=("Arial", 12, "bold"),
                                   bg="#f0f0f0",
                                   fg=self.code_examples[lang]["color"],
                                   selectcolor="#f0f0f0",
                                   command=self.update_description)
            radio.pack(anchor="w", pady=5)
            self.radios[lang] = radio
        
        # Метка для описания языка
        self.desc_label = tk.Label(self.root, text="",
                                   font=("Arial", 10, "italic"),
                                   bg="#f0f0f0", fg="#666")
        self.desc_label.pack(pady=5)
        
        # Кнопка показа кода
        self.show_button = tk.Button(self.root, 
                                     text="🔍 Показать код",
                                     font=("Arial", 12, "bold"),
                                     bg="#4CAF50", fg="white",
                                     padx=20, pady=5,
                                     command=self.show_code)
        self.show_button.pack(pady=10)
        
        # Рамка для вывода кода
        self.code_frame = tk.Frame(self.root, bg="#2d2d2d", relief="ridge", bd=2)
        self.code_frame.pack(pady=10, padx=20, fill="both", expand=True)
        
        # Метка для вывода кода (с моноширинным шрифтом)
        self.code_label = tk.Label(self.code_frame, 
                                   text="",
                                   font=("Consolas", 11),
                                   bg="#2d2d2d",
                                   fg="#f8f8f2",
                                   wraplength=450,
                                   justify="left",
                                   padx=15,
                                   pady=15)
        self.code_label.pack()
        
        # Кнопка копирования
        self.copy_button = tk.Button(self.root,
                                     text="📋 Копировать в буфер",
                                     font=("Arial", 10),
                                     bg="#FF9800", fg="white",
                                     command=self.copy_to_clipboard)
        self.copy_button.pack(pady=5)
        
        # Статусная строка
        self.status_label = tk.Label(self.root, text="",
                                     font=("Arial", 9),
                                     bg="#f0f0f0", fg="gray")
        self.status_label.pack(pady=5)
        
        # Обновляем описание для начального выбора
        self.update_description()
    
    def update_description(self):
        """Обновляет описание языка"""
        lang = self.selected_lang.get()
        if lang in self.code_examples:
            self.desc_label.config(text=self.code_examples[lang]["description"])
            for l, radio in self.radios.items():
                if l == lang:
                    radio.config(fg=self.code_examples[l]["color"])
                else:
                    radio.config(fg="black")
    
    def show_code(self):
        """Показывает код на выбранном языке"""
        language = self.selected_lang.get()
        
        if language in self.code_examples:
            code = self.code_examples[language]["code"]
            self.code_label.config(text=code)
            self.status_label.config(text=f"Показан код на {language}", fg="green")
            
            # Анимация: мигание фрейма
            self.code_frame.config(bg="#4CAF50")
            self.root.after(200, lambda: self.code_frame.config(bg="#2d2d2d"))
        else:
            self.code_label.config(text="❌ Ошибка: язык не выбран")
            self.status_label.config(text="Пожалуйста, выберите язык", fg="red")
    
    def copy_to_clipboard(self):
        """Копирует код в буфер обмена"""
        code = self.code_label.cget("text")
        if code and code != "❌ Ошибка: язык не выбран":
            self.root.clipboard_clear()
            self.root.clipboard_append(code)
            self.status_label.config(text="✅ Код скопирован в буфер обмена!", fg="green")
            self.root.after(2000, lambda: self.status_label.config(text="", fg="gray"))
        else:
            self.status_label.config(text="❌ Нет кода для копирования", fg="red")

# Запуск приложения
if __name__ == "__main__":
    root = tk.Tk()
    app = LanguageSelector(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class ShoppingList:
    def __init__(self, root):
        self.root = root
        self.root.title("Список покупок")
        self.root.geometry("500x500")
        self.root.configure(bg="#f0f0f0")
        
        self.items = []  # Для хранения элементов (резервное копирование)
        self.create_widgets()
        
    def create_widgets(self):
        # Заголовок
        title = tk.Label(self.root, text="🛒 СПИСОК ПОКУПОК 🛒",
                        font=("Arial", 18, "bold"),
                        bg="#f0f0f0", fg="#333")
        title.pack(pady=10)
        
        # Статистика
        self.stats_label = tk.Label(self.root, text="Всего: 0 | Куплено: 0",
                                    font=("Arial", 10),
                                    bg="#f0f0f0", fg="#666")
        self.stats_label.pack()
        
        # Фрейм для ввода
        input_frame = tk.LabelFrame(self.root, text="Новый товар",
                                    font=("Arial", 10, "bold"),
                                    bg="#f0f0f0", padx=10, pady=10)
        input_frame.pack(pady=10, padx=20, fill="x")
        
        self.entry_item = tk.Entry(input_frame, font=("Arial", 12), width=30)
        self.entry_item.pack(side=tk.LEFT, padx=5, expand=True, fill="x")
        
        btn_add = tk.Button(input_frame, text="Добавить", font=("Arial", 10, "bold"),
                           bg="#4CAF50", fg="white", command=self.add_item)
        btn_add.pack(side=tk.LEFT, padx=5)
        
        # Фрейм для управления списком
        control_frame = tk.Frame(self.root, bg="#f0f0f0")
        control_frame.pack(pady=5)
        
        btn_delete = tk.Button(control_frame, text="Удалить", font=("Arial", 10),
                              bg="#f44336", fg="white", command=self.delete_item)
        btn_delete.pack(side=tk.LEFT, padx=5)
        
        btn_mark_done = tk.Button(control_frame, text="✓ Отметить купленное", font=("Arial", 10),
                                  bg="#2196F3", fg="white", command=self.mark_done)
        btn_mark_done.pack(side=tk.LEFT, padx=5)
        
        btn_clear_all = tk.Button(control_frame, text="Очистить всё", font=("Arial", 10),
                                  bg="#FF9800", fg="white", command=self.clear_all)
        btn_clear_all.pack(side=tk.LEFT, padx=5)
        
        # Список с прокруткой
        list_frame = tk.Frame(self.root)
        list_frame.pack(pady=10, fill="both", expand=True, padx=20)
        
        self.listbox = tk.Listbox(list_frame, font=("Arial", 11),
                                 selectmode=tk.SINGLE, height=12,
                                 bg="white", fg="#333", selectbackground="#2196F3")
        self.listbox.pack(side=tk.LEFT, fill="both", expand=True)
        
        scrollbar = tk.Scrollbar(list_frame, orient=tk.VERTICAL)
        scrollbar.pack(side=tk.RIGHT, fill="y")
        self.listbox.config(yscrollcommand=scrollbar.set)
        scrollbar.config(command=self.listbox.yview)
        
        # Двойной клик для удаления
        self.listbox.bind("<Double-Button-1>", lambda e: self.delete_item())
        
        # Клавиша Delete для удаления
        self.root.bind("<Delete>", lambda e: self.delete_item())
        
        # Кнопка для сохранения в файл
        btn_save = tk.Button(self.root, text="💾 Сохранить в файл", font=("Arial", 10),
                            bg="#9C27B0", fg="white", command=self.save_to_file)
        btn_save.pack(pady=5)
        
        # Кнопка для загрузки из файла
        btn_load = tk.Button(self.root, text="📂 Загрузить из файла", font=("Arial", 10),
                            bg="#9C27B0", fg="white", command=self.load_from_file)
        btn_load.pack(pady=5)
        
        # Информация
        info = tk.Label(self.root, text="💡 Советы: Enter - добавить | Delete - удалить | Двойной клик - удалить",
                       font=("Arial", 8), bg="#f0f0f0", fg="#888")
        info.pack(pady=5)
        
        # Привязка Enter
        self.entry_item.bind("<Return>", lambda e: self.add_item())
        
        # Добавляем примеры
        self.add_sample_items()
        
        # Обновляем статистику
        self.update_stats()
    
    def add_item(self):
        """Добавляет элемент в список"""
        item = self.entry_item.get().strip()
        if item:
            self.listbox.insert(tk.END, f"• {item}")
            self.entry_item.delete(0, tk.END)
            self.update_stats()
        else:
            self.show_temp_message("Введите текст!", self.entry_item)
    
    def delete_item(self):
        """Удаляет выбранный элемент"""
        selected = self.listbox.curselection()
        if selected:
            item = self.listbox.get(selected[0])
            if messagebox.askyesno("Подтверждение", f'Удалить "{item}"?'):
                self.listbox.delete(selected[0])
                self.update_stats()
    
    def mark_done(self):
        """Отмечает выбранный элемент как купленный"""
        selected = self.listbox.curselection()
        if selected:
            item = self.listbox.get(selected[0])
            if not item.startswith("✅"):
                self.listbox.delete(selected[0])
                self.listbox.insert(selected[0], f"✅ {item[2:]}")
                self.listbox.itemconfig(selected[0], fg="gray")
                self.update_stats()
    
    def clear_all(self):
        """Очищает весь список"""
        if messagebox.askyesno("Подтверждение", "Очистить весь список?"):
            self.listbox.delete(0, tk.END)
            self.update_stats()
    
    def update_stats(self):
        """Обновляет статистику"""
        total = self.listbox.size()
        
        # Считаем отмеченные элементы (начинаются с ✅)
        marked = 0
        for i in range(total):
            if self.listbox.get(i).startswith("✅"):
                marked += 1
        
        self.stats_label.config(text=f"📊 Всего: {total}  |  ✅ Куплено: {marked}")
    
    def save_to_file(self):
        """Сохраняет список в файл"""
        try:
            with open("shopping_list.txt", "w", encoding="utf-8") as f:
                for i in range(self.listbox.size()):
                    item = self.listbox.get(i)
                    f.write(item + "\n")
            self.show_temp_message("✅ Сохранено в 'shopping_list.txt'", self.stats_label, is_error=False)
        except Exception as e:
            messagebox.showerror("Ошибка", f"Не удалось сохранить: {e}")
    
    def load_from_file(self):
        """Загружает список из файла"""
        try:
            with open("shopping_list.txt", "r", encoding="utf-8") as f:
                items = [line.strip() for line in f if line.strip()]
            
            if items:
                self.listbox.delete(0, tk.END)
                for item in items:
                    self.listbox.insert(tk.END, item)
                self.update_stats()
                self.show_temp_message("✅ Загружено из 'shopping_list.txt'", self.stats_label, is_error=False)
        except FileNotFoundError:
            self.show_temp_message("❌ Файл 'shopping_list.txt' не найден", self.stats_label)
        except Exception as e:
            messagebox.showerror("Ошибка", f"Не удалось загрузить: {e}")
    
    def add_sample_items(self):
        """Добавляет примеры товаров"""
        samples = ["🥛 Молоко", "🍞 Хлеб", "🥚 Яйца", "🧀 Сыр", "🍎 Яблоки", "🥕 Морковь"]
        for item in samples:
            self.listbox.insert(tk.END, f"• {item}")
    
    def show_temp_message(self, message, widget, is_error=True):
        """Показывает временное сообщение"""
        original_text = widget.cget("text") if isinstance(widget, tk.Label) else None
        original_bg = widget.cget("bg")
        
        if is_error:
            widget.config(bg="#ffcccc")
        elif isinstance(widget, tk.Label):
            widget.config(text=message, fg="green" if not is_error else "red")
        
        if isinstance(widget, tk.Entry):
            widget.insert(0, message)
            widget.config(fg="red")
            self.root.after(1500, lambda: [widget.delete(0, tk.END), widget.config(fg="black")])
        elif isinstance(widget, tk.Label) and original_text is not None:
            self.root.after(1500, lambda: widget.config(text=original_text, fg="#666"))
        elif isinstance(widget, tk.Entry):
            self.root.after(1500, lambda: widget.config(bg=original_bg))

# Запуск приложения
if __name__ == "__main__":
    root = tk.Tk()
    app = ShoppingList(root)
    root.mainloop()
```

```python
import tkinter as tk
from datetime import datetime

class DigitalClock:
    def __init__(self, root):
        self.root = root
        self.root.title("Цифровые часы")
        self.root.geometry("500x300")
        self.root.configure(bg="#1a1a2e")
        
        # Флаг для 24-часового формата (True) или 12-часового (False)
        self.is_24h = True
        
        self.create_widgets()
        self.update_time()
    
    def create_widgets(self):
        # Заголовок
        title_label = tk.Label(
            self.root,
            text="⏰ ЦИФРОВЫЕ ЧАСЫ ⏰",
            font=("Arial", 14, "bold"),
            bg="#1a1a2e",
            fg="#e0e0e0"
        )
        title_label.pack(pady=10)
        
        # Основные часы
        self.time_label = tk.Label(
            self.root,
            font=("DS-Digital", 60, "bold"),
            bg="#1a1a2e",
            fg="#00ffcc"
        )
        self.time_label.pack(expand=True)
        
        # Дата
        self.date_label = tk.Label(
            self.root,
            font=("Arial", 12),
            bg="#1a1a2e",
            fg="#888888"
        )
        self.date_label.pack(pady=5)
        
        # День недели
        self.weekday_label = tk.Label(
            self.root,
            font=("Arial", 11),
            bg="#1a1a2e",
            fg="#666666"
        )
        self.weekday_label.pack()
        
        # Фрейм для кнопок
        button_frame = tk.Frame(self.root, bg="#1a1a2e")
        button_frame.pack(pady=15)
        
        # Кнопка переключения формата
        self.format_btn = tk.Button(
            button_frame,
            text="24H",
            font=("Arial", 10),
            bg="#16213e",
            fg="#00ffcc",
            command=self.toggle_format,
            width=8
        )
        self.format_btn.pack(side=tk.LEFT, padx=5)
        
        # Кнопка выхода (ИСПРАВЛЕНО - используем destroy вместо quit)
        exit_btn = tk.Button(
            button_frame,
            text="Выход",
            font=("Arial", 10),
            bg="#e94560",
            fg="white",
            command=self.exit_app,  # Новая функция
            width=8
        )
        exit_btn.pack(side=tk.LEFT, padx=5)
        
        # Информация
        info_label = tk.Label(
            self.root,
            text="Цифровые часы | Обновление каждую секунду",
            font=("Arial", 8),
            bg="#1a1a2e",
            fg="#444444"
        )
        info_label.pack(pady=5)
    
    def toggle_format(self):
        """Переключает между 24-часовым и 12-часовым форматом"""
        self.is_24h = not self.is_24h
        self.format_btn.config(text="24H" if self.is_24h else "12H")
    
    def exit_app(self):
        """Корректно закрывает приложение"""
        self.root.destroy()  # Используем destroy вместо quit
    
    def update_time(self):
        """Обновляет время, дату и день недели"""
        now = datetime.now()
        
        # Форматирование времени
        if self.is_24h:
            time_str = now.strftime("%H:%M:%S")
        else:
            time_str = now.strftime("%I:%M:%S %p")
            if time_str.startswith("0"):
                time_str = time_str[1:]  # Убираем ведущий ноль
        
        # Форматирование даты
        date_str = now.strftime("%d.%m.%Y")
        
        # День недели на русском
        weekdays = {
            "Monday": "Понедельник",
            "Tuesday": "Вторник",
            "Wednesday": "Среда",
            "Thursday": "Четверг",
            "Friday": "Пятница",
            "Saturday": "Суббота",
            "Sunday": "Воскресенье"
        }
        english_weekday = now.strftime("%A")
        weekday_str = weekdays.get(english_weekday, english_weekday)
        
        # Обновляем метки
        self.time_label.config(text=time_str)
        self.date_label.config(text=date_str)
        self.weekday_label.config(text=weekday_str)
        
        # Запускаем обновление через 1 секунду
        self.root.after(1000, self.update_time)

# Запуск приложения
if __name__ == "__main__":
    root = tk.Tk()
    clock = DigitalClock(root)
    
    # Добавляем обработку закрытия окна крестиком
    root.protocol("WM_DELETE_WINDOW", clock.exit_app)
    
    root.mainloop()
```
