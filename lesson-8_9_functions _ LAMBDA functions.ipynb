```python
def addition(x, y):
    return x + y

def multiplication(x, y):
    return x * y

def exponentiation(x, y):
    return x ** y
```

```python
func = addition

operation_sign = input()

if operation_sign == '*':
    func = multiplication
elif operation_sign == '^':
    func = exponentiation

x, y = 3, 4
func(x, y)
```

```python
operation = {
    '+': addition,
    '*': multiplication,
    '^': exponentiation
}

operation_sign = input()

func = operation[operation_sign]

x, y = 3, 4
func(x, y)
```

```python
# Форматировать строку: 
# 1. Как в предложениях.
# 2. все строчные
# 3. ВСЕ ПРОПИСНЫЕ
# 4. Начинать С Прописных
# 5. иЗМЕНИТЬ РЕГИСТР

# начальное значение строки
s = '''функция в PYTHON - объект, принимающий аргументы и возвращающий значение. обычно функция определяется с помощью инструкции def. инструкция return говорит, что нужно вернуть значение.'''
# print(s)

def start_with_upper(s):
  return s[0].upper() + s[1:]

# пользовательская функция принимает строку возвращает строку, отформатированный как в предложениях. 
# Пусть все предложения заканчиваются точкой
def as_sentence(s):
  seq = s.split(sep='. ')
  for i in range(len(seq)):
    seq[i] = start_with_upper(seq[i]) # capitalize()

  return '. '.join(seq)

# тестиование работы функции start_with_upper
# first_upper = start_with_upper(s)
# print(first_upper)

# тестиование работы функции as_sentence
sentence = as_sentence(s)
print(sentence)
```

```python
def all_lower(s):
    return s.lower()

def all_upper(s):
    return s.upper()

print(all_lower(s))
print(all_upper(s))
```

```python
# пользовательская функция принимает текст, возвращает текст с измененным регистром
def change_register(s):
  str_as_list = list(s)
  for i in range(len(str_as_list)):
    if str_as_list[i].isalpha(): # swapcase()
      if str_as_list[i].isupper():
        str_as_list[i] = str_as_list[i].lower()
      else:
        str_as_list[i] = str_as_list[i].upper()
  return ''.join(str_as_list)

print(change_register(s))
```
