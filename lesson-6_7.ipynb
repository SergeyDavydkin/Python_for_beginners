```python
def analyze_word_occurrences(text):
    """
    Функция анализирует вхождения слов в тексте
    
    Args:
        text (str): Входной текст
    
    Returns:
        tuple: (список предыдущих вхождений, словарь итоговых счетчиков)
    """
    count = {}
    previous = []
    
    print("Анализ текста:", text)
    print("Предыдущие вхождения:", end=' ')
    
    for word in text.split():
        count[word] = count.get(word, 0) + 1
        prev_count = count[word] - 1
        previous.append(prev_count)
        print(prev_count, end=' ')
    
    print("\nИтоговая статистика:")
    for word, total in count.items():
        print(f"  '{word}': {total} раз")
    
    return previous, count

# Вызов функции
text = 'one two one tho three'
prev_counts, final_counts = analyze_word_occurrences(text)
```

```python
def solve(data_string):
    """
    Функция для обработки данных без использования input()
    
    Аргументы:
        data_string: строка с входными данными в формате:
            N
            Фамилия Имя паспорт номер1 номер2 ...
            ...
    
    Возвращает:
        строку с отсортированными записями
    """
    # Разбиваем входные данные на строки
    lines = data_string.strip().split('\n')
    
    if not lines or not lines[0].strip():
        return ""
    
    # Первая строка - количество записей
    n = int(lines[0].strip())
    
    records = []
    # Читаем следующие n строк
    for i in range(1, min(n, len(lines) - 1) + 1):
        line = lines[i].strip()
        if not line:
            continue
            
        parts = line.split()
        if len(parts) < 3:  # Минимум: фамилия, имя, паспорт
            continue
            
        surname = parts[0]
        name = parts[1]
        passport = parts[2]
        phones = parts[3:] if len(parts) > 3 else []
        
        records.append((surname, name, passport, phones))
    
    # Сортируем по паспортным данным
    records.sort(key=lambda x: x[2])
    
    # Формируем выходную строку
    output_lines = []
    for surname, name, passport, phones in records:
        line = f"{surname} {name} {passport}"
        if phones:
            line += f" {' '.join(phones)}"
        output_lines.append(line)
    
    return "\n".join(output_lines)


def pretty_solve(data_string):
    """
    Красивая обертка для функции solve с визуализацией
    """
    from datetime import datetime
    
    print("\n" + "═" * 70)
    print("📱 СИСТЕМА УЧЕТА ТЕЛЕФОННЫХ НОМЕРОВ")
    print("═" * 70)
    
    # Парсим исходные данные для отображения
    lines = data_string.strip().split('\n')
    if not lines or not lines[0].strip():
        print("❌ Нет данных для обработки")
        return ""
    
    n = int(lines[0].strip())
    original_records = []
    
    print(f"\n📊 Количество записей: {n}")
    print(f"🕐 Время обработки: {datetime.now().strftime('%H:%M:%S')}")
    
    # Собираем исходные записи
    for i in range(1, min(n, len(lines) - 1) + 1):
        line = lines[i].strip()
        if line:
            parts = line.split()
            if len(parts) >= 3:
                original_records.append((parts[0], parts[1], parts[2], parts[3:] if len(parts) > 3 else []))
    
    # Выводим исходные данные в красивой таблице
    print("\n📋 ИСХОДНЫЕ ДАННЫЕ (в порядке ввода):")
    print("─" * 70)
    print(f"│ {'№':<3} │ {'Фамилия':<12} │ {'Имя':<10} │ {'Паспорт':<8} │ {'Телефон(ы)':<25} │")
    print("─" * 70)
    
    for idx, (surname, name, passport, phones) in enumerate(original_records, 1):
        phone_str = ', '.join(phones) if phones else "нет номеров"
        if len(phone_str) > 25:
            phone_str = phone_str[:22] + "..."
        print(f"│ {idx:<3} │ {surname:<12} │ {name:<10} │ {passport:<8} │ {phone_str:<25} │")
    
    print("─" * 70)
    
    # Показываем паспорта до сортировки
    passports = [r[2] for r in original_records]
    print(f"\n🔢 Паспорта до сортировки: {' → '.join(passports)}")
    
    # Запускаем сортировку
    print("\n⚙️  Выполняется сортировка по паспортным данным...")
    
    result = solve(data_string)
    
    # Парсим результат
    result_records = []
    if result.strip():
        for line in result.strip().split('\n'):
            parts = line.split()
            if len(parts) >= 3:
                result_records.append((parts[0], parts[1], parts[2], parts[3:] if len(parts) > 3 else []))
    
    # Показываем паспорта после сортировки
    sorted_passports = [r[2] for r in result_records]
    print(f"🔢 Паспорта после сортировки: {' → '.join(sorted_passports)}")
    
    # Выводим результат в красивой таблице
    print("\n✅ РЕЗУЛЬТАТ (отсортировано по паспорту):")
    print("─" * 70)
    print(f"│ {'№':<3} │ {'Фамилия':<12} │ {'Имя':<10} │ {'Паспорт':<8} │ {'Телефон(ы)':<25} │")
    print("─" * 70)
    
    for idx, (surname, name, passport, phones) in enumerate(result_records, 1):
        phone_str = ', '.join(phones) if phones else "нет номеров"
        if len(phone_str) > 25:
            phone_str = phone_str[:22] + "..."
        print(f"│ {idx:<3} │ {surname:<12} │ {name:<10} │ {passport:<8} │ {phone_str:<25} │")
    
    print("─" * 70)
    
    # Статистика
    print(f"\n📊 Статистика:")
    print(f"   • Всего записей: {len(result_records)}")
    print(f"   • Всего номеров: {sum(len(r[3]) for r in result_records)}")
    
    print("\n" + "═" * 70)
    print("✅ Обработка завершена успешно!")
    print("═" * 70 + "\n")
    
    return result


# Демонстрация работы
if __name__ == "__main__":
    input_data = """3
Ivanov Ivan 352251 89112457891
Petrov Petr 351231 89112014573
Gartess Alena 201456 89111012354
"""
    
    # Запускаем с красивым выводом
    result = pretty_solve(input_data)
    
    # Можно также просто вывести результат без оформления
    print("\n" + "=" * 50)
    print("Простой вывод (без оформления):")
    print("=" * 50)
    print(solve(input_data))
```

```python
def pretty_solve_color(data_string):
    """
    Красивая обертка с поддержкой цветов (для терминалов)
    """
    # ANSI цветовые коды
    class Colors:
        HEADER = '\033[95m'
        BLUE = '\033[94m'
        CYAN = '\033[96m'
        GREEN = '\033[92m'
        WARNING = '\033[93m'
        FAIL = '\033[91m'
        END = '\033[0m'
        BOLD = '\033[1m'
        UNDERLINE = '\033[4m'
    
    print(f"\n{Colors.HEADER}{'='*60}{Colors.END}")
    print(f"{Colors.BOLD}{Colors.CYAN}📱 СИСТЕМА УЧЕТА ТЕЛЕФОННЫХ НОМЕРОВ{Colors.END}")
    print(f"{Colors.HEADER}{'='*60}{Colors.END}")
    
    lines = data_string.strip().split('\n')
    if not lines or not lines[0].strip():
        print(f"{Colors.FAIL}❌ Нет данных для обработки{Colors.END}")
        return ""
    
    n = int(lines[0].strip())
    original_records = []
    
    print(f"\n{Colors.BOLD}📊 Количество записей:{Colors.END} {n}")
    
    # Собираем исходные записи
    for i in range(1, min(n, len(lines) - 1) + 1):
        line = lines[i].strip()
        if line:
            parts = line.split()
            if len(parts) >= 3:
                original_records.append((parts[0], parts[1], parts[2], parts[3:] if len(parts) > 3 else []))
    
    # Выводим исходные данные
    print(f"\n{Colors.BOLD}{Colors.BLUE}📋 ИСХОДНЫЕ ДАННЫЕ:{Colors.END}")
    print(f"{Colors.CYAN}{'-'*60}{Colors.END}")
    
    for idx, (surname, name, passport, phones) in enumerate(original_records, 1):
        phone_str = ', '.join(phones) if phones else f"{Colors.WARNING}нет номеров{Colors.END}"
        print(f"  {idx}. {Colors.GREEN}{surname} {name}{Colors.END} "
              f"[пасспорт: {Colors.BOLD}{passport}{Colors.END}] -> {phone_str}")
    
    # Показываем паспорта до сортировки
    passports = [r[2] for r in original_records]
    print(f"\n{Colors.BOLD}🔢 Паспорта до сортировки:{Colors.END} ", end="")
    print(" → ".join([f"{Colors.BOLD}{p}{Colors.END}" for p in passports]))
    
    # Запускаем сортировку
    print(f"\n{Colors.CYAN}⚙️  Выполняется сортировка по паспортным данным...{Colors.END}")
    
    result = solve(data_string)
    
    # Парсим результат
    result_records = []
    if result.strip():
        for line in result.strip().split('\n'):
            parts = line.split()
            if len(parts) >= 3:
                result_records.append((parts[0], parts[1], parts[2], parts[3:] if len(parts) > 3 else []))
    
    # Показываем паспорта после сортировки
    sorted_passports = [r[2] for r in result_records]
    print(f"{Colors.BOLD}🔢 Паспорта после сортировки:{Colors.END} ", end="")
    print(" → ".join([f"{Colors.BOLD}{p}{Colors.END}" for p in sorted_passports]))
    
    # Выводим результат
    print(f"\n{Colors.BOLD}{Colors.GREEN}✅ РЕЗУЛЬТАТ (отсортировано по паспорту):{Colors.END}")
    print(f"{Colors.CYAN}{'-'*60}{Colors.END}")
    
    for idx, (surname, name, passport, phones) in enumerate(result_records, 1):
        phone_str = ', '.join(phones) if phones else f"{Colors.WARNING}нет номеров{Colors.END}"
        print(f"  {idx}. {Colors.GREEN}{surname} {name}{Colors.END} "
              f"[пасспорт: {Colors.BOLD}{passport}{Colors.END}] -> {phone_str}")
    
    # Статистика
    total_phones = sum(len(r[3]) for r in result_records)
    print(f"\n{Colors.BOLD}📊 Статистика:{Colors.END}")
    print(f"  • {Colors.CYAN}Всего записей:{Colors.END} {len(result_records)}")
    print(f"  • {Colors.CYAN}Всего номеров:{Colors.END} {total_phones}")
    
    print(f"\n{Colors.GREEN}{'='*60}{Colors.END}")
    print(f"{Colors.BOLD}{Colors.GREEN}✅ Обработка завершена успешно!{Colors.END}")
    print(f"{Colors.GREEN}{'='*60}{Colors.END}\n")
    
    return result


# Запуск
input_data = """3
Ivanov Ivan 352251 89112457891
Petrov Petr 351231 89112014573
Gartess Alena 201456 89111012354
"""

# Выбирайте любой вариант:
# pretty_solve(input_data)  # без цветов
print(pretty_solve_color(input_data))  # с цветами (для терминала)
```

```python
def process_phone_records_detailed():
    """
    Функция обрабатывает записи с детальным разбором данных
    """
    # Получаем количество записей
    n = int(input())
    
    # Создаем список для хранения записей в структурированном виде
    records = []
    
    # Читаем все записи
    for i in range(n):
        line = input().strip()
        # Разбиваем строку на части
        parts = line.split()
        surname = parts[0]      # Фамилия
        name = parts[1]         # Имя
        passport = int(parts[2]) # Паспортные данные (преобразуем в число для сортировки)
        phones = parts[3:]       # Список телефонных номеров
        
        # Сохраняем в структурированном виде
        records.append({
            'surname': surname,
            'name': name,
            'passport': passport,
            'phones': phones,
            'original': line     # Сохраняем исходную строку
        })
    
    # Сортируем по паспортным данным
    sorted_records = sorted(records, key=lambda x: x['passport'])
    
    # Выводим результат
    print("\nОтсортированные записи по паспортным данным:")
    print("-" * 60)
    for record in sorted_records:
        print(record['original'])
        
    # Дополнительная информация
    print("\n" + "=" * 60)
    print("Детальная информация:")
    for record in sorted_records:
        phone_count = len(record['phones'])
        phones_str = ', '.join(record['phones'])
        print(f"{record['surname']} {record['name']} (пасспорт: {record['passport']}): {phone_count} номер(ов) - {phones_str}")

# Вызов функции
process_phone_records_detailed()
```

```python
def solve(data):
    """
    Функция для обработки данных о покупках клиентов
    
    Аргументы:
        data: строка с входными данными в формате:
            N (количество клиентов)
            Фамилия Имя
            K (количество купленных товаров)
            товар цена
            ...
    
    Возвращает:
        строку с агрегированными данными по каждому клиенту
    """
    # 1. ПОДГОТОВКА ДАННЫХ
    # Разбиваем входную строку на отдельные строки, удаляем лишние пробелы в начале и конце
    lines = data.strip().split('\n')
    
    # Инициализируем указатель позиции (индекс текущей строки)
    # и читаем количество клиентов из первой строки
    pos, n = 0, int(lines[0])
    pos += 1  # Переходим к следующей строке
    
    # Словарь для хранения данных о покупках
    # Формат: {имя_клиента: {товар1: сумма1, товар2: сумма2, ...}}
    sales = {}
    
    # 2. ОБРАБОТКА КАЖДОГО КЛИЕНТА
    for _ in range(n):
        # Читаем имя клиента
        name = lines[pos]
        pos += 1
        
        # Читаем количество купленных товаров
        k = int(lines[pos])
        pos += 1
        
        # Инициализируем пустой словарь для клиента, если его еще нет
        # setdefault создает запись с пустым словарем, если ключ отсутствует
        sales.setdefault(name, {})
        
        # 3. ОБРАБОТКА ТОВАРОВ КЛИЕНТА
        for _ in range(k):
            # Разбиваем строку на название товара и цену
            prod, price = lines[pos].split()
            pos += 1
            
            # Агрегируем данные:
            # get(prod, 0) - получаем текущую сумму по товару или 0, если товара нет
            # добавляем цену и сохраняем обратно
            sales[name][prod] = sales[name].get(prod, 0) + int(price)
    
    # 4. ФОРМИРОВАНИЕ РЕЗУЛЬТАТА С СОРТИРОВКОЙ ТОВАРОВ
    result_lines = []
    
    for name, products in sales.items():
        # Добавляем имя клиента
        result_lines.append(name)
        
        # СОРТИРУЕМ ТОВАРЫ ПО АЛФАВИТУ
        # items() возвращает пары (товар, сумма), сортируем по названию товара
        sorted_products = sorted(products.items(), key=lambda x: x[0])
        
        # Добавляем каждый товар с суммой
        for product, total in sorted_products:
            result_lines.append(f"{product} {total}")
    
    # Объединяем все строки с переносами
    return "\n".join(result_lines)


def solve_compact(data):
    """
    Компактная версия с сортировкой товаров
    """
    lines = data.strip().split('\n')
    pos, n = 0, int(lines[0]); pos += 1
    sales = {}
    
    for _ in range(n):
        name = lines[pos]; pos += 1
        k = int(lines[pos]); pos += 1
        sales.setdefault(name, {})
        for _ in range(k):
            prod, price = lines[pos].split(); pos += 1
            sales[name][prod] = sales[name].get(prod, 0) + int(price)
    
    # Формируем результат с сортировкой товаров по алфавиту
    return '\n'.join(
        f"{name}\n" + '\n'.join(f"{p} {s}" for p, s in sorted(products.items(), key=lambda x: x[0]))
        for name, products in sales.items()
    )


def solve_with_details(data):
    """
    Детальная версия с подробным выводом и сортировкой
    """
    lines = data.strip().split('\n')
    pos, n = 0, int(lines[0]); pos += 1
    sales = {}
    
    print("=" * 60)
    print("НАЧАЛО ОБРАБОТКИ ДАННЫХ")
    print("=" * 60)
    print(f"Всего клиентов в базе: {n}")
    
    for client_num in range(n):
        name = lines[pos]; pos += 1
        k = int(lines[pos]); pos += 1
        sales.setdefault(name, {})
        
        print(f"\n📋 Клиент {client_num + 1}: {name}")
        print(f"   Куплено товаров: {k}")
        
        for item_num in range(k):
            prod, price = lines[pos].split(); pos += 1
            old_total = sales[name].get(prod, 0)
            new_total = old_total + int(price)
            sales[name][prod] = new_total
            
            print(f"   {item_num + 1}. {prod}: {price} руб. (всего: {new_total} руб.)")
    
    print("\n" + "=" * 60)
    print("АГРЕГИРОВАННЫЕ РЕЗУЛЬТАТЫ (С СОРТИРОВКОЙ ТОВАРОВ)")
    print("=" * 60)
    
    result_lines = []
    for name, products in sales.items():
        print(f"\n👤 {name}")
        result_lines.append(name)
        
        # СОРТИРУЕМ ТОВАРЫ ПО АЛФАВИТУ
        sorted_products = sorted(products.items(), key=lambda x: x[0])
        
        for product, total in sorted_products:
            print(f"   • {product}: {total} руб.")
            result_lines.append(f"{product} {total}")
    
    print("\n" + "=" * 60)
    print("ИТОГОВЫЙ ФОРМАТ (ДЛЯ ПРОГРАММЫ)")
    print("=" * 60)
    result = "\n".join(result_lines)
    print(result)
    
    return result


# ТЕСТОВЫЕ ДАННЫЕ
input_data = """3
Ivanov Ivan
3
Tomato 95
Cucumbers 75
Milk 70
Ivanov Ivan
3
Tomato 95
Cucumbers 75
Milk 70
Petrov Petr
4
Tomato 195
Cucumbers 120
Milk 70
Bred 45
"""

# print("\n" + "🔥" * 30)
# print("ЗАПУСК БАЗОВОЙ ВЕРСИИ")
# print("🔥" * 30)
# print(solve(input_data))

print("\n" + "🔥" * 30)
print("ЗАПУСК ВЕРСИИ С ДЕТАЛЬНЫМ ВЫВОДОМ")
print("🔥" * 30)
solve_with_details(input_data)
```
