````python
import json

file_name = path + '13.json'
user = {"firstName": "Иван", 
        "lastName": "Иванов", 
        "age": 28, 
        "married": False, 
        "phones": ["(908)123-45-67", "(980)123-45-67"]}

# Открываем в текстовом режиме с указанием кодировки
with open(file_name, 'w', encoding='utf-8') as f:
    json.dump(user, f, ensure_ascii=False, indent=4)  # записываем сразу объект user
````

````python
# Открываем файл в текстовом режиме для чтения с указанием кодировки UTF-8,
# так как JSON-файл содержит текст (в том числе русские символы)
with open(file_name, 'r', encoding='utf-8') as f:
    # Метод json.load() читает JSON-данные из файла и автоматически
    # преобразует их в соответствующий объект Python (словарь, список и т.д.)
    # В отличие от json.loads(), который принимает строку, json.load() работает
    # непосредственно с файловым объектом, что удобнее и эффективнее
    obj = json.load(f)

# После десериализации можно обращаться к данным как к обычному словарю Python
print(obj['firstName'])  # Выводит значение ключа 'firstName' -> "Иван"
print(obj['age'])        # Выводит значение ключа 'age' -> 28
print(obj['phones'][1])  # Выводит второй элемент списка 'phones' -> "(908)123-45-67"
````

### Задание 1.1 — Первая запись

Создайте файл `hello.txt` и запишите в него строку `"Привет, мир!"`. Затем прочитайте содержимое и выведите на экран.

````python
# Задание 1.1 — Первая запись
# Создайте файл hello.txt и запишите в него строку "Привет, мир!"
# Затем прочитайте содержимое и выведите на экран.

import os

# Указываем путь к папке
path = f"D:/Python/Pyton-2026/files/"

# Полный путь к файлу
file_name = os.path.join(path, "hello.txt")  # D:/Python/Pyton-2026/files/hello.txt

# Создаём папку, если её не существует
os.makedirs(path, exist_ok=True)

# Запись в файл
with open(file_name, 'w', encoding='utf-8') as file:
    file.write("Привет, мир!")
    print(f"✓ Записано в файл: {file_name}")

# Чтение из файла
with open(file_name, 'r', encoding='utf-8') as file:
    content = file.read()
    print(f"✓ Прочитано из файла: {content}")
````

````python
# Вариант 2: С проверкой существования файла
import os

# Запись в файл
with open(file_name, 'w', encoding='utf-8') as file:
    file.write("Привет, мир!")

# Проверка существования и чтение
if os.path.exists(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        content = file.read()
        print(f"Содержимое файла: {content}")
else:
    print("Файл не найден")
````

````python
# Вариант 3: С использованием pathlib (современный подход)
from pathlib import Path

# Создаём объект Path
path = Path("D:/Python/Pyton-2026/files/")
file_path = path / "hello.txt"

# Создаём папку, если её не существует
path.mkdir(parents=True, exist_ok=True)

# Запись в файл
file_path.write_text("Привет, мир!", encoding='utf-8')
print(f"✓ Записано в файл: {file_path}")

# Чтение из файла
content = file_path.read_text(encoding='utf-8')
print(f"✓ Прочитано из файла: {content}")
````

````python
# Вариант 4: С обработкой ошибок
import os

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "hello.txt")

try:
    # Создаём папку, если её не существует
    os.makedirs(path, exist_ok=True)
    
    # Запись в файл
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write("Привет, мир!")
    print(f"✓ Файл успешно создан: {file_name}")
    
    # Чтение из файла
    with open(file_name, 'r', encoding='utf-8') as file:
        content = file.read()
        print(f"✓ Содержимое файла: {content}")
        
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлу '{file_name}'")
except Exception as e:
    print(f"Произошла ошибка: {e}")
````

### Задание 1.2 — Многострочный текст

Создайте файл `poem.txt` и запишите в него стихотворение из 4 строк (с переносами). Прочитайте файл тремя способами:

- `read()` — целиком

- `readline()` — построчно в цикле

- `readlines()` — списком строк

````python
import os

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "poem.txt")

# Создаём папку, если её не существует
os.makedirs(path, exist_ok=True)

# Запись в файл
with open(file_name, 'w', encoding='utf-8') as file:
    file.write("Весна, весна! Как воздух чист!\nКак ясен небосклон!\nСвоей лазурию живой\nСлепит мне очи он.\n")
    print(f"✓ Записано в файл: {file_name}")

print("\n" + "="*50 + "\n")

# Способ 1: read() — чтение целиком
print("Способ 1: read() — чтение всего файла целиком")
print("-" * 40)
with open(file_name, 'r', encoding='utf-8') as file:
    content = file.read()
    print(content)
print("="*50 + "\n")

# Способ 2: readline() — построчное чтение в цикле
print("Способ 2: readline() — построчное чтение в цикле")
print("-" * 40)
with open(file_name, 'r', encoding='utf-8') as file:
    line = file.readline()
    while line:
        print(line, end='')  # end='' чтобы не добавлять лишний перевод строки
        line = file.readline()
print("\n" + "="*50 + "\n")

# Способ 3: readlines() — чтение списком строк
print("Способ 3: readlines() — чтение списком строк")
print("-" * 40)
with open(file_name, 'r', encoding='utf-8') as file:
    lines = file.readlines()
    print(f"Тип объекта: {type(lines)}")
    print(f"Количество строк: {len(lines)}")
    print(f"Список строк: {lines}")
    print("\nВывод строк по отдельности:")
    for i, line in enumerate(lines, 1):
        print(f"Строка {i}: {line}", end='')
````

````python
# Альтернативный вариант с использованием for-цикла (более Pythonic):
import os

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "poem.txt")

# Создаём папку, если её не существует
os.makedirs(path, exist_ok=True)

# Запись в файл
with open(file_name, 'w', encoding='utf-8') as file:
    file.write("Весна, весна! Как воздух чист!\n"
               "Как ясен небосклон!\n"
               "Своей лазурию живой\n"
               "Слепит мне очи он.\n")
    print(f"✓ Записано в файл: {file_name}\n")

# Способ 1: read() — целиком
with open(file_name, 'r', encoding='utf-8') as file:
    print("1. read() - весь файл как строка:")
    print(file.read())
print()

# Способ 2: итерация по файлу (самый Pythonic способ)
with open(file_name, 'r', encoding='utf-8') as file:
    print("2. Итерация по файлу - построчно:")
    for line in file:
        print(line, end='')
print("\n")

# Способ 3: readlines() — список строк
with open(file_name, 'r', encoding='utf-8') as file:
    print("3. readlines() - список строк:")
    lines = file.readlines()
    for i, line in enumerate(lines, 1):
        print(f"  {i}: {line}", end='')
````

Создайте файл `log.txt`, запишите в него текущую дату и время. Затем 5 раз дозапишите сообщение "Выполнена операция №X" (где X — номер операции). В конце выведите содержимое файла.

````python
import os
from datetime import datetime

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "log.txt")

# Создаём папку, если её не существует
os.makedirs(path, exist_ok=True)

# 1. Записываем текущую дату и время в файл
current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

with open(file_name, 'w', encoding='utf-8') as file:
    file.write(f"Лог-файл создан: {current_time}\n")
    print(f"✓ Создан файл: {file_name}")
    print(f"✓ Записана дата: {current_time}\n")

# 2. 5 раз дозаписываем сообщения
for i in range(1, 6):
    with open(file_name, 'a', encoding='utf-8') as file:
        file.write(f"Выполнена операция №{i}\n")
    print(f"✓ Добавлена операция №{i}")

# 3. Выводим содержимое файла
print("\n" + "="*50)
print("Содержимое файла log.txt:")
print("="*50)

with open(file_name, 'r', encoding='utf-8') as file:
    content = file.read()
    print(content)
````

### Задание 2.1 — Сумма чисел в файле

````python
# Вариант 4: С использованием pathlib и функциями
from pathlib import Path

def read_numbers(file_path):
    """Читает числа из файла, возвращает список чисел"""
    numbers = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if line:
                numbers.append(int(line))
    return numbers

def write_numbers(file_path, numbers):
    """Записывает список чисел в файл (по одному на строку)"""
    with open(file_path, 'w', encoding='utf-8') as file:
        for number in numbers:
            file.write(f"{number}\n")

def filter_numbers(numbers):
    """Разделяет числа на чётные и нечётные"""
    even = [n for n in numbers if n % 2 == 0]
    odd = [n for n in numbers if n % 2 != 0]
    return even, odd

# Основная программа
path = Path("D:/Python/Pyton-2026/files/")
path.mkdir(parents=True, exist_ok=True)

input_file = path / "input.txt"
even_file = path / "even.txt"
odd_file = path / "odd.txt"

# Создаём input.txt
input_file.write_text("15\n22\n37\n48\n51\n64\n73\n86\n99\n100\n", encoding='utf-8')

# Обрабатываем
numbers = read_numbers(input_file)
even_numbers, odd_numbers = filter_numbers(numbers)
write_numbers(even_file, even_numbers)
write_numbers(odd_file, odd_numbers)

# Выводим результат
print(f"Всего чисел: {len(numbers)}")
print(f"Чётные ({len(even_numbers)}): {even_numbers}")
print(f"Нечётные ({len(odd_numbers)}): {odd_numbers}")
print(f"\nФайлы созданы:")
print(f"  {even_file}")
print(f"  {odd_file}")
````

````python
# Вариант 3: С функцией для анализа зарплат
import os
import csv

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "employees.csv")

os.makedirs(path, exist_ok=True)

# Создаём файл
with open(file_name, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Имя", "Должность", "Зарплата"])
    writer.writerows([
        ["Иван Петров", "Разработчик", 85000],
        ["Анна Сидорова", "Менеджер", 65000],
        ["Сергей Иванов", "Тестировщик", 45000],
        ["Мария Смирнова", "Аналитик", 72000],
        ["Алексей Кузнецов", "Дизайнер", 48000],
        ["Елена Васильева", "HR-специалист", 55000]
    ])

def filter_by_salary(file_path, min_salary):
    """Возвращает список сотрудников с зарплатой выше min_salary"""
    employees = []
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            salary = int(row["Зарплата"])
            if salary > min_salary:
                employees.append(row)
    return employees

def add_employee(file_path, name, position, salary):
    """Добавляет нового сотрудника в файл"""
    with open(file_path, 'a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow([name, position, salary])

def display_employees(employees, title="Список сотрудников"):
    """Выводит список сотрудников в отформатированном виде"""
    if not employees:
        print("Нет сотрудников для отображения")
        return
    
    print("="*60)
    print(title)
    print("="*60)
    print(f"{'Имя':<20} {'Должность':<20} {'Зарплата':<10}")
    print("-"*60)
    
    for emp in employees:
        if isinstance(emp, dict):
            name = emp["Имя"]
            position = emp["Должность"]
            salary = int(emp["Зарплата"])  # Преобразуем в число
        else:
            name, position, salary = emp
            salary = int(salary)  # Преобразуем в число
        
        print(f"{name:<20} {position:<20} {salary:>10,}")
    
    print("-"*60)

# Основная программа
# 1. Выводим сотрудников с зарплатой выше 50 000
high_salary = filter_by_salary(file_name, 50000)
display_employees(high_salary, "СОТРУДНИКИ С ЗАРПЛАТОЙ ВЫШЕ 50 000:")

# 2. Добавляем нового сотрудника
add_employee(file_name, "Ольга Новикова", "Архитектор", 95000)
print(f"\n✓ Добавлен новый сотрудник: Ольга Новикова")

# 3. Показываем всех сотрудников
all_employees = []
with open(file_name, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # пропускаем заголовок
    all_employees = list(reader)

display_employees(all_employees, "ОБНОВЛЁННЫЙ СПИСОК ВСЕХ СОТРУДНИКОВ:")
````

````python
# Вариант 4: С функцией для преобразования разных форматов
import os
import csv

def convert_delimited_file(input_file, output_file, input_delimiter='|', output_delimiter=','):
    """
    Преобразует файл с одним разделителем в файл с другим разделителем
    
    Args:
        input_file: путь к исходному файлу
        output_file: путь к выходному файлу
        input_delimiter: разделитель в исходном файле
        output_delimiter: разделитель в выходном файле
    """
    converted_rows = 0
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        with open(output_file, 'w', newline='', encoding='utf-8') as outfile:
            writer = csv.writer(outfile, delimiter=output_delimiter)
            
            for line in infile:
                line = line.strip()
                if line:
                    # Разбиваем строку по входному разделителю
                    parts = line.split(input_delimiter)
                    # Записываем с выходным разделителем
                    writer.writerow(parts)
                    converted_rows += 1
                    
                    # Выводим информацию о преобразовании
                    print(f"Строка {converted_rows}:")
                    print(f"  Было: {line}")
                    print(f"  Стало: {output_delimiter.join(parts)}")
    
    return converted_rows

# Основная программа
path = f"D:/Python/Pyton-2026/files/"
input_file = os.path.join(path, "data.txt")
output_file = os.path.join(path, "data.csv")

os.makedirs(path, exist_ok=True)

# Создаём исходный файл
with open(input_file, 'w', encoding='utf-8') as file:
    file.write("Имя|Город|Возраст\n")
    file.write("Иван|Москва|25\n")
    file.write("Ольга|СПб|30\n")

print("="*50)
print("ПРЕОБРАЗОВАНИЕ ФАЙЛА:")
print("="*50)

# Преобразуем файл
count = convert_delimited_file(input_file, output_file, input_delimiter='|', output_delimiter=',')

print(f"\n✓ Преобразовано строк: {count}")
print(f"✓ Результат сохранён: {output_file}")

# Проверяем результат
print("\n" + "="*50)
print("РЕЗУЛЬТАТ (data.csv):")
print("="*50)
with open(output_file, 'r', encoding='utf-8') as file:
    print(file.read())
````

````python
# Вариант 5: С обработкой ошибок и дополнительными возможностями
import os
import csv

path = f"D:/Python/Pyton-2026/files/"
input_file = os.path.join(path, "data.txt")
output_file = os.path.join(path, "data.csv")

os.makedirs(path, exist_ok=True)

# Создаём исходный файл
with open(input_file, 'w', encoding='utf-8') as file:
    file.write("Имя|Город|Возраст\n")
    file.write("Иван|Москва|25\n")
    file.write("Ольга|СПб|30\n")

try:
    # Проверяем существование исходного файла
    if not os.path.exists(input_file):
        raise FileNotFoundError(f"Файл {input_file} не найден")
    
    # Получаем размер исходного файла
    input_size = os.path.getsize(input_file)
    print(f"Исходный файл: {input_file}")
    print(f"Размер: {input_size} байт\n")
    
    # Преобразуем файл
    converted_count = 0
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        with open(output_file, 'w', newline='', encoding='utf-8') as outfile:
            writer = csv.writer(outfile)
            
            for line_num, line in enumerate(infile, 1):
                # Удаляем лишние пробелы и символы перевода строки
                line = line.strip()
                
                if not line:
                    print(f"Строка {line_num}: пустая строка, пропущена")
                    continue
                
                # Разбиваем по разделителю
                parts = line.split('|')
                
                # Очищаем каждую часть от лишних пробелов
                parts = [part.strip() for part in parts]
                
                # Записываем в CSV
                writer.writerow(parts)
                converted_count += 1
                
                print(f"Строка {line_num}: {line} -> {', '.join(parts)}")
    
    # Получаем размер выходного файла
    output_size = os.path.getsize(output_file)
    
    print(f"\n" + "="*50)
    print("СТАТИСТИКА ПРЕОБРАЗОВАНИЯ:")
    print("="*50)
    print(f"Обработано строк: {converted_count}")
    print(f"Выходной файл: {output_file}")
    print(f"Размер: {output_size} байт")
    print(f"Экономия места: {input_size - output_size} байт")
    
    # Показываем результат
    print("\n" + "="*50)
    print("СОДЕРЖИМОЕ CSV-ФАЙЛА:")
    print("="*50)
    
    with open(output_file, 'r', encoding='utf-8') as file:
        for i, line in enumerate(file, 1):
            print(f"{i}: {line.strip()}")
    
except FileNotFoundError as e:
    print(f"Ошибка: {e}")
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлам")
except Exception as e:
    print(f"Произошла ошибка: {e}")
````

````python
# Вариант 3: С подробной статистикой
import os
import json

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "book.json")

def analyze_book(book):
    """Анализирует книгу и возвращает статистику"""
    stats = {
        "название": book.get("название", "Не указано"),
        "автор": book.get("автор", "Не указан"),
        "год": book.get("год", "Не указан"),
        "жанр": book.get("жанр", "Не указан"),
        "количество_глав": len(book.get("список_глав", [])),
        "есть_доп_информация": bool(book.get("дополнительная_информация")),
        "ключи": list(book.keys())
    }
    
    # Статистика по главам
    chapters = book.get("список_глав", [])
    if chapters:
        stats["средняя_длина_названия"] = sum(len(ch) for ch in chapters) / len(chapters)
        stats["самая_длинная_глава"] = max(chapters, key=len)
        stats["самая_короткая_глава"] = min(chapters, key=len)
    
    return stats

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        book = json.load(file)
    
    # Получаем статистику
    stats = analyze_book(book)
    
    # Выводим информацию
    print("="*70)
    print(f"📚 АНАЛИЗ КНИГИ: {stats['название']}")
    print("="*70)
    
    print("\n📋 ОСНОВНЫЕ ДАННЫЕ:")
    print(f"   ✍️  Автор: {stats['автор']}")
    print(f"   📅 Год издания: {stats['год']}")
    print(f"   🎭 Жанр: {stats['жанр']}")
    print(f"   📖 Количество глав: {stats['количество_глав']}")
    
    if stats['количество_глав'] > 0:
        print(f"\n📊 СТАТИСТИКА ПО ГЛАВАМ:")
        print(f"   📏 Средняя длина названия: {stats['средняя_длина_названия']:.1f} символов")
        print(f"   🔝 Самая длинная глава: {stats['самая_длинная_глава'][:50]}...")
        print(f"   🔽 Самая короткая глава: {stats['самая_короткая_глава']}")
    
    print(f"\n🗂️  Структура JSON:")
    print(f"   Всего полей: {len(stats['ключи'])}")
    print(f"   Поля: {', '.join(stats['ключи'])}")
    
    # Выводим первые 5 глав
    print(f"\n📑 ПЕРВЫЕ 5 ГЛАВ:")
    print("-"*70)
    chapters = book.get("список_глав", [])
    for i, chapter in enumerate(chapters[:5], 1):
        print(f"   {i}. {chapter}")
    
    if len(chapters) > 5:
        print(f"   ... и ещё {len(chapters) - 5} глав(ы)")
    
    print("\n" + "="*70)
    
except FileNotFoundError:
    print(f"❌ Файл не найден: {file_name}")
    print("   Выполните задание 4.1 для создания файла")
except json.JSONDecodeError:
    print(f"❌ Ошибка парсинга JSON в файле: {file_name}")
except Exception as e:
    print(f"❌ Ошибка: {e}")
````

````python
# Вариант 4: С интерактивным выводом
import os
import json

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "book.json")

def print_colored(text, color_code):
    """Выводит цветной текст (для терминала)"""
    try:
        print(f"\033[{color_code}m{text}\033[0m")
    except:
        print(text)

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        book = json.load(file)
    
    # Очистка экрана (опционально)
    # os.system('cls' if os.name == 'nt' else 'clear')
    
    title = book.get("название", "Книга")
    chapters = book.get("список_глав", [])
    
    print("\n" + "🌟" * 35)
    print(f"{' ' * 10}📖 {title.upper()} 📖")
    print("🌟" * 35)
    
    # Информационная панель
    print("\n┌" + "─" * 68 + "┐")
    print(f"│ {'Автор:':<12} {book.get('автор', 'Не указан'):<54}│")
    print(f"│ {'Год:':<12} {book.get('год', 'Не указан'):<54}│")
    print(f"│ {'Жанр:':<12} {book.get('жанр', 'Не указан'):<54}│")
    print(f"│ {'Глав:':<12} {len(chapters):<54}│")
    print("└" + "─" * 68 + "┘")
    
    # Меню выбора
    while True:
        print("\n" + "="*70)
        print("ВЫБЕРИТЕ ДЕЙСТВИЕ:")
        print("1 - Показать все главы")
        print("2 - Показать главы постранично")
        print("3 - Найти главу по названию")
        print("4 - Выйти")
        
        choice = input("\nВаш выбор (1-4): ").strip()
        
        if choice == "1":
            print("\n" + "="*70)
            print("ВСЕ ГЛАВЫ:")
            print("-"*70)
            for i, chapter in enumerate(chapters, 1):
                print(f"{i:3}. {chapter}")
            print("-"*70)
            
        elif choice == "2":
            page = 1
            per_page = 10
            total_pages = (len(chapters) + per_page - 1) // per_page
            
            while True:
                start = (page - 1) * per_page
                end = min(start + per_page, len(chapters))
                
                print(f"\n{'='*70}")
                print(f"ГЛАВЫ (страница {page} из {total_pages}):")
                print("-"*70)
                
                for i in range(start, end):
                    print(f"{i+1:3}. {chapters[i]}")
                
                print("-"*70)
                
                if page < total_pages:
                    print("n - следующая страница | p - предыдущая | q - выход")
                    cmd = input("-> ").lower()
                    if cmd == 'n':
                        page += 1
                    elif cmd == 'p' and page > 1:
                        page -= 1
                    elif cmd == 'q':
                        break
                else:
                    break
                    
        elif choice == "3":
            search = input("Введите текст для поиска в названиях глав: ").strip().lower()
            if search:
                found = [(i+1, ch) for i, ch in enumerate(chapters) 
                        if search in ch.lower()]
                
                if found:
                    print(f"\n🔍 Найдено глав: {len(found)}")
                    print("-"*70)
                    for num, chapter in found:
                        print(f"{num:3}. {chapter}")
                else:
                    print(f"\n❌ Главы с текстом '{search}' не найдены")
            else:
                print("\n❌ Введите текст для поиска")
                
        elif choice == "4":
            print("\nДо свидания! 👋")
            break
        else:
            print("\n❌ Неверный выбор. Попробуйте снова.")
    
except FileNotFoundError:
    print(f"\n❌ Файл не найден: {file_name}")
    print("   Сначала выполните задание 4.1 для создания файла book.json")
except json.JSONDecodeError:
    print(f"\n❌ Ошибка: Файл {file_name} содержит некорректный JSON")
except KeyboardInterrupt:
    print("\n\nПрограмма прервана пользователем")
except Exception as e:
    print(f"\n❌ Ошибка: {e}")
````

````python
# Вариант 2: С функциями для управления задачами
import os
import json
from datetime import datetime

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "tasks.json")

os.makedirs(path, exist_ok=True)

# Создаём исходный файл
initial_tasks = {
    "задачи": [
        {"id": 1, "task": "Изучить работу с файлами в Python", "status": "выполнено"},
        {"id": 2, "task": "Сделать домашнее задание по JSON", "status": "в процессе"},
        {"id": 3, "task": "Прочитать документацию по модулю json", "status": "не начата"}
    ]
}

with open(file_name, 'w', encoding='utf-8') as file:
    json.dump(initial_tasks, file, ensure_ascii=False, indent=4)

def load_tasks(filename):
    """Загружает задачи из JSON-файла"""
    with open(filename, 'r', encoding='utf-8') as file:
        return json.load(file)

def save_tasks(filename, data):
    """Сохраняет задачи в JSON-файл"""
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

def add_task(data, task_text, status="не начата"):
    """Добавляет новую задачу"""
    new_id = max([t["id"] for t in data["задачи"]], default=0) + 1
    new_task = {
        "id": new_id,
        "task": task_text,
        "status": status,
        "создана": datetime.now().strftime("%Y-%m-%d")
    }
    data["задачи"].append(new_task)
    print(f"✓ Добавлена задача: {task_text} (ID: {new_id})")
    return data

def update_task_status(data, task_id, new_status):
    """Обновляет статус задачи по ID"""
    for task in data["задачи"]:
        if task["id"] == task_id:
            old_status = task["status"]
            task["status"] = new_status
            print(f"✓ Обновлена задача ID={task_id}: {task['task']}")
            print(f"  Статус изменён: {old_status} -> {new_status}")
            return True
    print(f"❌ Задача с ID={task_id} не найдена")
    return False

def display_tasks(data, filter_status=None):
    """Отображает задачи с возможностью фильтрации"""
    tasks = data["задачи"]
    if filter_status:
        tasks = [t for t in tasks if t["status"] == filter_status]
    
    if not tasks:
        print("Задачи не найдены")
        return
    
    print("\n" + "="*70)
    print(f"{'ID':<5} {'Задача':<45} {'Статус':<15}")
    print("-"*70)
    
    for task in tasks:
        status_icon = "✅" if task["status"] == "выполнено" else "🔄" if task["status"] == "в процессе" else "⏳"
        print(f"{status_icon} {task['id']:<4} {task['task']:<45} {task['status']:<15}")
    
    print("="*70)

# Основная программа
print("="*70)
print("УПРАВЛЕНИЕ ЗАДАЧАМИ")
print("="*70)

# Загружаем данные
tasks_data = load_tasks(file_name)

# Показываем текущие задачи
display_tasks(tasks_data)

# Добавляем новую задачу
tasks_data = add_task(tasks_data, "Написать программу для обновления JSON", "не начата")

# Обновляем статус задачи
update_task_status(tasks_data, 2, "выполнено")

# Сохраняем изменения
save_tasks(file_name, tasks_data)

# Показываем обновлённый список
display_tasks(tasks_data)

print(f"\n✓ Данные сохранены в файл: {file_name}")
````

````python
# Вариант 3: С интерактивным управлением
import os
import json
from datetime import datetime

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "tasks.json")

os.makedirs(path, exist_ok=True)

# Создаём исходный файл
initial_tasks = {
    "задачи": [
        {"id": 1, "task": "Изучить работу с файлами в Python", "status": "выполнено", "приоритет": "высокий"},
        {"id": 2, "task": "Сделать домашнее задание по JSON", "status": "в процессе", "приоритет": "средний"},
        {"id": 3, "task": "Прочитать документацию по модулю json", "status": "не начата", "приоритет": "низкий"}
    ],
    "метаданные": {
        "версия": "1.0",
        "последнее_обновление": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
}

with open(file_name, 'w', encoding='utf-8') as file:
    json.dump(initial_tasks, file, ensure_ascii=False, indent=4)

def load_data():
    """Загружает данные из JSON-файла"""
    with open(file_name, 'r', encoding='utf-8') as file:
        return json.load(file)

def save_data(data):
    """Сохраняет данные в JSON-файл"""
    data["метаданные"]["последнее_обновление"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(file_name, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

def show_tasks(data):
    """Отображает все задачи"""
    print("\n" + "="*80)
    print(f"{'ID':<5} {'Задача':<45} {'Статус':<15} {'Приоритет':<10}")
    print("-"*80)
    
    for task in data["задачи"]:
        status_icon = "✅" if task["status"] == "выполнено" else "🔄" if task["status"] == "в процессе" else "⏳"
        print(f"{status_icon} {task['id']:<4} {task['task']:<45} {task['status']:<15} {task['приоритет']:<10}")
    
    print("="*80)
    print(f"Всего задач: {len(data['задачи'])}")
    print(f"Последнее обновление: {data['метаданные']['последнее_обновление']}")

def add_task_interactive(data):
    """Интерактивное добавление задачи"""
    print("\n--- ДОБАВЛЕНИЕ НОВОЙ ЗАДАЧИ ---")
    task_text = input("Введите описание задачи: ").strip()
    
    if not task_text:
        print("❌ Задача не может быть пустой!")
        return data
    
    print("\nВыберите статус:")
    print("1 - Не начата")
    print("2 - В процессе")
    print("3 - Выполнено")
    status_choice = input("Ваш выбор (1-3): ").strip()
    
    status_map = {"1": "не начата", "2": "в процессе", "3": "выполнено"}
    status = status_map.get(status_choice, "не начата")
    
    print("\nВыберите приоритет:")
    print("1 - Высокий")
    print("2 - Средний")
    print("3 - Низкий")
    priority_choice = input("Ваш выбор (1-3): ").strip()
    
    priority_map = {"1": "высокий", "2": "средний", "3": "низкий"}
    priority = priority_map.get(priority_choice, "средний")
    
    new_id = max([t["id"] for t in data["задачи"]], default=0) + 1
    new_task = {
        "id": new_id,
        "task": task_text,
        "status": status,
        "приоритет": priority
    }
    
    data["задачи"].append(new_task)
    print(f"\n✓ Задача добавлена (ID: {new_id})")
    return data

def update_status_interactive(data):
    """Интерактивное обновление статуса задачи"""
    print("\n--- ОБНОВЛЕНИЕ СТАТУСА ЗАДАЧИ ---")
    
    try:
        task_id = int(input("Введите ID задачи: ").strip())
    except ValueError:
        print("❌ Некорректный ID!")
        return data
    
    # Ищем задачу
    task = None
    for t in data["задачи"]:
        if t["id"] == task_id:
            task = t
            break
    
    if not task:
        print(f"❌ Задача с ID={task_id} не найдена!")
        return data
    
    print(f"\nТекущая задача: {task['task']}")
    print(f"Текущий статус: {task['status']}")
    
    print("\nВыберите новый статус:")
    print("1 - Не начата")
    print("2 - В процессе")
    print("3 - Выполнено")
    
    status_choice = input("Ваш выбор (1-3): ").strip()
    status_map = {"1": "не начата", "2": "в процессе", "3": "выполнено"}
    new_status = status_map.get(status_choice)
    
    if new_status:
        old_status = task["status"]
        task["status"] = new_status
        print(f"\n✓ Статус обновлён: {old_status} -> {new_status}")
    else:
        print("❌ Неверный выбор!")
    
    return data

# Основная программа
print("="*80)
print("📋 СИСТЕМА УПРАВЛЕНИЯ ЗАДАЧАМИ")
print("="*80)

# Загружаем данные
tasks_data = load_data()

# Показываем текущие задачи
show_tasks(tasks_data)

# Меню действий
while True:
    print("\n" + "-"*80)
    print("ДОСТУПНЫЕ ДЕЙСТВИЯ:")
    print("1 - Добавить новую задачу")
    print("2 - Изменить статус задачи")
    print("3 - Показать все задачи")
    print("4 - Сохранить и выйти")
    print("5 - Выйти без сохранения")
    
    choice = input("\nВаш выбор (1-5): ").strip()
    
    if choice == "1":
        tasks_data = add_task_interactive(tasks_data)
        save_data(tasks_data)
        show_tasks(tasks_data)
        
    elif choice == "2":
        tasks_data = update_status_interactive(tasks_data)
        save_data(tasks_data)
        show_tasks(tasks_data)
        
    elif choice == "3":
        show_tasks(tasks_data)
        
    elif choice == "4":
        save_data(tasks_data)
        print(f"\n✓ Данные сохранены в файл: {file_name}")
        print("До свидания! 👋")
        break
        
    elif choice == "5":
        print("\nВыход без сохранения изменений...")
        print("До свидания! 👋")
        break
        
    else:
        print("❌ Неверный выбор! Попробуйте снова.")
````

````python
# Вариант 5: С обработкой ошибок и информацией о файле
import os
import pickle
import random
from datetime import datetime

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "data.pkl")

os.makedirs(path, exist_ok=True)

def save_to_pickle(data, filename):
    """Сохраняет данные в pickle-файл с обработкой ошибок"""
    try:
        with open(filename, 'wb') as file:
            pickle.dump(data, file)
        
        file_size = os.path.getsize(filename)
        print(f"✓ Данные сохранены:")
        print(f"  Файл: {filename}")
        print(f"  Размер: {file_size} байт ({file_size / 1024:.2f} КБ)")
        print(f"  Время: {datetime.now().strftime('%H:%M:%S')}")
        return True
        
    except PermissionError:
        print(f"❌ Ошибка: Нет прав на запись в файл {filename}")
        return False
    except Exception as e:
        print(f"❌ Ошибка при сохранении: {e}")
        return False

def load_from_pickle(filename):
    """Загружает данные из pickle-файла с обработкой ошибок"""
    try:
        if not os.path.exists(filename):
            print(f"❌ Ошибка: Файл {filename} не найден")
            return None
            
        with open(filename, 'rb') as file:
            data = pickle.load(file)
        
        file_size = os.path.getsize(filename)
        print(f"✓ Данные загружены:")
        print(f"  Файл: {filename}")
        print(f"  Размер: {file_size} байт ({file_size / 1024:.2f} КБ)")
        print(f"  Время: {datetime.now().strftime('%H:%M:%S')}")
        return data
        
    except pickle.UnpicklingError:
        print(f"❌ Ошибка: Файл {filename} повреждён или имеет неверный формат")
        return None
    except Exception as e:
        print(f"❌ Ошибка при загрузке: {e}")
        return None

# Создаём список случайных чисел
random_numbers = [random.randint(1, 100) for _ in range(10)]
print("="*60)
print("СОЗДАН СПИСОК СЛУЧАЙНЫХ ЧИСЕЛ:")
print("="*60)
print(f"Список: {random_numbers}")
print(f"Сумма: {sum(random_numbers)}")
print(f"Среднее: {sum(random_numbers) / len(random_numbers):.2f}\n")

# Сохраняем
print("СОХРАНЕНИЕ:")
print("-"*40)
save_to_pickle(random_numbers, file_name)

# Загружаем
print("\nЗАГРУЗКА:")
print("-"*40)
loaded_numbers = load_from_pickle(file_name)

# Выводим результат
if loaded_numbers is not None:
    print("\n" + "="*60)
    print("РЕЗУЛЬТАТ:")
    print("="*60)
    print(f"Исходный список:    {random_numbers}")
    print(f"Загруженный список: {loaded_numbers}")
    print(f"Списки идентичны:   {random_numbers == loaded_numbers}")
    
    if random_numbers == loaded_numbers:
        print("\n✅ Операция выполнена успешно!")
    else:
        print("\n❌ Обнаружено несоответствие данных!")
````

````python
# Вариант 2: С рекурсивным обходом всех поддиректорий
import os
from datetime import datetime

def get_size_format(size):
    """Форматирует размер в удобочитаемый вид"""
    for unit in ['B', 'KB', 'MB', 'GB']:
        if size < 1024.0:
            return f"{size:.2f} {unit}"
        size /= 1024.0
    return f"{size:.2f} TB"

def scan_directory(path, indent=0):
    """Рекурсивно сканирует директорию"""
    try:
        items = os.listdir(path)
        
        for item in sorted(items):
            item_path = os.path.join(path, item)
            prefix = "  " * indent
            
            if os.path.isdir(item_path):
                print(f"{prefix}📁 {item}/")
                scan_directory(item_path, indent + 1)
            else:
                size = os.path.getsize(item_path)
                size_str = get_size_format(size)
                print(f"{prefix}📄 {item} ({size_str})")
                
    except PermissionError:
        print(f"{'  ' * indent}⚠️ Нет доступа к папке")

# Текущая директория
current_dir = "."

print("="*70)
print(f"РЕКУРСИВНЫЙ ОБХОД ДИРЕКТОРИИ: {os.path.abspath(current_dir)}")
print("="*70)

scan_directory(current_dir)

print("\n" + "="*70)
print("Обход завершён")
print("="*70)
````

````python
# Вариант 3: С подробной информацией (дата, права доступа)
import os
import stat
from datetime import datetime

def format_permissions(mode):
    """Форматирует права доступа в читаемый вид"""
    perms = []
    
    # Владелец
    perms.append('r' if mode & stat.S_IRUSR else '-')
    perms.append('w' if mode & stat.S_IWUSR else '-')
    perms.append('x' if mode & stat.S_IXUSR else '-')
    
    # Группа
    perms.append('r' if mode & stat.S_IRGRP else '-')
    perms.append('w' if mode & stat.S_IWGRP else '-')
    perms.append('x' if mode & stat.S_IXGRP else '-')
    
    # Остальные
    perms.append('r' if mode & stat.S_IROTH else '-')
    perms.append('w' if mode & stat.S_IWOTH else '-')
    perms.append('x' if mode & stat.S_IXOTH else '-')
    
    return ''.join(perms)

def format_size(size):
    """Форматирует размер файла"""
    if size < 1024:
        return f"{size} B"
    elif size < 1024 * 1024:
        return f"{size / 1024:.1f} KB"
    elif size < 1024 * 1024 * 1024:
        return f"{size / (1024 * 1024):.1f} MB"
    else:
        return f"{size / (1024 * 1024 * 1024):.1f} GB"

# Текущая директория
current_dir = "."

print("="*100)
print(f"ДЕТАЛЬНЫЙ ОБХОД ДИРЕКТОРИИ: {os.path.abspath(current_dir)}")
print("="*100)

print(f"{'Имя':<40} {'Тип':<10} {'Размер':<12} {'Изменён':<20} {'Права':<10}")
print("-"*100)

items = os.listdir(current_dir)

for item in sorted(items):
    item_path = os.path.join(current_dir, item)
    
    # Получаем информацию
    if os.path.isdir(item_path):
        item_type = "Папка"
        size = "-"
    else:
        item_type = "Файл"
        size = format_size(os.path.getsize(item_path))
    
    # Время последнего изменения
    mtime = os.path.getmtime(item_path)
    modified = datetime.fromtimestamp(mtime).strftime("%Y-%m-%d %H:%M:%S")
    
    # Права доступа
    mode = os.stat(item_path).st_mode
    permissions = format_permissions(mode)
    
    print(f"{item:<40} {item_type:<10} {size:<12} {modified:<20} {permissions:<10}")

print("-"*100)

# Статистика
total_files = sum(1 for item in items if os.path.isfile(os.path.join(current_dir, item)))
total_dirs = sum(1 for item in items if os.path.isdir(os.path.join(current_dir, item)))

print(f"\nСтатистика:")
print(f"  Всего: {len(items)} элементов")
print(f"  Папок: {total_dirs}")
print(f"  Файлов: {total_files}")
````

````python
# Вариант 5: С сохранением результатов в файл
import os
from datetime import datetime

def find_txt_files(directory):
    """Находит все .txt файлы в директории и поддиректориях"""
    txt_files = []
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.txt'):
                full_path = os.path.join(root, file)
                size = os.path.getsize(full_path)
                modified = os.path.getmtime(full_path)
                txt_files.append({
                    'path': full_path,
                    'name': file,
                    'size': size,
                    'modified': modified
                })
    
    # Сортируем по пути
    txt_files.sort(key=lambda x: x['path'])
    return txt_files

def save_results_to_file(txt_files, output_file):
    """Сохраняет результаты поиска в файл"""
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("="*80 + "\n")
        f.write(f"РЕЗУЛЬТАТЫ ПОИСКА .txt ФАЙЛОВ\n")
        f.write(f"Дата: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write("="*80 + "\n\n")
        
        if txt_files:
            f.write(f"Найдено файлов: {len(txt_files)}\n\n")
            
            for i, file_info in enumerate(txt_files, 1):
                size_str = f"{file_info['size']} B"
                if file_info['size'] >= 1024:
                    size_str = f"{file_info['size'] / 1024:.2f} KB"
                if file_info['size'] >= 1024 * 1024:
                    size_str = f"{file_info['size'] / (1024 * 1024):.2f} MB"
                
                modified_str = datetime.fromtimestamp(file_info['modified']).strftime("%Y-%m-%d %H:%M:%S")
                
                f.write(f"{i:3}. {file_info['path']}\n")
                f.write(f"     Имя: {file_info['name']}\n")
                f.write(f"     Размер: {size_str}\n")
                f.write(f"     Изменён: {modified_str}\n\n")
        else:
            f.write("Файлы с расширением .txt не найдены.\n")
        
        f.write("="*80 + "\n")

# Задаём директорию для поиска
path = f"D:/Python/Pyton-2026/files/"
search_dir = os.path.join(path, "project")
output_file = os.path.join(path, "search_results.txt")

if not os.path.exists(search_dir):
    print(f"⚠️ Директория {search_dir} не существует!")
    print("Сначала выполните задание 6.2 для создания структуры папок.")
    exit()

print("="*80)
print(f"ПОИСК .txt ФАЙЛОВ В: {search_dir}")
print("="*80)

# Выполняем поиск
txt_files = find_txt_files(search_dir)

# Сохраняем результаты
save_results_to_file(txt_files, output_file)

# Выводим на экран
if txt_files:
    print(f"\n✅ Найдено {len(txt_files)} файлов:\n")
    for i, file_info in enumerate(txt_files, 1):
        size_str = f"{file_info['size']} B"
        if file_info['size'] >= 1024:
            size_str = f"{file_info['size'] / 1024:.2f} KB"
        
        print(f"{i:2}. {file_info['path']} ({size_str})")
else:
    print(f"\n❌ Файлы с расширением .txt не найдены")

print(f"\n📄 Результаты сохранены в файл: {output_file}")
print("="*80)
````

````python
# Вариант 2: С расширенными возможностями (редактирование, поиск)
import os
from datetime import datetime

path = f"D:/Python/Pyton-2026/files/"
file_name = os.path.join(path, "journal.txt")

os.makedirs(path, exist_ok=True)

def load_notes():
    """Загружает заметки из файла"""
    if not os.path.exists(file_name):
        return []
    
    with open(file_name, 'r', encoding='utf-8') as file:
        notes = file.readlines()
    
    # Удаляем пустые строки
    return [note.strip() for note in notes if note.strip()]

def save_notes(notes):
    """Сохраняет заметки в файл"""
    with open(file_name, 'w', encoding='utf-8') as file:
        for note in notes:
            file.write(f"{note}\n")

def add_note():
    """Добавляет новую заметку"""
    print("\n--- ДОБАВЛЕНИЕ ЗАМЕТКИ ---")
    note_text = input("Введите текст заметки: ").strip()
    
    if not note_text:
        print("❌ Заметка не может быть пустой!")
        return
    
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    note_entry = f"[{timestamp}] {note_text}"
    
    notes = load_notes()
    notes.append(note_entry)
    save_notes(notes)
    
    print(f"✓ Заметка добавлена")

def view_notes():
    """Просматривает все заметки"""
    print("\n--- ПРОСМОТР ЗАМЕТОК ---")
    
    notes = load_notes()
    
    if not notes:
        print("Журнал пуст. Заметок пока нет.")
        return
    
    print(f"\n{'='*70}")
    print(f"ВСЕГО ЗАМЕТОК: {len(notes)}")
    print(f"{'='*70}")
    
    for i, note in enumerate(notes, 1):
        print(f"{i:3}. {note}")

def edit_note():
    """Редактирует существующую заметку"""
    print("\n--- РЕДАКТИРОВАНИЕ ЗАМЕТКИ ---")
    
    notes = load_notes()
    
    if not notes:
        print("Журнал пуст. Нечего редактировать.")
        return
    
    # Показываем заметки
    for i, note in enumerate(notes, 1):
        print(f"{i:3}. {note}")
    
    try:
        choice = int(input("\nВыберите номер заметки для редактирования: ").strip())
        if choice < 1 or choice > len(notes):
            print("❌ Неверный номер!")
            return
        
        print(f"\nТекущая заметка: {notes[choice-1]}")
        new_text = input("Введите новый текст заметки: ").strip()
        
        if not new_text:
            print("❌ Заметка не может быть пустой!")
            return
        
        # Сохраняем старую дату
        old_timestamp = notes[choice-1].split(']')[0] + "]"
        notes[choice-1] = f"{old_timestamp} {new_text}"
        
        save_notes(notes)
        print(f"✓ Заметка {choice} отредактирована")
        
    except ValueError:
        print("❌ Введите корректный номер!")

def search_notes():
    """Поиск заметок по ключевому слову"""
    print("\n--- ПОИСК ЗАМЕТОК ---")
    
    keyword = input("Введите ключевое слово для поиска: ").strip().lower()
    
    if not keyword:
        print("❌ Введите ключевое слово!")
        return
    
    notes = load_notes()
    
    if not notes:
        print("Журнал пуст.")
        return
    
    found = []
    for note in notes:
        if keyword in note.lower():
            found.append(note)
    
    if found:
        print(f"\n🔍 Найдено {len(found)} заметок с ключевым словом '{keyword}':")
        print(f"{'='*70}")
        for i, note in enumerate(found, 1):
            print(f"{i:3}. {note}")
    else:
        print(f"❌ Заметки с ключевым словом '{keyword}' не найдены")

def delete_all_notes():
    """Удаляет все заметки"""
    print("\n--- УДАЛЕНИЕ ВСЕХ ЗАМЕТОК ---")
    
    notes = load_notes()
    
    if not notes:
        print("Журнал уже пуст.")
        return
    
    print(f"⚠️ ВНИМАНИЕ! Будет удалено {len(notes)} заметок.")
    confirm = input("Вы уверены? (y/n): ").strip().lower()
    
    if confirm == 'y':
        save_notes([])
        print("✓ Все заметки удалены. Журнал очищен.")
    else:
        print("Операция отменена.")

def delete_one_note():
    """Удаляет одну заметку"""
    print("\n--- УДАЛЕНИЕ ЗАМЕТКИ ---")
    
    notes = load_notes()
    
    if not notes:
        print("Журнал пуст. Нечего удалять.")
        return
    
    # Показываем заметки
    for i, note in enumerate(notes, 1):
        print(f"{i:3}. {note}")
    
    try:
        choice = int(input("\nВыберите номер заметки для удаления: ").strip())
        if choice < 1 or choice > len(notes):
            print("❌ Неверный номер!")
            return
        
        deleted_note = notes.pop(choice - 1)
        save_notes(notes)
        print(f"✓ Заметка удалена: {deleted_note[:50]}...")
        
    except ValueError:
        print("❌ Введите корректный номер!")

def show_menu():
    """Показывает главное меню"""
    print("\n" + "="*70)
    print("📝 ЖУРНАЛ ОПЕРАЦИЙ")
    print("="*70)
    print("1. Добавить заметку")
    print("2. Просмотреть все заметки")
    print("3. Редактировать заметку")
    print("4. Найти заметку")
    print("5. Удалить одну заметку")
    print("6. Удалить все заметки")
    print("7. Выход")
    print("="*70)

# Основная программа
print("Добро пожаловать в программу Журнал операций!")

while True:
    show_menu()
    choice = input("\nВыберите действие (1-7): ").strip()
    
    if choice == '1':
        add_note()
    elif choice == '2':
        view_notes()
    elif choice == '3':
        edit_note()
    elif choice == '4':
        search_notes()
    elif choice == '5':
        delete_one_note()
    elif choice == '6':
        delete_all_notes()
    elif choice == '7':
        print("\nДо свидания! 👋")
        break
    else:
        print("❌ Неверный выбор! Пожалуйста, выберите 1-7.")
````

````python
# Вариант 2: С созданием подпапок по дате
import os
import shutil
from datetime import datetime

def backup_txt_files(source_dir, backup_root):
    """
    Создаёт резервные копии всех .txt файлов с организацией по датам
    
    Args:
        source_dir: исходная папка
        backup_root: корневая папка для бэкапов
    """
    # Создаём папку с текущей датой
    current_date = datetime.now().strftime("%Y-%m-%d")
    backup_dir = os.path.join(backup_root, current_date)
    os.makedirs(backup_dir, exist_ok=True)
    
    print(f"Папка для бэкапа: {backup_dir}")
    print("-"*70)
    
    copied_files = []
    errors = []
    
    for root, dirs, files in os.walk(source_dir):
        for file_name in files:
            if file_name.endswith('.txt'):
                source_path = os.path.join(root, file_name)
                
                # Сохраняем относительный путь для сохранения структуры
                rel_path = os.path.relpath(source_path, source_dir)
                backup_path = os.path.join(backup_dir, rel_path)
                
                # Создаём подпапки, если нужно
                os.makedirs(os.path.dirname(backup_path), exist_ok=True)
                
                try:
                    # Копируем файл
                    shutil.copy2(source_path, backup_path)
                    copied_files.append((rel_path, backup_path))
                    
                    size = os.path.getsize(source_path)
                    size_str = f"{size} B" if size < 1024 else f"{size/1024:.2f} KB"
                    print(f"✓ {rel_path} -> {os.path.basename(backup_path)} ({size_str})")
                    
                except Exception as e:
                    errors.append((rel_path, str(e)))
                    print(f"❌ Ошибка при копировании {rel_path}: {e}")
    
    return copied_files, errors

# Задаём пути
path = f"D:/Python/Pyton-2026/files/"
source_dir = os.path.join(path, "source")
backup_root = os.path.join(path, "backups")

# Создаём папки
os.makedirs(source_dir, exist_ok=True)
os.makedirs(backup_root, exist_ok=True)

# Создаём тестовые файлы
sample_files = [
    ("file1.txt", "Содержимое первого файла"),
    ("file2.txt", "Содержимое второго файла"),
    ("subfolder/file3.txt", "Файл в подпапке"),
    ("subfolder/notes.txt", "Заметки в подпапке")
]

print("="*70)
print("СОЗДАНИЕ ТЕСТОВЫХ ФАЙЛОВ:")
print("="*70)

for file_name, content in sample_files:
    file_path = os.path.join(source_dir, file_name)
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(content)
    print(f"✓ Создан: {file_path}")

print("\n" + "="*70)
print("РЕЗЕРВНОЕ КОПИРОВАНИЕ:")
print("="*70)

# Выполняем бэкап
copied, errors = backup_txt_files(source_dir, backup_root)

# Выводим статистику
print("\n" + "="*70)
print("СТАТИСТИКА:")
print("="*70)
print(f"Скопировано файлов: {len(copied)}")
print(f"Ошибок: {len(errors)}")
print(f"Папка назначения: {os.path.join(backup_root, datetime.now().strftime('%Y-%m-%d'))}")
````

````python
# Вариант 3: С добавлением временной метки и проверкой на дубликаты
import os
import shutil
from datetime import datetime

def backup_with_timestamp(source_dir, backup_dir):
    """
    Создаёт резервные копии .txt файлов с временной меткой,
    предотвращая перезапись существующих копий
    """
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # Создаём папку для бэкапа
    backup_folder = os.path.join(backup_dir, f"backup_{timestamp}")
    os.makedirs(backup_folder, exist_ok=True)
    
    print(f"Папка бэкапа: {backup_folder}")
    print("-"*70)
    
    stats = {
        "copied": 0,
        "skipped": 0,
        "errors": 0,
        "total_size": 0
    }
    
    for root, dirs, files in os.walk(source_dir):
        for file_name in files:
            if file_name.endswith('.txt'):
                source_path = os.path.join(root, file_name)
                
                # Формируем имя с временной меткой
                name_without_ext = os.path.splitext(file_name)[0]
                backup_name = f"{name_without_ext}_{timestamp}.txt"
                backup_path = os.path.join(backup_folder, backup_name)
                
                # Проверяем, существует ли уже такая копия
                if os.path.exists(backup_path):
                    print(f"⚠️ Пропущен {file_name}: копия уже существует")
                    stats["skipped"] += 1
                    continue
                
                try:
                    # Копируем с сохранением метаданных
                    shutil.copy2(source_path, backup_path)
                    
                    size = os.path.getsize(source_path)
                    stats["copied"] += 1
                    stats["total_size"] += size
                    
                    size_str = f"{size} B" if size < 1024 else f"{size/1024:.2f} KB"
                    print(f"✓ {file_name} -> {backup_name} ({size_str})")
                    
                except Exception as e:
                    stats["errors"] += 1
                    print(f"❌ Ошибка при копировании {file_name}: {e}")
    
    return stats, backup_folder

# Задаём пути
path = f"D:/Python/Pyton-2026/files/"
source_dir = os.path.join(path, "source")
backup_dir = os.path.join(path, "backup")

os.makedirs(source_dir, exist_ok=True)
os.makedirs(backup_dir, exist_ok=True)

# Создаём тестовые файлы
test_files = [
    ("data1.txt", "Данные 1"),
    ("data2.txt", "Данные 2"),
    ("report.txt", "Отчёт")
]

print("="*70)
print("СОЗДАНИЕ ТЕСТОВЫХ ФАЙЛОВ:")
print("="*70)

for file_name, content in test_files:
    file_path = os.path.join(source_dir, file_name)
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(content)
    print(f"✓ Создан: {file_path}")

print("\n" + "="*70)
print("ВЫПОЛНЕНИЕ РЕЗЕРВНОГО КОПИРОВАНИЯ:")
print("="*70)

# Выполняем первый бэкап
stats1, backup_folder1 = backup_with_timestamp(source_dir, backup_dir)

# Выполняем второй бэкап (должен пропустить уже скопированные файлы)
print("\n" + "="*70)
print("ПОВТОРНОЕ КОПИРОВАНИЕ (проверка на дубликаты):")
print("="*70)

stats2, backup_folder2 = backup_with_timestamp(source_dir, backup_dir)

# Выводим статистику
print("\n" + "="*70)
print("СТАТИСТИКА ПЕРВОГО БЭКАПА:")
print("="*70)
print(f"Скопировано: {stats1['copied']}")
print(f"Пропущено: {stats1['skipped']}")
print(f"Ошибок: {stats1['errors']}")
print(f"Общий размер: {stats1['total_size']} B")

print("\n" + "="*70)
print("СТАТИСТИКА ВТОРОГО БЭКАПА:")
print("="*70)
print(f"Скопировано: {stats2['copied']}")
print(f"Пропущено: {stats2['skipped']}")
print(f"Ошибок: {stats2['errors']}")

print("\n✅ Резервное копирование завершено!")
````

````python
# Вариант 4: С использованием pathlib и логированием
from pathlib import Path
from datetime import datetime
import shutil

def backup_txt_files_pathlib(source_dir, backup_dir):
    """
    Создаёт резервные копии .txt файлов с использованием pathlib
    """
    source_path = Path(source_dir)
    backup_path = Path(backup_dir)
    
    # Создаём папки
    source_path.mkdir(parents=True, exist_ok=True)
    backup_path.mkdir(parents=True, exist_ok=True)
    
    # Получаем текущую дату
    date_str = datetime.now().strftime("%Y%m%d")
    
    # Создаём лог-файл
    log_file = backup_path / f"backup_log_{date_str}.txt"
    
    print("="*70)
    print(f"РЕЗЕРВНОЕ КОПИРОВАНИЕ .txt ФАЙЛОВ")
    print(f"Источник: {source_path.absolute()}")
    print(f"Назначение: {backup_path.absolute()}")
    print("="*70)
    
    # Поиск всех .txt файлов
    txt_files = list(source_path.rglob("*.txt"))
    
    if not txt_files:
        print("❌ .txt файлы не найдены")
        return
    
    print(f"\nНайдено .txt файлов: {len(txt_files)}")
    print("-"*70)
    
    backup_stats = {
        "copied": [],
        "failed": [],
        "total_size": 0
    }
    
    for file_path in txt_files:
        # Формируем имя для бэкапа
        backup_name = f"{file_path.stem}_{date_str}{file_path.suffix}"
        backup_file = backup_path / backup_name
        
        try:
            # Копируем файл
            shutil.copy2(file_path, backup_file)
            
            size = file_path.stat().st_size
            backup_stats["copied"].append({
                "source": file_path.name,
                "backup": backup_name,
                "size": size,
                "path": str(file_path)
            })
            backup_stats["total_size"] += size
            
            size_str = f"{size} B" if size < 1024 else f"{size/1024:.2f} KB"
            print(f"✓ {file_path.name} -> {backup_name} ({size_str})")
            
        except Exception as e:
            backup_stats["failed"].append({
                "source": file_path.name,
                "error": str(e)
            })
            print(f"❌ Ошибка при копировании {file_path.name}: {e}")
    
    # Записываем лог
    with open(log_file, 'w', encoding='utf-8') as f:
        f.write(f"ЛОГ РЕЗЕРВНОГО КОПИРОВАНИЯ\n")
        f.write(f"Дата: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Источник: {source_path.absolute()}\n")
        f.write(f"Назначение: {backup_path.absolute()}\n")
        f.write("="*70 + "\n\n")
        
        f.write(f"Успешно скопировано: {len(backup_stats['copied'])}\n")
        for item in backup_stats["copied"]:
            f.write(f"  - {item['source']} -> {item['backup']} ({item['size']} B)\n")
        
        if backup_stats["failed"]:
            f.write(f"\nОшибок: {len(backup_stats['failed'])}\n")
            for item in backup_stats["failed"]:
                f.write(f"  - {item['source']}: {item['error']}\n")
        
        f.write(f"\nОбщий размер скопированных файлов: {backup_stats['total_size']} B\n")
    
    # Выводим статистику
    print("\n" + "="*70)
    print("СТАТИСТИКА:")
    print("="*70)
    print(f"Успешно скопировано: {len(backup_stats['copied'])}")
    print(f"Ошибок: {len(backup_stats['failed'])}")
    print(f"Общий размер: {backup_stats['total_size']} B")
    print(f"Лог сохранён: {log_file.name}")
    
    return backup_stats

# Задаём пути
path = Path("D:/Python/Pyton-2026/files/")
source_dir = path / "source"
backup_dir = path / "backup"

# Создаём тестовые файлы
print("="*70)
print("СОЗДАНИЕ ТЕСТОВЫХ ФАЙЛОВ:")
print("="*70)

source_dir.mkdir(parents=True, exist_ok=True)

test_files = {
    "doc1.txt": "Первый документ\nСодержимое",
    "doc2.txt": "Второй документ\nБольше данных",
    "report.txt": "Отчёт за март\nИтоги работы",
    "readme.txt": "Инструкция по использованию"
}

for name, content in test_files.items():
    file_path = source_dir / name
    file_path.write_text(content, encoding='utf-8')
    print(f"✓ Создан: {file_path}")

# Выполняем бэкап
backup_stats = backup_txt_files_pathlib(source_dir, backup_dir)

print("\n✅ Резервное копирование завершено!")
````

````python
# Вариант 5: С архивированием в ZIP
import os
import zipfile
from datetime import datetime

def backup_to_zip(source_dir, backup_dir):
    """
    Создаёт резервную копию всех .txt файлов в ZIP-архив
    """
    # Создаём папку для бэкапов
    os.makedirs(backup_dir, exist_ok=True)
    
    # Формируем имя архива с датой
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    zip_name = f"backup_{timestamp}.zip"
    zip_path = os.path.join(backup_dir, zip_name)
    
    print("="*70)
    print(f"СОЗДАНИЕ ZIP-АРХИВА С .txt ФАЙЛАМИ")
    print(f"Источник: {source_dir}")
    print(f"Архив: {zip_path}")
    print("="*70)
    
    # Собираем все .txt файлы
    txt_files = []
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            if file.endswith('.txt'):
                full_path = os.path.join(root, file)
                # Сохраняем относительный путь для структуры в архиве
                rel_path = os.path.relpath(full_path, source_dir)
                txt_files.append((full_path, rel_path))
    
    if not txt_files:
        print("❌ .txt файлы не найдены")
        return None
    
    print(f"\nНайдено .txt файлов: {len(txt_files)}")
    print("-"*70)
    
    # Создаём ZIP-архив
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for full_path, rel_path in txt_files:
            # Добавляем файл в архив
            zipf.write(full_path, rel_path)
            
            size = os.path.getsize(full_path)
            size_str = f"{size} B" if size < 1024 else f"{size/1024:.2f} KB"
            print(f"✓ Добавлен: {rel_path} ({size_str})")
    
    # Получаем размер архива
    archive_size = os.path.getsize(zip_path)
    archive_size_str = f"{archive_size} B" if archive_size < 1024 else f"{archive_size/1024:.2f} KB"
    
    print("\n" + "="*70)
    print("СТАТИСТИКА:")
    print("="*70)
    print(f"Файлов в архиве: {len(txt_files)}")
    print(f"Размер архива: {archive_size_str}")
    print(f"Архив сохранён: {zip_path}")
    
    return zip_path

# Задаём пути
path = f"D:/Python/Pyton-2026/files/"
source_dir = os.path.join(path, "source")
backup_dir = os.path.join(path, "backup_zip")

os.makedirs(source_dir, exist_ok=True)

# Создаём тестовые файлы
test_files = [
    ("file1.txt", "Содержимое 1"),
    ("file2.txt", "Содержимое 2"),
    ("subfolder/file3.txt", "Файл в подпапке"),
    ("subfolder/data.txt", "Данные")
]

print("="*70)
print("СОЗДАНИЕ ТЕСТОВЫХ ФАЙЛОВ:")
print("="*70)

for file_name, content in test_files:
    file_path = os.path.join(source_dir, file_name)
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(content)
    print(f"✓ Создан: {file_path}")

# Создаём ZIP-архив
zip_path = backup_to_zip(source_dir, backup_dir)

if zip_path:
    print("\n✅ Резервное копирование в ZIP завершено!")
````

````python
# Вариант 2: С поддержкой разных разделителей и кодировок
import os
import csv
import json
from datetime import datetime

class CSVToJSONConverter:
    """Класс для конвертации CSV в JSON с логированием"""
    
    def __init__(self, input_file, output_file, log_file, delimiter=',', encoding='utf-8'):
        self.input_file = input_file
        self.output_file = output_file
        self.log_file = log_file
        self.delimiter = delimiter
        self.encoding = encoding
        self.stats = {
            "start_time": None,
            "end_time": None,
            "rows_read": 0,
            "rows_success": 0,
            "rows_error": 0,
            "errors": []
        }
    
    def write_log(self, message, level="INFO"):
        """Записывает сообщение в лог"""
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open(self.log_file, 'a', encoding='utf-8') as f:
            f.write(f"[{timestamp}] [{level}] {message}\n")
    
    def convert_value(self, value):
        """Преобразует значение в соответствующий тип"""
        if not value or value.strip() == '':
            return None
        
        value = value.strip()
        
        # Проверяем булевы значения
        if value.lower() in ('true', 'yes', '1'):
            return True
        if value.lower() in ('false', 'no', '0'):
            return False
        
        # Проверяем числа
        try:
            if '.' in value:
                return float(value)
            else:
                return int(value)
        except ValueError:
            pass
        
        # Возвращаем как строку
        return value
    
    def read_csv(self):
        """Читает CSV-файл и возвращает данные"""
        data = []
        headers = []
        
        self.write_log(f"Начало чтения CSV-файла (разделитель: '{self.delimiter}', кодировка: {self.encoding})")
        
        try:
            with open(self.input_file, 'r', encoding=self.encoding) as csvfile:
                reader = csv.reader(csvfile, delimiter=self.delimiter)
                
                # Читаем заголовки
                headers = next(reader)
                self.write_log(f"Заголовки ({len(headers)}): {headers}")
                
                # Читаем строки
                for row_num, row in enumerate(reader, start=2):
                    self.stats["rows_read"] += 1
                    
                    if not any(row):  # Пропускаем полностью пустые строки
                        continue
                    
                    try:
                        # Создаём словарь для строки
                        row_dict = {}
                        for i, value in enumerate(row):
                            if i < len(headers):
                                row_dict[headers[i]] = self.convert_value(value)
                            else:
                                # Если полей больше, чем заголовков
                                row_dict[f"extra_field_{i+1}"] = self.convert_value(value)
                        
                        data.append(row_dict)
                        self.stats["rows_success"] += 1
                        
                    except Exception as e:
                        self.stats["rows_error"] += 1
                        error_msg = f"Строка {row_num}: {e}"
                        self.stats["errors"].append(error_msg)
                        self.write_log(error_msg, "ERROR")
            
            self.write_log(f"Чтение завершено. Успешно: {self.stats['rows_success']}, Ошибок: {self.stats['rows_error']}")
            
        except Exception as e:
            self.write_log(f"Ошибка чтения CSV: {e}", "ERROR")
            raise
        
        return headers, data
    
    def save_json(self, headers, data):
        """Сохраняет данные в JSON-файл"""
        self.write_log("Начало сохранения в JSON")
        
        json_data = {
            "metadata": {
                "source_file": os.path.basename(self.input_file),
                "conversion_date": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "total_rows": self.stats["rows_success"],
                "total_fields": len(headers),
                "delimiter": self.delimiter,
                "encoding": self.encoding,
                "statistics": {
                    "rows_read": self.stats["rows_read"],
                    "rows_success": self.stats["rows_success"],
                    "rows_error": self.stats["rows_error"]
                }
            },
            "headers": headers,
            "data": data
        }
        
        # Добавляем ошибки в метаданные, если есть
        if self.stats["errors"]:
            json_data["metadata"]["errors"] = self.stats["errors"]
        
        try:
            with open(self.output_file, 'w', encoding='utf-8') as jsonfile:
                json.dump(json_data, jsonfile, ensure_ascii=False, indent=2)
            
            file_size = os.path.getsize(self.output_file)
            self.write_log(f"JSON-файл сохранён. Размер: {file_size} байт")
            
        except Exception as e:
            self.write_log(f"Ошибка сохранения JSON: {e}", "ERROR")
            raise
        
        return file_size
    
    def convert(self):
        """Выполняет конвертацию"""
        self.stats["start_time"] = datetime.now()
        self.write_log("="*50)
        self.write_log("НАЧАЛО КОНВЕРТАЦИИ")
        self.write_log(f"Входной файл: {self.input_file}")
        self.write_log(f"Выходной файл: {self.output_file}")
        
        try:
            # Чтение CSV
            headers, data = self.read_csv()
            
            # Сохранение JSON
            file_size = self.save_json(headers, data)
            
            # Завершение
            self.stats["end_time"] = datetime.now()
            duration = (self.stats["end_time"] - self.stats["start_time"]).total_seconds()
            
            # Записываем итоги
            self.write_log("-"*50)
            self.write_log("ИТОГИ КОНВЕРТАЦИИ:")
            self.write_log(f"Обработано строк: {self.stats['rows_success']}")
            self.write_log(f"Количество полей: {len(headers)}")
            self.write_log(f"Размер JSON-файла: {file_size} байт")
            self.write_log(f"Время выполнения: {duration:.3f} сек")
            self.write_log("КОНВЕРТАЦИЯ ЗАВЕРШЕНА УСПЕШНО")
            self.write_log("="*50)
            
            return True
            
        except Exception as e:
            self.write_log(f"КОНВЕРТАЦИЯ ПРЕРВАНА: {e}", "ERROR")
            self.write_log("="*50)
            return False

# Задаём пути
path = f"D:/Python/Pyton-2026/files/"
input_file = os.path.join(path, "data.csv")
output_file = os.path.join(path, "data.json")
log_file = os.path.join(path, "conversion_log.txt")

# Создаём тестовый CSV-файл
csv_data = """Имя;Возраст;Город;Профессия;Зарплата
Иван Петров;28;Москва;Разработчик;85000
Анна Сидорова;25;Санкт-Петербург;Дизайнер;65000
Сергей Иванов;32;Новосибирск;Аналитик;90000
Мария Смирнова;29;Екатеринбург;Менеджер;75000
"""

with open(input_file, 'w', encoding='utf-8') as f:
    f.write(csv_data)

# Создаём конвертер и выполняем конвертацию
converter = CSVToJSONConverter(
    input_file=input_file,
    output_file=output_file,
    log_file=log_file,
    delimiter=';',  # Используем точку с запятой как разделитель
    encoding='utf-8'
)

success = converter.convert()

if success:
    print("\n✅ Конвертация выполнена успешно!")
    print(f"JSON-файл: {output_file}")
    print(f"Лог-файл: {log_file}")
    
    # Показываем лог
    print("\n" + "="*70)
    print("ЛОГ-ФАЙЛ:")
    print("="*70)
    with open(log_file, 'r', encoding='utf-8') as f:
        print(f.read())
else:
    print("\n❌ Конвертация не выполнена!")
````

````python
# Вариант 3: С поддержкой больших файлов (построчная обработка)
import os
import csv
import json
from datetime import datetime

def convert_csv_to_json_large(input_file, output_file, log_file, chunk_size=1000):
    """
    Конвертирует большой CSV-файл в JSON с построчной обработкой
    """
    start_time = datetime.now()
    
    def write_log(message):
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open(log_file, 'a', encoding='utf-8') as f:
            f.write(f"[{timestamp}] {message}\n")
    
    write_log("="*50)
    write_log("НАЧАЛО КОНВЕРТАЦИИ (построчная обработка)")
    write_log(f"Входной файл: {input_file}")
    write_log(f"Выходной файл: {output_file}")
    
    try:
        # Открываем JSON-файл для записи
        with open(output_file, 'w', encoding='utf-8') as jsonfile:
            jsonfile.write('{\n  "data": [\n')
            
            row_count = 0
            first_row = True
            
            with open(input_file, 'r', encoding='utf-8') as csvfile:
                reader = csv.reader(csvfile)
                headers = next(reader)
                write_log(f"Заголовки: {headers}")
                
                for row_num, row in enumerate(reader, start=2):
                    if not any(row):
                        continue
                    
                    # Создаём словарь для строки
                    row_dict = {}
                    for i, value in enumerate(row):
                        if i < len(headers):
                            # Пробуем преобразовать числовые значения
                            try:
                                if value and value.replace('.', '').replace('-', '').isdigit():
                                    if '.' in value:
                                        row_dict[headers[i]] = float(value)
                                    else:
                                        row_dict[headers[i]] = int(value)
                                else:
                                    row_dict[headers[i]] = value
                            except:
                                row_dict[headers[i]] = value
                    
                    # Записываем в JSON
                    if not first_row:
                        jsonfile.write(',\n')
                    first_row = False
                    
                    json.dump(row_dict, jsonfile, ensure_ascii=False, indent=4)
                    row_count += 1
                    
                    # Логируем прогресс
                    if row_count % chunk_size == 0:
                        write_log(f"Обработано строк: {row_count}")
            
            # Закрываем JSON
            jsonfile.write('\n  ],\n')
            jsonfile.write(f'  "metadata": {{\n')
            jsonfile.write(f'    "total_rows": {row_count},\n')
            jsonfile.write(f'    "conversion_date": "{datetime.now().strftime("%Y-%m-%d %H:%M:%S")}",\n')
            jsonfile.write(f'    "source_file": "{os.path.basename(input_file)}"\n')
            jsonfile.write(f'  }}\n')
            jsonfile.write('}\n')
        
        end_time = datetime.now()
        duration = (end_time - start_time).total_seconds()
        
        write_log("-"*50)
        write_log("ИТОГИ КОНВЕРТАЦИИ:")
        write_log(f"Обработано строк: {row_count}")
        write_log(f"Время выполнения: {duration:.3f} сек")
        write_log("КОНВЕРТАЦИЯ ЗАВЕРШЕНА УСПЕШНО")
        write_log("="*50)
        
        print(f"\n✅ Конвертация завершена!")
        print(f"Обработано строк: {row_count}")
        print(f"Время выполнения: {duration:.2f} сек")
        
        return True
        
    except Exception as e:
        write_log(f"ОШИБКА: {e}")
        write_log("="*50)
        print(f"\n❌ Ошибка: {e}")
        return False

# Задаём пути
path = f"D:/Python/Pyton-2026/files/"
input_file = os.path.join(path, "large_data.csv")
output_file = os.path.join(path, "large_data.json")
log_file = os.path.join(path, "conversion_log.txt")

# Создаём тестовый CSV-файл
csv_data = """ID,Имя,Возраст,Город,Зарплата
1,Иван Петров,28,Москва,85000
2,Анна Сидорова,25,Санкт-Петербург,65000
3,Сергей Иванов,32,Новосибирск,90000
4,Мария Смирнова,29,Екатеринбург,75000
5,Алексей Кузнецов,35,Казань,70000
"""

with open(input_file, 'w', encoding='utf-8') as f:
    f.write(csv_data)

# Выполняем конвертацию
convert_csv_to_json_large(input_file, output_file, log_file)

# Показываем лог
print("\n" + "="*70)
print("ЛОГ-ФАЙЛ:")
print("="*70)
with open(log_file, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# С функцией для разных режимов чтения
import os
import chardet  # Требуется установка: pip install chardet

def detect_encoding(filename):
    """Определяет кодировку файла"""
    try:
        with open(filename, 'rb') as f:
            raw_data = f.read(10000)  # Читаем первые 10KB
            result = chardet.detect(raw_data)
            return result['encoding']
    except:
        return None

def safe_read_file_advanced():
    """
    Расширенная версия безопасного чтения файла
    """
    print("="*70)
    print("БЕЗОПАСНОЕ ЧТЕНИЕ ФАЙЛА (расширенная версия)")
    print("="*70)
    
    filename = input("\nВведите имя файла для чтения: ").strip()
    
    if not filename:
        print("❌ Имя файла не может быть пустым!")
        return
    
    print(f"\nПытаюсь прочитать файл: {filename}")
    print("-"*70)
    
    # Проверяем существование файла
    if not os.path.exists(filename):
        print(f"❌ Файл '{filename}' не найден!")
        return
    
    # Проверяем права доступа
    if not os.access(filename, os.R_OK):
        print(f"❌ Нет прав на чтение файла '{filename}'!")
        return
    
    # Проверяем, является ли файл директорией
    if os.path.isdir(filename):
        print(f"❌ '{filename}' является директорией, а не файлом!")
        return
    
    # Пробуем определить кодировку
    detected_encoding = detect_encoding(filename)
    if detected_encoding:
        print(f"Определена кодировка: {detected_encoding}")
    
    # Список кодировок для попыток
    encodings = [
        detected_encoding,
        'utf-8',
        'cp1251',
        'latin-1',
        'cp866',
        'koi8-r',
        'utf-16',
        'utf-32'
    ]
    # Убираем None и дубликаты
    encodings = [enc for enc in encodings if enc]
    encodings = list(dict.fromkeys(encodings))
    
    # Пробуем прочитать с разными кодировками
    for encoding in encodings:
        try:
            with open(filename, 'r', encoding=encoding) as file:
                content = file.read()
            
            print(f"\n✅ Файл успешно прочитан с кодировкой: {encoding}")
            print(f"Размер файла: {os.path.getsize(filename)} байт")
            print(f"Количество строк: {len(content.splitlines())}")
            print(f"Количество символов: {len(content)}")
            
            # Показываем первые 10 строк
            print("\n" + "="*70)
            print("ПЕРВЫЕ 10 СТРОК ФАЙЛА:")
            print("="*70)
            lines = content.splitlines()
            for i, line in enumerate(lines[:10], 1):
                print(f"{i:3}: {line}")
            
            if len(lines) > 10:
                print(f"\n... и ещё {len(lines) - 10} строк")
            
            return
            
        except UnicodeDecodeError:
            continue
        except Exception as e:
            print(f"Ошибка при чтении с кодировкой {encoding}: {e}")
            continue
    
    # Если ни одна кодировка не подошла
    print("\n❌ Не удалось прочитать файл ни с одной из кодировок!")
    
    # Предлагаем прочитать в бинарном режиме
    choice = input("\nПрочитать файл в бинарном режиме? (y/n): ").strip().lower()
    if choice == 'y':
        try:
            with open(filename, 'rb') as file:
                binary_data = file.read(1000)
            print(f"\nПервые 200 байт файла (в шестнадцатеричном формате):")
            print("-"*50)
            for i in range(0, min(len(binary_data), 200), 16):
                hex_part = ' '.join(f'{b:02x}' for b in binary_data[i:i+16])
                ascii_part = ''.join(chr(b) if 32 <= b < 127 else '.' for b in binary_data[i:i+16])
                print(f"{i:04x}: {hex_part:<48} {ascii_part}")
        except Exception as e:
            print(f"Ошибка при бинарном чтении: {e}")

# Запуск расширенной версии
if __name__ == "__main__":
    safe_read_file_advanced()
````

### Задание 8.2 — Проверка перед записью

Перед записью в файл проверяйте:

- Существует ли директория (если нет — создайте)

- Достаточно ли свободного места (имитируйте проверку)

- Не пустой ли файл (спрашивайте подтверждение на перезапись)

````python
import os
import shutil
from datetime import datetime

def check_directory_exists(directory):
    """
    Проверяет существование директории, если нет - создаёт
    """
    if os.path.exists(directory):
        if os.path.isdir(directory):
            print(f"✓ Директория существует: {directory}")
            return True
        else:
            print(f"❌ Путь существует, но это не директория: {directory}")
            return False
    else:
        print(f"⚠️ Директория не существует: {directory}")
        try:
            os.makedirs(directory, exist_ok=True)
            print(f"✓ Директория успешно создана: {directory}")
            return True
        except Exception as e:
            print(f"❌ Ошибка при создании директории: {e}")
            return False

def check_free_space(directory, required_size_bytes):
    """
    Проверяет наличие свободного места (имитация)
    """
    print(f"\n--- ПРОВЕРКА СВОБОДНОГО МЕСТА ---")
    print(f"Требуется места: {required_size_bytes} байт ({required_size_bytes / 1024:.2f} KB)")
    
    try:
        # Получаем информацию о свободном месте на диске
        stat = os.statvfs(directory)
        free_space = stat.f_frsize * stat.f_bavail
        
        print(f"Доступно места: {free_space} байт ({free_space / 1024 / 1024:.2f} MB)")
        
        if free_space >= required_size_bytes:
            print(f"✓ Достаточно свободного места")
            return True
        else:
            print(f"❌ Недостаточно свободного места!")
            print(f"   Требуется: {required_size_bytes / 1024:.2f} KB")
            print(f"   Доступно: {free_space / 1024:.2f} KB")
            return False
            
    except Exception as e:
        print(f"⚠️ Не удалось проверить свободное место: {e}")
        print("   Продолжаем выполнение (имитация успешной проверки)")
        return True  # В случае ошибки считаем, что места достаточно

def check_file_not_empty(filepath):
    """
    Проверяет, существует ли файл и не пустой ли он
    """
    if not os.path.exists(filepath):
        print(f"✓ Файл не существует, можно создавать новый")
        return False  # Файл не существует
    
    if os.path.getsize(filepath) == 0:
        print(f"⚠️ Файл существует, но пустой")
        return True  # Файл пустой
    
    print(f"⚠️ Файл существует и содержит данные")
    print(f"   Размер файла: {os.path.getsize(filepath)} байт")
    return True  # Файл не пустой

def ask_overwrite_confirmation(filepath):
    """
    Спрашивает подтверждение на перезапись файла
    """
    print(f"\n--- ПОДТВЕРЖДЕНИЕ ПЕРЕЗАПИСИ ---")
    print(f"Файл уже существует: {filepath}")
    
    # Показываем содержимое существующего файла
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            content = f.read()
            if content:
                print("\nСодержимое существующего файла:")
                print("-"*50)
                lines = content.splitlines()
                for i, line in enumerate(lines[:5], 1):
                    print(f"{i:2}: {line}")
                if len(lines) > 5:
                    print(f"... и ещё {len(lines) - 5} строк")
                print("-"*50)
    except:
        pass
    
    while True:
        choice = input("\nФайл будет перезаписан. Продолжить? (y/n): ").strip().lower()
        if choice == 'y':
            print("✓ Подтверждено: файл будет перезаписан")
            return True
        elif choice == 'n':
            print("❌ Операция отменена пользователем")
            return False
        else:
            print("Пожалуйста, введите 'y' или 'n'")

def safe_write_to_file(filepath, content, simulate_space=True):
    """
    Безопасная запись в файл с проверками
    
    Args:
        filepath: путь к файлу
        content: содержимое для записи
        simulate_space: имитировать проверку свободного места
    """
    print("="*70)
    print("БЕЗОПАСНАЯ ЗАПИСЬ В ФАЙЛ")
    print("="*70)
    print(f"Целевой файл: {filepath}")
    print("-"*70)
    
    # 1. Проверяем директорию
    directory = os.path.dirname(filepath)
    if directory and not check_directory_exists(directory):
        return False
    
    # 2. Проверяем свободное место (имитация)
    content_size = len(content.encode('utf-8'))
    if simulate_space:
        if not check_free_space(directory or '.', content_size):
            return False
    else:
        print("ℹ️ Проверка свободного места пропущена")
    
    # 3. Проверяем, существует ли файл и не пустой ли он
    if check_file_not_empty(filepath):
        # Файл существует, запрашиваем подтверждение
        if not ask_overwrite_confirmation(filepath):
            return False
    
    # 4. Выполняем запись
    try:
        # Создаём временный файл для безопасной записи
        temp_file = filepath + ".tmp"
        
        print(f"\n--- ЗАПИСЬ В ФАЙЛ ---")
        print(f"Записываем данные...")
        
        with open(temp_file, 'w', encoding='utf-8') as f:
            f.write(content)
        
        # Заменяем оригинальный файл временным
        if os.path.exists(filepath):
            os.remove(filepath)
        os.rename(temp_file, filepath)
        
        print(f"✓ Данные успешно записаны в файл")
        print(f"  Размер записи: {content_size} байт ({content_size / 1024:.2f} KB)")
        print(f"  Количество строк: {len(content.splitlines())}")
        
        return True
        
    except Exception as e:
        print(f"❌ Ошибка при записи в файл: {e}")
        # Удаляем временный файл в случае ошибки
        if os.path.exists(temp_file):
            os.remove(temp_file)
        return False

def simulate_low_space():
    """
    Имитирует ситуацию с недостатком места
    """
    print("\n" + "="*70)
    print("СИМУЛЯЦИЯ: НЕДОСТАТОЧНО МЕСТА")
    print("="*70)
    
    filepath = "D:/Python/Pyton-2026/files/test_low_space.txt"
    content = "X" * 1024 * 1024 * 100  # 100 MB данных
    
    # Специальная проверка с имитацией недостатка места
    def low_space_check(directory, required_size):
        print(f"Требуется места: {required_size / 1024 / 1024:.2f} MB")
        print(f"Доступно (симуляция): 50 MB")
        print("❌ Недостаточно свободного места!")
        return False
    
    # Подменяем функцию проверки
    global check_free_space
    original_check = check_free_space
    try:
        check_free_space = low_space_check
        success = safe_write_to_file(filepath, content)
        if not success:
            print("\n✅ Симуляция успешна: запись отклонена из-за недостатка места")
    finally:
        check_free_space = original_check

def main():
    """
    Основная программа с демонстрацией различных сценариев
    """
    # Создаём тестовые данные
    path = "D:/Python/Pyton-2026/files/"
    test_file = os.path.join(path, "test_output.txt")
    
    test_content = """Это тестовое содержимое файла.
Создано: {timestamp}
Цель: демонстрация безопасной записи.

Это вторая строка.
И третья строка.
И ещё несколько строк для проверки.
""".format(timestamp=datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    
    # Сценарий 1: Первая запись (директория существует, файла нет)
    print("\n" + "="*70)
    print("СЦЕНАРИЙ 1: ПЕРВАЯ ЗАПИСЬ")
    print("="*70)
    success = safe_write_to_file(test_file, test_content)
    if success:
        print("\n✅ Сценарий 1 выполнен успешно")
    
    # Сценарий 2: Запись в существующий файл (должен спросить подтверждение)
    print("\n" + "="*70)
    print("СЦЕНАРИЙ 2: ЗАПИСЬ В СУЩЕСТВУЮЩИЙ ФАЙЛ")
    print("="*70)
    
    new_content = test_content.replace("тестовое", "ОБНОВЛЁННОЕ")
    new_content += "\n\nЭто обновлённое содержимое, добавленное позже."
    
    success = safe_write_to_file(test_file, new_content)
    if success:
        print("\n✅ Сценарий 2 выполнен успешно")
    
    # Сценарий 3: Запись в новую директорию
    print("\n" + "="*70)
    print("СЦЕНАРИЙ 3: ЗАПИСЬ В НОВУЮ ДИРЕКТОРИЮ")
    print("="*70)
    
    new_dir_file = "D:/Python/Pyton-2026/files/new_folder/subfolder/test_new.txt"
    content_small = "Небольшое содержимое для нового файла"
    
    success = safe_write_to_file(new_dir_file, content_small)
    if success:
        print("\n✅ Сценарий 3 выполнен успешно")
    
    # Сценарий 4: Попытка записи в защищённую директорию
    print("\n" + "="*70)
    print("СЦЕНАРИЙ 4: ЗАПИСЬ В ЗАЩИЩЁННУЮ ДИРЕКТОРИЮ")
    print("="*70)
    
    protected_file = "C:/Windows/test_protected.txt"
    success = safe_write_to_file(protected_file, "Тестовый контент")
    if not success:
        print("\n✅ Сценарий 4: запись отклонена (ожидаемо)")
    
    # Сценарий 5: Пустой файл
    print("\n" + "="*70)
    print("СЦЕНАРИЙ 5: ЗАПИСЬ В ПУСТОЙ ФАЙЛ")
    print("="*70)
    
    empty_file = os.path.join(path, "empty.txt")
    
    # Создаём пустой файл
    with open(empty_file, 'w', encoding='utf-8') as f:
        pass
    
    success = safe_write_to_file(empty_file, "Новое содержимое для пустого файла")
    if success:
        print("\n✅ Сценарий 5 выполнен успешно")
    
    # Сценарий 6: Имитация недостатка места
    simulate_low_space()
    
    # Выводим итоговую статистику
    print("\n" + "="*70)
    print("ИТОГОВАЯ СТАТИСТИКА")
    print("="*70)
    
    if os.path.exists(test_file):
        size = os.path.getsize(test_file)
        print(f"Файл: {test_file}")
        print(f"  Размер: {size} байт ({size / 1024:.2f} KB)")
        with open(test_file, 'r', encoding='utf-8') as f:
            lines = len(f.readlines())
        print(f"  Количество строк: {lines}")
    
    print("\n✅ Все проверки выполнены!")

# Запуск программы
if __name__ == "__main__":
    main()
````
