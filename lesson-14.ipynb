````python
# 5. Линейная алгебра
a = np.array([[0, 1], [2, 3]])
print(np.linalg.det(a))
print('='*50)

# Обратная матрица
a1 = np.linalg.inv(a)
print(a1)
print(a @ a1)
print(a1 @ a)
print('='*50)

# Решение линейной системы au=v
v = np.array([0, 1], dtype=np.float64)
print(a1 @ v)
u = np.linalg.solve(a, v)
print(u)
print('='*50)

# Проверим
print(a @ u - v)
print('='*50)

# Собственные значения и собственные векторы
l, u = np.linalg.eig(a)
print(l)
print(u)
print('='*50)

# Проверим
for i in range(2):
    print(a @ u[:, i] - l[i] * u[:, i])
````

````python
# Функция diag от одномерного массива строит диагональную матрицу
L = np.diag(l)
print(L)
print(np.diag(L))
print('='*50)

print(a @ u - u @ L)
print('='*50)

print(np.linalg.inv(u) @ a @ u)
print('='*50)

# Найдём теперь левые собственные векторы
l, v = np.linalg.eig(a.T)
print(l)
print(v)
print('='*50)

# Собственные векторы нормированы на 1
print(u.T @ u)
print(v.T @ v)
print('='*50)

# Левые и правые собственные векторы, соответствующие разным собственнымзначениям, ортогональны
print(v.T @ u)
````

### 3. Стандартные массивы:

Создайте следующие массивы без использования ручного ввода чисел:

Массив из 10 нулей.

Массив из 5 единиц.

Единичную матрицу размером 3x3.

Массив, заполненный константой (например, числом 7), размером 2x4.

````python
import numpy as np

# 1. Массив из 10 нулей
zeros_arr = np.zeros(10)
print("Массив из 10 нулей:")
print(zeros_arr)
print(f"Форма: {zeros_arr.shape}")
print()

# 2. Массив из 5 единиц
ones_arr = np.ones(5)
print("Массив из 5 единиц:")
print(ones_arr)
print(f"Форма: {ones_arr.shape}")
print()

# 3. Единичная матрица размером 3x3
identity_matrix = np.eye(3)
print("Единичная матрица 3x3:")
print(identity_matrix)
print(f"Форма: {identity_matrix.shape}")
print()

# 4. Массив, заполненный константой (числом 7), размером 2x4
constant_arr = np.full((2, 4), 7)
print("Массив 2x4, заполненный числом 7:")
print(constant_arr)
print(f"Форма: {constant_arr.shape}")
````

### 7. Срезы (Slicing):

Создайте массив от 0 до 20. Выберите из него все элементы со 2-го по 10-й с шагом 2.

````python
import numpy as np

# Создаём массив от 0 до 20
arr = np.arange(21)  # от 0 до 20 включительно
print("Исходный массив от 0 до 20:")
print(arr)
print()

# Выбираем элементы со 2-го по 10-й с шагом 2
# Индексация: start=2, stop=11 (так как 10-й элемент имеет индекс 10, а stop не включается), step=2
result = arr[2:11:2]
print("Элементы со 2-го по 10-й с шагом 2:")
print(result)
````

````python
# Дополнительные примеры срезов:
import numpy as np

arr = np.arange(21)
print("Исходный массив:", arr)

# Различные варианты срезов:

# 1. Без указания start (начинаем с начала)
print(f"Первые 10 элементов: {arr[:10]}")

# 2. Без указания stop (и до конца)
print(f"Элементы с 10-го до конца: {arr[10:]}")

# 3. Только с шагом
print(f"Каждый третий элемент: {arr[::3]}")

# 4. Отрицательный шаг (обратный порядок)
print(f"Массив в обратном порядке: {arr[::-1]}")

# 5. Отрицательные индексы
print(f"Последние 5 элементов: {arr[-5:]}")

# 6. Сложный срез
print(f"С 5-го по 15-й элемент с шагом 3: {arr[5:16:3]}")

# 7. Срез с начала до 10-го элемента с шагом 2
print(f"Первые 10 элементов с шагом 2: {arr[:10:2]}")
````

**Ключевые моменты:**

- Синтаксис: `[start:stop:step]`

- `start` — включается в результат

- `stop` — не включается в результат

- `step` — шаг (может быть отрицательным)

- Если `start` опущен — начало массива

- Если `stop` опущен — до конца массива

- Если `step` опущен — шаг равен 1

- Срезы создают представление (view) исходного массива, а не копию (изменение среза изменит исходный массив)

### 8. Изменение формы (reshape):

Создайте одномерный массив из 12 элементов (от 0 до 11). Преобразуйте его в:

- Массив формы (3, 4).

- Массив формы (2, 2, 3) (3-мерный куб).

````python
import numpy as np

# Создаём одномерный массив из 12 элементов (от 0 до 11)
arr_1d = np.arange(12)
print("Исходный одномерный массив:")
print(arr_1d)
print(f"Форма: {arr_1d.shape}")
print(f"Размерность: {arr_1d.ndim}")
print()

# 1. Преобразуем в массив формы (3, 4)
arr_3x4 = arr_1d.reshape(3, 4)
print("Массив формы (3, 4):")
print(arr_3x4)
print(f"Форма: {arr_3x4.shape}")
print(f"Размерность: {arr_3x4.ndim}")
print()

# 2. Преобразуем в массив формы (2, 2, 3) - 3-мерный куб
arr_2x2x3 = arr_1d.reshape(2, 2, 3)
print("Массив формы (2, 2, 3) - 3-мерный:")
print(arr_2x2x3)
print(f"Форма: {arr_2x2x3.shape}")
print(f"Размерность: {arr_2x2x3.ndim}")
print()

# Демонстрация доступа к элементам 3-мерного массива
print("Доступ к элементам 3-мерного массива:")
print(f"arr_2x2x3[0, 0, 0] = {arr_2x2x3[0, 0, 0]}")  # первый элемент
print(f"arr_2x2x3[0, 1, 2] = {arr_2x2x3[0, 1, 2]}")  # элемент в первой "строке", второй "строке", третьем "столбце"
print(f"arr_2x2x3[1, 0, 1] = {arr_2x2x3[1, 0, 1]}")  # элемент во втором "слое", первой "строке", втором "столбце"
````

````python
# Дополнительные способы использования reshape:
import numpy as np

arr_1d = np.arange(12)

# 1. Использование -1 для автоматического определения размера
# NumPy сам вычислит размер по одному из измерений
arr_auto = arr_1d.reshape(3, -1)  # -1 означает "вычисли автоматически"
print("Использование -1 (3, -1):")
print(arr_auto)
print(f"Форма: {arr_auto.shape}")
print()

arr_auto2 = arr_1d.reshape(-1, 4)  # 4 столбца, строки вычисляются автоматически
print("Использование -1 (-1, 4):")
print(arr_auto2)
print(f"Форма: {arr_auto2.shape}")
print()

# 2. reshape с порядком заполнения (order)
# 'C' - построчно (C-style, по умолчанию)
# 'F' - постолбцово (Fortran-style)
arr_c = arr_1d.reshape(3, 4, order='C')
arr_f = arr_1d.reshape(3, 4, order='F')
print("Порядок заполнения 'C' (построчно):")
print(arr_c)
print("\nПорядок заполнения 'F' (постолбцово):")
print(arr_f)

# 3. Проверка, что reshape создаёт представление (view), а не копию
arr_2d = arr_1d.reshape(3, 4)
arr_1d[0] = 99  # изменяем исходный массив
print("\nПосле изменения исходного массива:")
print(f"Исходный: {arr_1d}")
print(f"Изменённый (reshape): {arr_2d}")  # тоже изменился!
# Будьте осторожны: reshape возвращает view, если это возможно

# 4. Использование reshape с разными размерностями
arr_3d = np.arange(24).reshape(2, 3, 4)
print(f"\nМассив 2x3x4:\n{arr_3d}")
print(f"Форма: {arr_3d.shape}")
````

**Ключевые моменты:**

- `reshape()` не изменяет исходный массив, а возвращает новый (часто представление/view)

- Общее количество элементов должно совпадать: `12 = 3×4 = 2×2×3`

- `-1` в reshape означает "вычислить автоматически"

- По умолчанию заполнение идёт построчно (C-style, параметр `order='C'`)

- Для постолбцового заполнения используйте `order='F'` (Fortran-style)

- Изменение исходного массива может повлиять на reshaped-массив (если это view)

### 9. Условная индексация (Boolean indexing):

Создайте массив случайных целых чисел от 1 до 20 размером (5, 5). Выведите на экран все элементы, которые больше 10.

````python
import numpy as np

# Устанавливаем seed для воспроизводимости результатов
np.random.seed(42)

# Создаём массив случайных целых чисел от 1 до 20 размером (5, 5)
arr = np.random.randint(1, 21, size=(5, 5))
print("Исходный массив 5x5 (случайные числа от 1 до 20):")
print(arr)
print()

# 1. Создаём булевую маску (элементы > 10)
mask = arr > 10
print("Булевая маска (True где элементы > 10):")
print(mask)
print()

# 2. Выводим все элементы, которые больше 10 (с помощью маски)
elements_greater_than_10 = arr[mask]
print("Все элементы, которые больше 10:")
print(elements_greater_than_10)
print()

# 3. Альтернативный способ (без создания отдельной переменной маски)
print("Альтернативный способ (напрямую):")
print(arr[arr > 10])
````

````python
# Дополнительные примеры условной индексации:
import numpy as np

np.random.seed(42)
arr = np.random.randint(1, 21, size=(5, 5))
print("Исходный массив:")
print(arr)
print()

# 1. Подсчёт количества элементов, удовлетворяющих условию
count = np.sum(arr > 10)
print(f"Количество элементов > 10: {count}")

# 2. Процент элементов > 10
percentage = (count / arr.size) * 100
print(f"Процент элементов > 10: {percentage:.1f}%")
print()

# 3. Использование нескольких условий (логические операторы)
# Элементы больше 5 И меньше 15
mask_and = (arr > 5) & (arr < 15)
print("Элементы > 5 и < 15:")
print(arr[mask_and])
print()

# Элементы меньше 5 ИЛИ больше 15
mask_or = (arr < 5) | (arr > 15)
print("Элементы < 5 или > 15:")
print(arr[mask_or])
print()

# 4. Инвертирование условия (элементы НЕ больше 10)
mask_not = ~(arr > 10)  # или arr <= 10
print("Элементы <= 10:")
print(arr[mask_not])
print()

# 5. Замена элементов по условию с помощью np.where
# Заменяем все элементы > 10 на 0
arr_mod = arr.copy()
arr_mod[arr_mod > 10] = 0
print("Массив после замены элементов > 10 на 0:")
print(arr_mod)
print()

# 6. Использование np.where для более сложной замены
# Заменяем элементы > 10 на 100, остальные оставляем как есть
arr_where = np.where(arr > 10, 100, arr)
print("np.where: элементы > 10 заменены на 100:")
print(arr_where)
print()

# 7. Получение индексов элементов, удовлетворяющих условию
indices = np.where(arr > 10)
print("Индексы элементов > 10:")
print(f"Строки: {indices[0]}")
print(f"Столбцы: {indices[1]}")
print()

# 8. Вывод элементов с их индексами
print("Элементы > 10 с их координатами:")
for i in range(len(indices[0])):
    row = indices[0][i]
    col = indices[1][i]
    value = arr[row, col]
    print(f"  arr[{row}, {col}] = {value}")
````

**Ключевые моменты:**

- Булева маска — это массив из `True/False` той же формы, что и исходный

- `arr[mask]` возвращает одномерный массив элементов, где маска равна `True`

- Для комбинирования условий используйте `& (И)`, `| (ИЛИ)`, `~ (НЕ)`

- `np.where(условие, значение_если_True, значение_если_False)` — более гибкий способ замены

- `np.where(условие)` возвращает кортеж индексов, где условие истинно

- Булева индексация всегда создаёт **копию**, а не представление (view)

### 10. Flatten (сглаживание):

Возьмите массив формы (4, 3) и превратите его в одномерный массив с помощью метода flatten() или ravel().

````python
# Важное различие между flatten() и ravel():
import numpy as np

arr = np.arange(12).reshape(4, 3)
print("Исходный массив:")
print(arr)
print()

# flatten() - ВСЕГДА возвращает КОПИЮ данных
arr_flatten = arr.flatten()
print(f"flatten() - это копия? {arr_flatten.base is None}")  # True - нет связи с исходным

# ravel() - возвращает ПРЕДСТАВЛЕНИЕ (view) если возможно, иначе копию
arr_ravel = arr.ravel()
print(f"ravel() - это представление? {arr_ravel.base is arr}")  # True - есть связь

# Демонстрация различий:
print("\n--- Демонстрация различий ---")
print("Изменяем исходный массив:")
arr[0, 0] = 99
print(f"Исходный массив:\n{arr}")

print(f"\nflatten() результат (не изменился, т.к. это копия):\n{arr_flatten}")
print(f"ravel() результат (изменился, т.к. это представление):\n{arr_ravel}")

# ravel() иногда может вернуть копию (например, для непоследовательных массивов)
print("\n--- Когда ravel() возвращает копию ---")
arr_transposed = arr.T  # Транспонированный массив (не последовательный в памяти)
print(f"Транспонированный массив:\n{arr_transposed}")
print(f"Транспонированный массив в памяти: {arr_transposed.flags['C_CONTIGUOUS']}")

ravel_transposed = arr_transposed.ravel()
print(f"ravel() на транспонированном вернул копию? {ravel_transposed.base is None}")
# В этом случае ravel() может вернуть копию, т.к. не может создать представление
````

````python
# Дополнительные способы сглаживания:
import numpy as np

arr = np.arange(12).reshape(4, 3)
print("Исходный массив:")
print(arr)
print()

# 1. reshape с -1
arr_reshape = arr.reshape(-1)
print("Сглаживание с reshape(-1):")
print(arr_reshape)
print()

# 2. flatten с порядком обхода
print("flatten с разными порядками обхода:")
print(f"Построчно (C-style): {arr.flatten(order='C')}")
print(f"Постолбцово (Fortran-style): {arr.flatten(order='F')}")
print()

# 3. ravel с порядком обхода
print("ravel с разными порядками обхода:")
print(f"Построчно (C-style): {arr.ravel(order='C')}")
print(f"Постолбцово (Fortran-style): {arr.ravel(order='F')}")
````

**Сравнение методов сглаживания:**

|Метод	|Возвращает	|Когда использовать	|Производительность
|-------|-----------|-------------------|-------------------
|flatten()	|Всегда копию	|Когда нужна независимая копия	|Медленнее (выделяет новую память)
|ravel()	|Представление (если возможно)	|Когда нужно быстро и нет опасности изменений	|Быстрее (часто без копирования)
|reshape(-1)	|Представление (если возможно)	|Когда нужно сгладить и возможно изменить форму	|Аналогично ravel()

**Ключевые моменты:**

- `flatten()` — всегда создаёт копию данных

- `ravel()` — создаёт представление (view) по возможности (быстрее, экономит память)

- Изменение через `ravel()` (если это view) изменит исходный массив

- `reshape(-1)` — альтернативный способ сглаживания (также возвращает view)

- Для постолбцового сглаживания используйте параметр `order='F'`

- Максимум и его индекс (max, argmax).

### 14. Скалярное произведение:

Создайте два вектора (одномерные массива) длиной 5 и вычислите их скалярное произведение двумя способами: np.dot и методом @.

````python
import numpy as np

# Создаём два вектора (одномерные массивы) длиной 5
a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])

print("Вектор a:", a)
print("Вектор b:", b)
print()

# Вычисляем скалярное произведение двумя способами

# Способ 1: с помощью функции np.dot()
dot_product_1 = np.dot(a, b)
print("1. Скалярное произведение (np.dot(a, b)):")
print(f"   {dot_product_1}")
print()

# Способ 2: с помощью оператора @
dot_product_2 = a @ b
print("2. Скалярное произведение (a @ b):")
print(f"   {dot_product_2}")
print()

# Проверка, что результаты совпадают
print(f"Результаты совпадают: {dot_product_1 == dot_product_2}")
print()

# Детальный расчёт (для проверки)
print("Детальный расчёт (сумма произведений элементов):")
calculation = 0
for i in range(len(a)):
    product = a[i] * b[i]
    calculation += product
    print(f"   a[{i}] * b[{i}] = {a[i]} * {b[i]} = {product}")
print(f"   Сумма = {calculation}")
````

````python
# Сохранение размерности (keepdims):
import numpy as np

np.random.seed(42)
matrix = np.random.randint(1, 100, size=(3, 4))

print("Исходная матрица:")
print(matrix)
print(f"Форма: {matrix.shape}")
print()

# Без keepdims
sum_rows = np.sum(matrix, axis=1)
print("Сумма по строкам (без keepdims):")
print(f"Результат: {sum_rows}")
print(f"Форма результата: {sum_rows.shape}")  # (3,) - одномерный
print()

# С keepdims=True
sum_rows_keep = np.sum(matrix, axis=1, keepdims=True)
print("Сумма по строкам (с keepdims=True):")
print(f"Результат:\n{sum_rows_keep}")
print(f"Форма результата: {sum_rows_keep.shape}")  # (3, 1) - двумерный
print()

# Практическое применение keepdims
# Позволяет выполнять операции вещания (broadcasting)
normalized = matrix / sum_rows_keep  # Делим каждую строку на её сумму
print("Нормализованная матрица (каждая строка делится на сумму элементов строки):")
print(normalized)
print(f"Сумма элементов в каждой строке: {np.sum(normalized, axis=1)}")
````

**Ключевые моменты:**

- `axis=0` — операция выполняется **вниз по строкам** (вертикально), удаляя строки. Результат имеет размерность количества столбцов (одна строка из сумм/средних по каждому столбцу).

- `axis=1` — операция выполняется **поперёк столбцов** (горизонтально), удаляя столбцы. Результат имеет размерность количества строк (один столбец из сумм/средних по каждой строке)

- `axis=None` или без axis — операция применяется ко всем элементам (по умолчанию)

- `keepdims=True` сохраняет размерность массива (полезно для broadcasting)

- В многомерных массивах axis указывает измерение, по которому выполняется свёртка

- Понимание axis критически важно для работы с многомерными данными (изображения, временные ряды, таблицы)

#### Объяснение механизма вещания (Broadcasting)

**Почему это сработало?**

Вещание сработало, потому что формы массивов совместимы по правилам NumPy:

1. **Матрица имеет форму (4, 3)**

2. **Одномерный массив имеет форму (3,)**

NumPy автоматически "растягивает" одномерный массив, чтобы он соответствовал форме матрицы:

```text
Матрица (4, 3):     Массив (3,)        Результат (4, 3)
[[0, 1, 2]          [10, 20, 30]      [[10, 21, 32]
 [3, 4, 5]    +     [10, 20, 30]  =   [13, 24, 35]
 [6, 7, 8]          [10, 20, 30]      [16, 27, 38]
 [9, 10, 11]]       [10, 20, 30]]     [19, 30, 41]]
 ```

#### Правила вещания NumPy

Два массива совместимы для вещания, если:

1. Их размерности совпадают, или

2. Одна из размерностей равна 1, или

3. Одному из массивов не хватает размерностей (он дополняется слева единицами)

````python
# Пошаговое объяснение для нашего случая
import numpy as np

matrix = np.arange(12).reshape(4, 3)
array_1d = np.array([10, 20, 30])

print("Пошаговое объяснение вещания:")
print()

# Шаг 1: Сравниваем формы
print(f"Шаг 1: Форма матрицы: {matrix.shape}")
print(f"        Форма массива: {array_1d.shape}")
print()

# Шаг 2: NumPy дополняет форму массива слева единицами
print(f"Шаг 2: Дополнение формы массива слева единицами:")
print(f"        Исходная форма: {array_1d.shape}")
print(f"        Дополненная форма: (1, {array_1d.shape[0]}) -> (1, 3)")
print()

# Шаг 3: Сравниваем размерности справа налево
print("Шаг 3: Сравнение размерностей (справа налево):")
print(f"        Размерность 1 (столбцы): 3 и 3 -> совпадают ✓")
print(f"        Размерность 0 (строки):  4 и 1 -> 1 может быть расширена ✓")
print()

# Шаг 4: Массив "растягивается" по размерности, где у него 1
print("Шаг 4: Массив 'растягивается' (копируется) 4 раза по вертикали:")
print("        [10, 20, 30]  →  [[10, 20, 30],")
print("                           [10, 20, 30],")
print("                           [10, 20, 30],")
print("                           [10, 20, 30]]")
````

````python
# Визуализация правил вещания
import numpy as np

def explain_broadcasting(shape1, shape2):
    """Объясняет, совместимы ли формы для вещания"""
    print(f"\nПроверка совместимости форм {shape1} и {shape2}:")
    
    # Дополняем меньшую форму единицами слева
    len1, len2 = len(shape1), len(shape2)
    if len1 > len2:
        shape2_padded = (1,) * (len1 - len2) + shape2
        shape1_padded = shape1
    else:
        shape1_padded = (1,) * (len2 - len1) + shape1
        shape2_padded = shape2
    
    print(f"  После дополнения: {shape1_padded} и {shape2_padded}")
    
    compatible = True
    for i, (d1, d2) in enumerate(zip(shape1_padded[::-1], shape2_padded[::-1])):
        if d1 != d2 and d1 != 1 and d2 != 1:
            compatible = False
            print(f"  ✗ Размерность {len(shape1_padded)-i-1}: {d1} и {d2} - несовместимы")
            break
        else:
            print(f"  ✓ Размерность {len(shape1_padded)-i-1}: {d1} и {d2} - совместимы")
    
    if compatible:
        result_shape = tuple(max(d1, d2) for d1, d2 in zip(shape1_padded, shape2_padded))
        print(f"\n  Результат: формы совместимы → {result_shape}")
    else:
        print(f"\n  Результат: формы НЕ совместимы")
    
    return compatible

# Тестируем разные пары форм
explain_broadcasting((4, 3), (3,))
explain_broadcasting((4, 3), (4, 1))
explain_broadcasting((4, 3), (1, 3))
explain_broadcasting((4, 3), (2,))
explain_broadcasting((3, 4, 5), (4, 5))
explain_broadcasting((3, 4, 5), (3, 5))
````

**Ключевые моменты о вещании (Broadcasting)**

1. **Что такое вещание?** — Механизм NumPy, позволяющий выполнять операции с массивами разных форм без явного копирования данных.

2. **Основное правило:** Два измерения совместимы, если они:

    - равны, или

    - одно из них равно 1

3. **Процесс вещания:**

    - NumPy сравнивает формы справа налево

    - Недостающие размерности дополняются единицами слева

    - Массивы "растягиваются" по размерностям, где у них 1

4. **Преимущества:**

    - Экономия памяти (нет явного копирования)

    - Более быстрый код (векторизация)

    - Чистый и читаемый код

5. **Когда НЕ работает:**

    - Когда соответствующие размерности не равны и ни одна не равна 1

6. **Полезные функции:**

    - `np.newaxis` или `None` — для добавления новых размерностей

    - `np.broadcast_to()` — явное вещание

    - `np.broadcast_arrays()` — просмотр результата вещания

Используя вещание, создайте таблицу умножения (матрицу 10x10), где элемент [i, j] равен i * j. (Подсказка: используйте np.arange и reshape).

````python
import numpy as np

# Создаём одномерный массив чисел от 0 до 9
numbers = np.arange(10)
print("Массив чисел от 0 до 9:")
print(numbers)
print()

# Способ 1: Используем reshape для создания вектор-столбца и вектор-строки
# Превращаем в вектор-столбец (10, 1) и вектор-строку (1, 10)
col_vector = numbers.reshape(10, 1)  # или numbers[:, np.newaxis]
row_vector = numbers.reshape(1, 10)  # или numbers[np.newaxis, :]

print("Вектор-столбец (10, 1):")
print(col_vector)
print(f"Форма: {col_vector.shape}")
print()

print("Вектор-строка (1, 10):")
print(row_vector)
print(f"Форма: {row_vector.shape}")
print()

# Создаём таблицу умножения с помощью вещания
multiplication_table = col_vector * row_vector

print("Таблица умножения 10x10 (i * j):")
print(multiplication_table)
print(f"Форма таблицы: {multiplication_table.shape}")
````

````python
# Визуализация вещания для внешнего произведения
import numpy as np

# Создаём маленький пример для наглядности
small_numbers = np.arange(1, 4)  # [1, 2, 3]
col_small = small_numbers.reshape(3, 1)
row_small = small_numbers.reshape(1, 3)

print("Наглядный пример с числами [1, 2, 3]:")
print("=" * 50)
print()
print(f"Вектор-столбец (3, 1):\n{col_small}")
print()
print(f"Вектор-строка (1, 3):\n{row_small}")
print()
print("Операция умножения с вещанием:")
print()
print("Шаг 1: Растягивание вектор-столбца:")
col_expanded = np.broadcast_to(col_small, (3, 3))
print(col_expanded)
print()
print("Шаг 2: Растягивание вектор-строки:")
row_expanded = np.broadcast_to(row_small, (3, 3))
print(row_expanded)
print()
print("Шаг 3: Поэлементное умножение:")
result = col_expanded * row_expanded
print(result)
print()
print("Итоговая таблица умножения:")
for i in range(3):
    for j in range(3):
        print(f"{small_numbers[i]} × {small_numbers[j]} = {result[i, j]}")
````

````python
# Практическое применение: ковариационная матрица
import numpy as np

# Внешнее произведение используется для вычисления ковариационной матрицы
# Центрированные данные (3 признака, 5 образцов)
data_centered = np.random.randn(3, 5)
print("Центрированные данные (3 признака, 5 образцов):")
print(data_centered)
print()

# Вычисляем ковариационную матрицу через внешнее произведение
# (1/n) * (X @ X.T) где X - центрированные данные
cov_matrix = (data_centered @ data_centered.T) / (data_centered.shape[1] - 1)
print("Ковариационная матрица (3x3):")
print(cov_matrix)
print()

# Сравнение с встроенной функцией
cov_builtin = np.cov(data_centered)
print("Ковариационная матрица (np.cov):")
print(cov_builtin)
print(f"Матрицы совпадают: {np.allclose(cov_matrix, cov_builtin)}")
````

````python
# Многомерное fancy indexing: сетка индексов
import numpy as np

# Создаём матрицу
matrix = np.arange(12).reshape(3, 4)
print("Исходная матрица 3x4:")
print(matrix)
print()

# Создаём массивы индексов строк и столбцов
row_indices = np.array([0, 1, 2, 0, 1, 2])
col_indices = np.array([0, 1, 2, 3, 3, 3])

selected = matrix[row_indices, col_indices]
print("Выборка элементов по заданным координатам:")
print(f"Строки: {row_indices}")
print(f"Столбцы: {col_indices}")
print(f"Результат: {selected}")
print()

# Визуализация выбранных элементов
print("Визуализация выбранных позиций (X):")
for i in range(3):
    row_str = ""
    for j in range(4):
        is_selected = any((i == r and j == c) for r, c in zip(row_indices, col_indices))
        if is_selected:
            row_str += f"  X  "
        else:
            row_str += f"  {matrix[i, j]:2d} "
    print(row_str)
````

**Ключевые моменты о Fancy Indexing**

1. Определение: Fancy indexing — это использование массивов индексов для выборки элементов.

2. Результат: Всегда возвращает копию данных (не представление/view).

3. Форма результата: Форма результата соответствует форме массива индексов, а не исходного массива.

4. Поддерживаемые типы индексов:

    - Списки Python `([1, 2, 3])`

    - Массивы NumPy `(np.array([1, 2, 3]))`

    - Булевы массивы (маски)

5. Многомерное использование:

    - Для 2D: `arr[row_indices, col_indices]`

    - Индексы должны быть одинаковой формы

6. Применения:

    - Выборка случайных элементов

    - Перестановка данных

    - Извлечение подмножеств по условиям

    - Bootstrap и семплирование

7. Преимущества:

    - Быстрее циклов Python

    - Гибкость в выборке элементов

    - Поддержка нерегулярных паттернов выборки

Создайте массив случайных чисел от -5 до 5. С помощью np.where замените все отрицательные числа на 0, оставив положительные без изменений.

````python
import numpy as np

# Устанавливаем seed для воспроизводимости
np.random.seed(42)

# Создаём массив случайных чисел от -5 до 5 (размером 5x5 для наглядности)
arr = np.random.randint(-5, 6, size=(5, 5))
print("Исходный массив случайных чисел от -5 до 5:")
print(arr)
print()

# С помощью np.where заменяем отрицательные числа на 0
# np.where(условие, значение_если_True, значение_если_False)
result = np.where(arr < 0, 0, arr)

print("Массив после замены отрицательных чисел на 0:")
print(result)
print()

# Проверка: показываем, что отрицательных чисел больше нет
print("Проверка: отрицательные числа отсутствуют?")
print(f"Есть ли отрицательные числа: {np.any(result < 0)}")
print(f"Количество нулей: {np.sum(result == 0)}")
print(f"Количество положительных чисел: {np.sum(result > 0)}")
````

````python
# Различные способы использования np.where
import numpy as np

np.random.seed(42)
arr = np.random.randint(-5, 6, size=(5, 5))
print("Исходный массив:")
print(arr)
print()

# 1. Базовое использование: замена отрицательных на 0
result1 = np.where(arr < 0, 0, arr)
print("1. Отрицательные → 0:")
print(result1)
print()

# 2. Замена с разными значениями
result2 = np.where(arr < 0, -1, 1)
print("2. Отрицательные → -1, остальные → 1:")
print(result2)
print()

# 3. Замена отрицательных на их абсолютное значение
result3 = np.where(arr < 0, -arr, arr)
print("3. Отрицательные → |значение|:")
print(result3)
print()

# 4. Замена положительных на 100, отрицательные оставляем
result4 = np.where(arr > 0, 100, arr)
print("4. Положительные → 100, остальные без изменений:")
print(result4)
print()

# 5. Сложное условие: числа между -2 и 2 заменяем на 0
result5 = np.where((arr >= -2) & (arr <= 2), 0, arr)
print("5. Числа от -2 до 2 → 0:")
print(result5)
````

**Ключевые моменты о np.where**

1. Синтаксис:

    - `np.where(condition)` — возвращает индексы элементов, где condition истинно

    - `np.where(condition, x, y)` — возвращает массив, где элементы равны x если condition истинно, иначе y

2. Возвращаемое значение:

    - С одним аргументом: кортеж массивов индексов

    - С двумя/тремя аргументами: массив той же формы, что и condition

3. Преимущества:

    - Быстрее циклов Python

    - Читаемый и лаконичный код

    - Векторизованная операция

4. Типичные применения:

    - Замена значений по условию

    - Обрезка (clipping) данных

    - Условные вычисления

    - Категоризация данных

    - Обработка выбросов

5. Внимание:

    - `np.where(condition, x, y)` вычисляет и x, и y полностью перед применением

    - Для сложной логики можно вкладывать `np.where` (но осторожно с читаемостью)

6. Альтернативы:

    - Булева индексация: `arr[arr < 0] = 0`

    - `np.clip()` — для обрезки значений

    - `np.select()` — для множественных условий

### 20. clip:

Ограничьте значения того же массива диапазоном от -2 до 2 (все, что меньше -2, станет -2; больше 2 — станет 2).

````python
import numpy as np

# Устанавливаем seed для воспроизводимости
np.random.seed(42)

# Создаём массив случайных чисел от -5 до 5
arr = np.random.randint(-5, 6, size=(5, 5))
print("Исходный массив случайных чисел от -5 до 5:")
print(arr)
print()

# Ограничиваем значения массива диапазоном от -2 до 2 с помощью clip
clipped_arr = np.clip(arr, -2, 2)

print("Массив после clip (значения ограничены диапазоном [-2, 2]):")
print(clipped_arr)
print()

# Проверяем результат
print("Проверка:")
print(f"Минимальное значение: {np.min(clipped_arr)} (должно быть ≥ -2)")
print(f"Максимальное значение: {np.max(clipped_arr)} (должно быть ≤ 2)")
print(f"Значения < -2: {np.sum(clipped_arr < -2)}")
print(f"Значения > 2: {np.sum(clipped_arr > 2)}")
````

````python
# Различные способы использования clip
import numpy as np

np.random.seed(42)
arr = np.random.randint(-5, 6, size=(5, 5))
print("Исходный массив:")
print(arr)
print()

# 1. Базовое использование np.clip()
result1 = np.clip(arr, -2, 2)
print("1. np.clip(arr, -2, 2):")
print(result1)
print()

# 2. Метод clip() массива
result2 = arr.clip(-2, 2)
print("2. arr.clip(-2, 2):")
print(result2)
print()

# 3. Разные границы для разных операций
# Ограничиваем только снизу (все значения < 0 заменяем на 0)
result_lower = arr.clip(min=0)
print("3. arr.clip(min=0) - только нижняя граница:")
print(result_lower)
print()

# Ограничиваем только сверху (все значения > 3 заменяем на 3)
result_upper = arr.clip(max=3)
print("4. arr.clip(max=3) - только верхняя граница:")
print(result_upper)
print()

# 4. Ограничение с вычисляемыми границами
lower_bound = -1
upper_bound = 3
result_dynamic = arr.clip(lower_bound, upper_bound)
print(f"5. Ограничение с динамическими границами ({lower_bound}, {upper_bound}):")
print(result_dynamic)
````

````python
# Визуальное сравнение clip и where
import numpy as np

np.random.seed(42)
arr = np.random.randint(-10, 11, size=20)
print("Исходный массив:")
print(arr)
print()

# Clip (простой и быстрый)
clipped = np.clip(arr, -3, 3)
print("1. Clip (границы -3 и 3):")
print(clipped)
print()

# Эмуляция clip через where
clipped_where = np.where(arr < -3, -3, np.where(arr > 3, 3, arr))
print("2. Эмуляция через np.where:")
print(clipped_where)
print()

# Проверка, что результаты совпадают
print(f"Результаты совпадают: {np.array_equal(clipped, clipped_where)}")
print()

# Демонстрация отличий
print("Сравнение методов:")
print("-" * 50)
print("Значение  |  Clip  |  Where")
print("-" * 50)
for val in [-10, -5, -3, -2, 0, 2, 3, 5, 10]:
    clip_val = np.clip(val, -3, 3)
    where_val = np.where(val < -3, -3, np.where(val > 3, 3, val))
    print(f"{val:6}   |  {clip_val:4}   |  {where_val:4}")
````

````python
# Clip с массивами границ (broadcasting)
import numpy as np

# Создаём матрицу
matrix = np.arange(12).reshape(3, 4) * 2 - 5  # значения от -5 до 6
print("Исходная матрица:")
print(matrix)
print()

# 1. Одинаковые границы для всех элементов
uniform_clip = np.clip(matrix, -2, 3)
print("Одинаковые границы (-2, 3):")
print(uniform_clip)
print()

# 2. Разные нижние границы для разных строк
lower_bounds = np.array([-3, -1, 0]).reshape(3, 1)
clipped_lower = np.where(matrix < lower_bounds, lower_bounds, matrix)
print("Разные нижние границы для строк:")
print(f"Нижние границы: {lower_bounds.flatten()}")
print(clipped_lower)
print()

# 3. Разные границы для каждого элемента
min_vals = np.random.randint(-3, 0, size=(3, 4))
max_vals = np.random.randint(2, 5, size=(3, 4))
custom_clip = np.clip(matrix, min_vals, max_vals)
print("Индивидуальные границы для каждого элемента:")
print(f"Нижние границы:\n{min_vals}")
print(f"Верхние границы:\n{max_vals}")
print(f"Результат:\n{custom_clip}")
````

````python
# Сравнение производительности clip и where
import numpy as np
import time

# Создаём большой массив
np.random.seed(42)
arr_large = np.random.randn(10_000_000) * 10

# Метод 1: np.clip
start = time.time()
result1 = np.clip(arr_large, -3, 3)
time1 = time.time() - start

# Метод 2: np.where (эмуляция clip)
start = time.time()
result2 = np.where(arr_large < -3, -3, np.where(arr_large > 3, 3, arr_large))
time2 = time.time() - start

# Метод 3: Булева индексация с присваиванием (требует копирования)
arr_copy = arr_large.copy()
start = time.time()
arr_copy[arr_copy < -3] = -3
arr_copy[arr_copy > 3] = 3
time3 = time.time() - start

print("Сравнение производительности для 10,000,000 элементов:")
print(f"np.clip:              {time1:.4f} сек")
print(f"np.where (вложенный): {time2:.4f} сек")
print(f"Булева индексация:    {time3:.4f} сек")
print(f"\nnp.clip быстрее where в {time2/time1:.2f} раза")

# Проверка идентичности результатов
print(f"\nРезультаты совпадают: {np.array_equal(result1, result2)}")
````

````python
# Clip в цепочках операций
import numpy as np

# Создаём данные
data = np.array([-5, -2, 0, 3, 7, 10, 15])

# Clip часто используется для предотвращения численных проблем
print("Пример: Безопасное логарифмирование")
print(f"Исходные данные: {data}")

# Логарифм отрицательных чисел не определён
# Используем clip для гарантии положительных значений
safe_data = np.clip(data, 0.1, None)  # Минимум 0.1, максимум без ограничений
log_data = np.log(safe_data)
print(f"После clip(min=0.1): {safe_data}")
print(f"log(данных): {log_data}")
print()

# Clip при нормализации
print("Пример: Нормализация с защитой от выбросов")
values = np.array([10, 20, 30, 1000, 2000, 3000])
# Сначала отсекаем выбросы на уровне 95-го процентиля
p95 = np.percentile(values, 95)
values_clipped = np.clip(values, None, p95)
normalized = (values_clipped - np.min(values_clipped)) / (np.max(values_clipped) - np.min(values_clipped))
print(f"Исходные: {values}")
print(f"95-й процентиль: {p95}")
print(f"После clip(max={p95:.1f}): {values_clipped}")
print(f"Нормализованные: {normalized}")
````

**Ключевые моменты о clip**

1. Синтаксис:

    - `np.clip(arr, min, max)` — функция NumPy

    - `arr.clip(min, max)` — метод массива

    - Можно использовать `min=None` или `max=None` для ограничения только с одной стороны

2. Что делает:

    - Все значения < `min` заменяются на `min`

    - Все значения > `max` заменяются на `max`

    - Остальные значения остаются без изменений

3. Возвращаемое значение:

    - Возвращает новый массив (не изменяет исходный)

    - Тот же тип данных, что и исходный массив

4. Преимущества:

    - Простой и читаемый код

    - Высокая производительность (оптимизирован на C)

    - Быстрее, чем эмуляция через np.where

    - Поддерживает broadcasting

5. Применения:

    - Обработка выбросов

    - Нормализация данных

    - Ограничение значений датчиков

    - Коррекция цветовых диапазонов в изображениях

    - Предотвращение численных проблем (например, log(0))

6. Сравнение с альтернативами:

    - `np.clip` — самый быстрый и читаемый способ

    - `np.where` — более гибкий, но медленнее для простого clipping

    - Булева индексация — тоже быстрая, но изменяет исходный массив

````python
# Практический пример: Воспроизводимые эксперименты
import numpy as np

def run_experiment(seed=None):
    """Запускает эксперимент с опциональным seed"""
    if seed is not None:
        np.random.seed(seed)
    
    # Генерируем случайные данные
    data = np.random.randn(100, 5)
    
    # Вычисляем статистику
    mean = np.mean(data)
    std = np.std(data)
    
    return mean, std

print("Практический пример: Воспроизводимые эксперименты")
print("=" * 60)

# Эксперимент 1 с seed=42
print("\nЭксперимент 1 (seed=42):")
mean1, std1 = run_experiment(42)
print(f"  Среднее: {mean1:.6f}, Std: {std1:.6f}")

# Эксперимент 2 с тем же seed (должен дать тот же результат)
print("\nЭксперимент 2 (seed=42):")
mean2, std2 = run_experiment(42)
print(f"  Среднее: {mean2:.6f}, Std: {std2:.6f}")
print(f"  Результаты совпадают: {mean1 == mean2 and std1 == std2}")

# Эксперимент 3 с другим seed
print("\nЭксперимент 3 (seed=123):")
mean3, std3 = run_experiment(123)
print(f"  Среднее: {mean3:.6f}, Std: {std3:.6f}")
print(f"  Отличается от эксперимента 1: {mean1 != mean3}")

# Эксперимент 4 без seed (каждый раз разный)
print("\nЭксперимент 4 (без seed):")
mean4, std4 = run_experiment()
print(f"  Среднее: {mean4:.6f}, Std: {std4:.6f}")
````

**Ключевые моменты о воспроизводимости (Seed)**

1. Что такое seed?

    - Seed (зерно) — начальное число для генератора псевдослучайных чисел

    - При одинаковом seed генерируется одинаковая последовательность чисел

2. Как установить seed:

    - `np.random.seed(42)` — устанавливает глобальный seed

    - `np.random.RandomState(42)` — создаёт изолированный генератор

3. Важные свойства:

    - Seed определяет всю последующую последовательность случайных чисел

    - После установки seed все последующие вызовы случайных функций будут детерминированы

    - Сброс seed на то же значение перезапускает последовательность с начала

4. Когда использовать seed:

    - Для отладки и тестирования

    - Для воспроизводимости научных экспериментов

    - Для сравнения разных алгоритмов на одинаковых данных

    - В обучении с учителем (для разделения данных)

5. Современные рекомендации:

    - Использовать `np.random.RandomState` для изолированных генераторов

    - В новых проектах использовать np.random.Generator (NumPy 1.17+)

    - Для параллельных вычислений использовать разные seed для разных потоков

6. Новый API (NumPy 1.17+):

```python
# Современный способ
rng = np.random.default_rng(seed=42)
arr = rng.random(10)
```

### 23. Решение системы уравнений:
Решите систему линейных уравнений:

```text
2x+y=5
x+3y=10
``` 
Используйте np.linalg.solve(A, b), где A — матрица коэффициентов, b — вектор свободных членов.

````python
import numpy as np

# Система уравнений:
# 2x + y = 5
# x + 3y = 10

# Матрица коэффициентов A (коэффициенты при x и y)
A = np.array([[2, 1],
              [1, 3]])

# Вектор свободных членов b
b = np.array([5, 10])

print("Матрица коэффициентов A:")
print(A)
print()
print("Вектор свободных членов b:")
print(b)
print()

# Решаем систему уравнений
solution = np.linalg.solve(A, b)

print("Решение системы:")
print(f"x = {solution[0]}")
print(f"y = {solution[1]}")
print()

# Проверка: подставляем решение в уравнения
x, y = solution
check1 = 2*x + y
check2 = x + 3*y

print("Проверка:")
print(f"2x + y = 2*{x:.1f} + {y:.1f} = {check1:.1f} (должно быть 5)")
print(f"x + 3y = {x:.1f} + 3*{y:.1f} = {check2:.1f} (должно быть 10)")
````

````python
# Альтернативные способы решения
import numpy as np

A = np.array([[2, 1],
              [1, 3]])
b = np.array([5, 10])

# Способ 1: np.linalg.solve (рекомендуемый)
solution1 = np.linalg.solve(A, b)
print(f"1. np.linalg.solve(A, b): x={solution1[0]}, y={solution1[1]}")

# Способ 2: Через обратную матрицу (A^(-1) * b)
A_inv = np.linalg.inv(A)
solution2 = A_inv @ b
print(f"2. Обратная матрица: x={solution2[0]}, y={solution2[1]}")

# Способ 3: Через метод наименьших квадратов (для переопределённых систем)
solution3, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)
print(f"3. Наименьшие квадраты: x={solution3[0]}, y={solution3[1]}")

# Проверка, что все способы дают одинаковый результат
print(f"\nВсе способы дают одинаковый результат: {np.allclose(solution1, solution2)}")
````

````python
# Проверка точности решения
import numpy as np

# Создаём случайную систему
np.random.seed(42)
n = 5
A = np.random.rand(n, n)
x_true = np.random.rand(n)
b = A @ x_true

# Решаем систему
x_solution = np.linalg.solve(A, b)

print("Проверка точности решения:")
print("=" * 50)
print(f"Истинное решение x_true: {x_true}")
print(f"Найденное решение x_sol: {x_solution}")
print()

# Вычисляем погрешность
error_abs = np.abs(x_true - x_solution)
error_rel = error_abs / np.abs(x_true)
max_error = np.max(error_abs)

print(f"Максимальная абсолютная погрешность: {max_error:.2e}")
print(f"Максимальная относительная погрешность: {np.max(error_rel):.2e}")
print()

# Проверка: A * x_solution должно равняться b
b_check = A @ x_solution
b_diff = np.abs(b - b_check)

print(f"Максимальная погрешность в правой части: {np.max(b_diff):.2e}")
print(f"Решение найдено корректно: {np.allclose(b, b_check)}")
````

````python
# Практический пример: Балансировка химической реакции
import numpy as np

# Балансировка реакции горения пропана:
# a·C3H8 + b·O2 → c·CO2 + d·H2O

# Составляем уравнения баланса по каждому элементу:
# Углерод (C): 3a = c          → 3a - c = 0
# Водород (H): 8a = 2d        → 8a - 2d = 0  
# Кислород (O): 2b = 2c + d   → 2b - 2c - d = 0

# Переносим всё в левую часть, получаем систему:
# 3a + 0b - c + 0d = 0
# 8a + 0b + 0c - 2d = 0
# 0a + 2b - 2c - d = 0

# Записываем в матричной форме: A * [a, b, c, d]^T = 0
# Но система однородная (все правые части = 0), поэтому нужно добавить
# дополнительное условие (например, a = 1) или использовать метод наименьших квадратов

# Способ 1: Фиксируем a = 1 и решаем для b, c, d
print("Балансировка реакции горения пропана:")
print("a·C3H8 + b·O2 → c·CO2 + d·H2O")
print()

# Из уравнений при a = 1:
# c = 3a = 3
# d = 8a/2 = 4
# 2b = 2c + d = 2*3 + 4 = 10 → b = 5

print("Способ 1 (ручной подстановкой a=1):")
print(f"a = 1, b = 5, c = 3, d = 4")
print("C3H8 + 5O2 → 3CO2 + 4H2O")
print()

# Способ 2: Использование np.linalg.solve с нормализацией
# Перепишем уравнения в виде: 
# 3a - c = 0
# 8a - 2d = 0  
# 2b - 2c - d = 0

# Добавляем условие a = 1 (или любую другую нормировку)
# Получаем систему:
# 1·a + 0·b + 0·c + 0·d = 1  (нормировка)
# 3·a + 0·b - 1·c + 0·d = 0
# 8·a + 0·b + 0·c - 2·d = 0
# 0·a + 2·b - 2·c - 1·d = 0

A = np.array([
    [1, 0, 0, 0],   # a = 1
    [3, 0, -1, 0],  # 3a - c = 0
    [8, 0, 0, -2],  # 8a - 2d = 0
    [0, 2, -2, -1]  # 2b - 2c - d = 0
])

b = np.array([1, 0, 0, 0])  # Правая часть

solution = np.linalg.solve(A, b)
a, b_coef, c, d = solution

print("Способ 2 (решение через np.linalg.solve):")
print(f"a = {a:.0f}, b = {b_coef:.0f}, c = {c:.0f}, d = {d:.0f}")
print(f"{a:.0f}C3H8 + {b_coef:.0f}O2 → {c:.0f}CO2 + {d:.0f}H2O")
````

````python
# Решение системы с комплексными числами
import numpy as np

# Система с комплексными коэффициентами
# (1+2i)x + (3- i)y = 5+3i
# (2- i)x + (1+ i)y = 2- i

A_complex = np.array([[1+2j, 3-1j],
                      [2-1j, 1+1j]])

b_complex = np.array([5+3j, 2-1j])

solution_complex = np.linalg.solve(A_complex, b_complex)

print("Решение системы с комплексными числами:")
print("=" * 50)
print(f"Матрица A:\n{A_complex}")
print(f"Вектор b: {b_complex}")
print()
print(f"Решение: x = {solution_complex[0]}, y = {solution_complex[1]}")
print()

# Проверка
check = A_complex @ solution_complex
print(f"Проверка A*x = {check}")
print(f"Совпадает с b: {np.allclose(check, b_complex)}")
````

````python
# Экономическая модель
import numpy as np

print("=" * 60)
print("ЭКОНОМИЧЕСКИЙ ПРИМЕР: Равновесие рынка")
print("=" * 60)

# Задача: Найти равновесную цену и количество для двух товаров
# Товар 1: Qd1 = 10 - 2P1 + P2 (спрос)
#          Qs1 = 2 + 3P1 - P2  (предложение)
# Товар 2: Qd2 = 15 + P1 - 3P2
#          Qs2 = 5 - P1 + 2P2

# В равновесии: Qd1 = Qs1 и Qd2 = Qs2

# Уравнение для товара 1:
# 10 - 2P1 + P2 = 2 + 3P1 - P2
# 10 - 2P1 + P2 - 2 - 3P1 + P2 = 0
# 8 - 5P1 + 2P2 = 0
# -5P1 + 2P2 = -8

# Уравнение для товара 2:
# 15 + P1 - 3P2 = 5 - P1 + 2P2
# 15 + P1 - 3P2 - 5 + P1 - 2P2 = 0
# 10 + 2P1 - 5P2 = 0
# 2P1 - 5P2 = -10

# Система уравнений:
# -5P1 + 2P2 = -8
# 2P1 - 5P2 = -10

A = np.array([
    [-5, 2],
    [2, -5]
])

b = np.array([-8, -10])

print("Система уравнений для равновесных цен:")
print("-5P₁ + 2P₂ = -8")
print("2P₁ - 5P₂ = -10")
print()

# Решаем систему
P1, P2 = np.linalg.solve(A, b)

print(f"Равновесные цены:")
print(f"P₁ = {P1:.2f}")
print(f"P₂ = {P2:.2f}")
print()

# Находим равновесные количества
Q1 = 10 - 2*P1 + P2
Q2 = 15 + P1 - 3*P2

print(f"Равновесные количества:")
print(f"Q₁ = {Q1:.2f}")
print(f"Q₂ = {Q2:.2f}")
````

**Ключевые моменты о решении систем уравнений**

1. Формат вызова:

```python
np.linalg.solve(A, b)
```

    - A — квадратная матрица коэффициентов (n×n)

    - b — вектор правых частей (n,)

2. Требования:

    - Матрица A должна быть квадратной

    - Матрица A должна быть невырожденной (det ≠ 0)

    - Система должна иметь единственное решение

3. Альтернативные методы:

    - `np.linalg.inv(A) @ b` — через обратную матрицу (менее устойчиво)

    - `np.linalg.lstsq(A, b)` — для переопределённых систем

    - `np.linalg.pinv(A) @ b` — через псевдообратную матрицу

4. Преимущества `np.linalg.solve`:

    - Численно устойчив

    - Быстрее, чем вычисление обратной матрицы

    - Оптимизирован для решения линейных систем

5. Ошибки:

    - `LinAlgError` — если матрица вырождена или не квадратная

6. Применения:

    - Решение систем уравнений

    - Нахождение равновесий

    - Интерполяция и аппроксимация

    - Экономические модели

    - Инженерные расчёты

### 24. Определитель и обратная матрица:

Создайте матрицу 3x3 со случайными целыми числами. Вычислите её определитель (det) и обратную матрицу (inv). Проверьте, что произведение матрицы на обратную дает единичную матрицу (с учетом погрешности вычислений).

````python
import numpy as np

# Устанавливаем seed для воспроизводимости
np.random.seed(42)

# 1. Создаём матрицу 3x3 со случайными целыми числами
matrix = np.random.randint(1, 10, size=(3, 3))
print("Исходная матрица:")
print(matrix)
print()

# 2. Вычисляем определитель
det = np.linalg.det(matrix)
print(f"Определитель: {det:.2f}")
print()

# 3. Вычисляем обратную матрицу
inv_matrix = np.linalg.inv(matrix)
print("Обратная матрица:")
print(inv_matrix)
print()

# 4. Проверяем: матрица * обратная матрица = единичная матрица
identity = matrix @ inv_matrix
print("Произведение матрицы на обратную:")
print(identity)
print()

# Округляем до 2 знаков для наглядности
print("Проверка (с округлением):")
print(np.round(identity, 2))
print()

# Проверяем с учётом погрешности вычислений
is_identity = np.allclose(identity, np.eye(3))
print(f"Произведение даёт единичную матрицу? {is_identity}")
````

````python
# Краткое объяснение
import numpy as np

# Создаём матрицу
M = np.array([[2, 1],
              [1, 3]])

# Определитель (det) — число, характеризующее матрицу
det = np.linalg.det(M)
print(f"Определитель: {det}")  # 5.0

# Обратная матрица (inv) — такая матрица, что M * M⁻¹ = I
M_inv = np.linalg.inv(M)
print(f"Обратная матрица:\n{M_inv}")

# Проверка: M * M_inv ≈ единичной матрице
check = M @ M_inv
print(f"Проверка:\n{np.round(check, 2)}")  # [[1. 0.] [0. 1.]]

# np.allclose() учитывает погрешности вычислений
print(f"Всё верно? {np.allclose(check, np.eye(2))}")
````

**Ключевые моменты:**

* `np.linalg.det()` — вычисляет определитель (не должен быть 0 для обратимости)

* `np.linalg.inv()` — находит обратную матрицу

* `np.allclose()` — проверяет равенство с учётом погрешности

* `np.eye(n)` — создаёт единичную матрицу размера n
