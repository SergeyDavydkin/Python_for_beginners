````python
class Employee:
  def __init__(self, name, salary, experience):
    self.name = name # ФИО
    self.salary = salary # оклад
    self.experience = experience # стаж

  # Универсальный суперкласс сотрудников
  def compute_salary(self) : 
    # Cтандартный расчет зарплаты
    return self.salary * (100 + self.__get_persent()) / 100

  def __get_persent(self):
    if (self.experience < 5):
      return 0
    elif  (self.experience < 15):
      return 10
    else:
      return 15

Ivanov = Employee('Ivanov I.I.', 90000, 7)
print(Ivanov.compute_salary())
````

````python
class Manager(Employee):
  def __init__(self, name, salary, experience, contract):
    super().__init__(name, salary, experience)
    self.contract = contract

  # Универсальный суперкласс сотрудников
  def compute_salary (self) : 
    # Cтандартный расчет зарплаты
    return super().compute_salary() + self.contract * self.__get_contract_coast()

  def __get_contract_coast(self):
    if (self.contract < 50):
      return 500
    elif  (self.contract < 100):
      return 1000
    else:
      return 1500

Petrov = Manager('Petrov P.P.', 90000, 7, 13)
print(Petrov.compute_salary())
Manager.compute_salary(Petrov)
````

````python
import math
class Fraction():
  def __init__(self, numerator, denominator):
    gcd = math.gcd(numerator, denominator)
    self.numerator = numerator // gcd
    self.denominator = denominator // gcd
  
  def __mul__(self, fraction):
    if isinstance(fraction, Fraction):
      return Fraction(self.numerator * fraction.numerator, (self.denominator * fraction.denominator))
    

  def __str__(self):
    return str(self.numerator) + '/' + str(self.denominator)

b = Fraction(2, 4)
a = Fraction(2, 4)
a.__str__()

print(a * b)
````

````python
from datetime import datetime

class Book:
    """Класс, представляющий книгу."""
    
    def __init__(self, title, author, year):
        """Конструктор: создаёт книгу с названием, автором и годом."""
        self.title = title
        self.author = author
        self.year = year
    
    def get_age(self):
        """Возвращает возраст книги в годах."""
        current_year = datetime.now().year
        return current_year - self.year
    
    def __str__(self):
        """Возвращает красивое строковое представление книги."""
        return f'"{self.title}" ({self.author}, {self.year})'
````

````python
class Vehicle:
    """Базовый класс для всех транспортных средств."""
    
    def __init__(self, brand, speed):
        """
        Конструктор транспортного средства.
        
        Args:
            brand: Марка транспортного средства
            speed: Скорость (км/ч)
        """
        self.brand = brand
        self.speed = speed
    
    def info(self):
        """
        Возвращает информацию о транспортном средстве.
        
        Returns:
            Строка с информацией
        """
        return f"Транспорт: {self.brand}, Скорость: {self.speed} км/ч"


class Car(Vehicle):
    """Класс автомобиля, наследуется от Vehicle."""
    
    def __init__(self, brand, speed, fuel_type):
        """
        Конструктор автомобиля.
        
        Args:
            brand: Марка автомобиля
            speed: Скорость (км/ч)
            fuel_type: Тип топлива (бензин, дизель, электричество)
        """
        # Вызываем конструктор родителя (Vehicle)
        super().__init__(brand, speed)
        # Добавляем новый атрибут
        self.fuel_type = fuel_type
    
    def info(self):
        """
        Переопределяем метод info() для автомобиля.
        
        Returns:
            Расширенная информация об автомобиле
        """
        # Используем родительский метод и добавляем новое
        parent_info = super().info()
        return f"{parent_info}, Топливо: {self.fuel_type}"


class Bicycle(Vehicle):
    """Класс велосипеда, наследуется от Vehicle."""
    
    def __init__(self, brand, speed, has_bell):
        """
        Конструктор велосипеда.
        
        Args:
            brand: Марка велосипеда
            speed: Скорость (км/ч)
            has_bell: Есть ли звонок (True/False)
        """
        # Вызываем конструктор родителя (Vehicle)
        super().__init__(brand, speed)
        # Добавляем новый атрибут
        self.has_bell = has_bell
    
    def info(self):
        """
        Переопределяем метод info() для велосипеда.
        
        Returns:
            Информация о велосипеде с указанием наличия звонка
        """
        parent_info = super().info()
        bell_status = "со звонком" if self.has_bell else "без звонка"
        return f"{parent_info}, {bell_status}"
    
    def ring_bell(self):
        """
        Метод звонка велосипеда.
        
        Returns:
            Звук звонка или сообщение об отсутствии звонка
        """
        if self.has_bell:
            return "Дзынь-дзынь! 🚲"
        else:
            return "У велосипеда нет звонка!"


# ========== ПРИМЕРЫ ИСПОЛЬЗОВАНИЯ ==========

if __name__ == "__main__":
    # Создаем автомобили
    car1 = Car("Toyota", 180, "бензин")
    car2 = Car("Tesla", 220, "электричество")
    car3 = Car("BMW", 200, "дизель")
    
    # Создаем велосипеды
    bike1 = Bicycle("Stels", 25, True)
    bike2 = Bicycle("GT", 30, False)
    
    # Выводим информацию
    print("=== АВТОМОБИЛИ ===")
    print(car1.info())
    print(car2.info())
    print(car3.info())
    
    print("\n=== ВЕЛОСИПЕДЫ ===")
    print(bike1.info())
    print(bike2.info())
    
    # Используем метод звонка
    print("\n=== ЗВОНОК ===")
    print(f"Велосипед {bike1.brand}: {bike1.ring_bell()}")
    print(f"Велосипед {bike2.brand}: {bike2.ring_bell()}")
    
    # Демонстрация наследования
    print("\n=== ДОКАЗАТЕЛЬСТВО НАСЛЕДОВАНИЯ ===")
    print(f"Car наследует от Vehicle: {issubclass(Car, Vehicle)}")  # True
    print(f"Bicycle наследует от Vehicle: {issubclass(Bicycle, Vehicle)}")  # True
    print(f"car1 - это Vehicle? {isinstance(car1, Vehicle)}")  # True
    print(f"bike1 - это Vehicle? {isinstance(bike1, Vehicle)}")  # True
````

```python

#   +  — сложение векторов

**🔍 Ключевые концепции магических методов**

|Магический метод	|Оператор	|Вызывается когда	|Пример
|-------------------|-----------|-------------------|---------
|`__add__(self, other)`	|`+`	|`v1 + v2`	|Сложение векторов

|`__mul__(self, scalar)`	|`*`	|`v * 2`	|Умножение на число
|`__rmul__(self, scalar)`	|`*`	|`2 * v`	|Число умножить на вектор
|`__eq__(self, other)`	`==`	|`v1 == v2`	|Сравнение векторов
|`__repr__(self)`	|`repr()`	|`print(repr(v))`	|Отладочное представление
|`__str__(self)`	|`str()`	|`print(v)`	|Пользовательское представление

````python
class Vector2D:
    """Двумерный вектор с координатами x и y."""
    
    def __init__(self, x, y):
        """
        Конструктор вектора.
        
        Args:
            x: Координата по оси X
            y: Координата по оси Y
        """
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """
        Сложение векторов: v1 + v2
        
        Args:
            other: Другой вектор Vector2D
        
        Returns:
            Новый вектор с суммой координат
        """
        if isinstance(other, Vector2D):
            return Vector2D(self.x + other.x, self.y + other.y)
        return NotImplemented
    
    def __sub__(self, other):
        """
        Вычитание векторов: v1 - v2
        
        Args:
            other: Другой вектор Vector2D
        
        Returns:
            Новый вектор с разностью координат
        """
        if isinstance(other, Vector2D):
            return Vector2D(self.x - other.x, self.y - other.y)
        return NotImplemented
    
    def __mul__(self, scalar):
        """
        Умножение вектора на число (скаляр): v * n
        
        Args:
            scalar: Число (int или float)
        
        Returns:
            Новый вектор с умноженными координатами
        """
        if isinstance(scalar, (int, float)):
            return Vector2D(self.x * scalar, self.y * scalar)
        return NotImplemented
    
    def __rmul__(self, scalar):
        """
        Умножение числа на вектор (для поддержки n * v)
        
        Args:
            scalar: Число (int или float)
        
        Returns:
            Новый вектор с умноженными координатами
        """
        return self.__mul__(scalar)
    
    def __eq__(self, other):
        """
        Сравнение векторов: v1 == v2
        
        Args:
            other: Другой вектор Vector2D
        
        Returns:
            True если координаты равны, иначе False
        """
        if isinstance(other, Vector2D):
            return self.x == other.x and self.y == other.y
        return False
    
    def __repr__(self):
        """
        Представление вектора для разработчиков.
        
        Returns:
            Строка вида "Vector2D(3, 4)"
        """
        return f"Vector2D({self.x}, {self.y})"
    
    def __str__(self):
        """
        Понятное строковое представление для пользователей.
        
        Returns:
            Строка вида "(3, 4)"
        """
        return f"({self.x}, {self.y})"


# ========== ПРИМЕРЫ ИСПОЛЬЗОВАНИЯ ==========


# Создаем векторы
v1 = Vector2D(3, 4)
v2 = Vector2D(1, 2)

print("=== СОЗДАНИЕ ВЕКТОРОВ ===")
print(f"v1 = {v1}")          # (3, 4)
print(f"v2 = {v2}")          # (1, 2)
print(f"repr(v1) = {repr(v1)}")  # Vector2D(3, 4)

print("\n=== СЛОЖЕНИЕ ===")
v3 = v1 + v2
print(f"v1 + v2 = {v3}")     # (4, 6)
print(f"repr: {repr(v3)}")    # Vector2D(4, 6)

print("\n=== ВЫЧИТАНИЕ ===")
v4 = v1 - v2
print(f"v1 - v2 = {v4}")     # (2, 2)

print("\n=== УМНОЖЕНИЕ НА ЧИСЛО ===")
v5 = v1 * 2
print(f"v1 * 2 = {v5}")      # (6, 8)

v6 = 3 * v1
print(f"3 * v1 = {v6}")      # (9, 12) - работает благодаря __rmul__

print("\n=== СРАВНЕНИЕ ===")
v7 = Vector2D(3, 4)
print(f"v1 == v2 -> {v1 == v2}")  # False
print(f"v1 == v7 -> {v1 == v7}")  # True

print("\n=== ЦЕПОЧКИ ОПЕРАЦИЙ ===")
result = (v1 + v2) * 2
print(f"(v1 + v2) * 2 = {result}")  # (8, 12)

result2 = (v1 * 2) + (v2 * 3)
print(f"(v1 * 2) + (v2 * 3) = {result2}")  # (9, 14)
````

````python
class Time:
    """Класс для работы со временем (часы и минуты)."""
    
    def __init__(self, hours, minutes):
        """
        Конструктор времени.
        
        Args:
            hours: Часы (0-23)
            minutes: Минуты (0-59)
        """
        # Нормализуем время при создании
        self.hours = hours
        self.minutes = minutes
        self._normalize()
    
    def _normalize(self):
        """Приводит время к нормальному формату (минуты < 60, часы 0-23)."""
        # Переносим лишние минуты в часы
        extra_hours = self.minutes // 60
        self.minutes = self.minutes % 60
        self.hours += extra_hours
        
        # Приводим часы к диапазону 0-23 (сутки)
        self.hours = self.hours % 24
    
    def __eq__(self, other):
        """Сравнение: time1 == time2."""
        if isinstance(other, Time):
            return self.hours == other.hours and self.minutes == other.minutes
        return False
    
    def __lt__(self, other):
        """Сравнение: time1 < time2."""
        if isinstance(other, Time):
            # Переводим всё в минуты для сравнения
            self_total = self.hours * 60 + self.minutes
            other_total = other.hours * 60 + other.minutes
            return self_total < other_total
        return NotImplemented
    
    def __le__(self, other):
        """Сравнение: time1 <= time2."""
        return self < other or self == other
    
    def __gt__(self, other):
        """Сравнение: time1 > time2."""
        return not (self < other or self == other)
    
    def __ge__(self, other):
        """Сравнение: time1 >= time2."""
        return not (self < other)
    
    def __add__(self, minutes):
        """
        Прибавляет минуты ко времени.
        
        Args:
            minutes: Количество минут для прибавления
        
        Returns:
            Новый объект Time с прибавленными минутами
        """
        if isinstance(minutes, (int, float)):
            new_minutes = self.hours * 60 + self.minutes + minutes
            new_hours = new_minutes // 60
            new_minutes = new_minutes % 60
            new_hours = new_hours % 24
            return Time(new_hours, new_minutes)
        return NotImplemented
    
    def __radd__(self, minutes):
        """Поддерживает запись: minutes + time (для симметрии)."""
        return self.__add__(minutes)
    
    def __str__(self):
        """Красивый вывод времени."""
        return f"{self.hours:02d}:{self.minutes:02d}"
    
    def __repr__(self):
        """Представление для разработчиков."""
        return f"Time({self.hours}, {self.minutes})"


# ========== ПРИМЕРЫ ИСПОЛЬЗОВАНИЯ ==========

# Создаем время
t1 = Time(10, 30)
t2 = Time(14, 45)
t3 = Time(10, 30)

print("=== СОЗДАНИЕ ВРЕМЕНИ ===")
print(f"t1 = {t1}")  # 10:30
print(f"t2 = {t2}")  # 14:45
print(f"t3 = {t3}")  # 10:30

print("\n=== СРАВНЕНИЕ ===")
print(f"t1 == t2 -> {t1 == t2}")  # False
print(f"t1 == t3 -> {t1 == t3}")  # True
print(f"t1 < t2  -> {t1 < t2}")   # True
print(f"t1 > t2  -> {t1 > t2}")   # False
print(f"t1 <= t3 -> {t1 <= t3}")  # True
print(f"t2 >= t1 -> {t2 >= t1}")  # True

print("\n=== СЛОЖЕНИЕ МИНУТ ===")
t4 = Time(23, 50)
print(f"{t4} + 20 = {t4 + 20}")   # 00:10 (перенос через сутки)

t5 = Time(1, 45)
print(f"{t5} + 30 = {t5 + 30}")   # 02:15

t6 = Time(0, 0)
print(f"{t6} + 1440 = {t6 + 1440}")  # 00:00 (ровно сутки)

print("\n=== ЦЕПОЧКИ ОПЕРАЦИЙ ===")
t7 = Time(22, 30)
result = t7 + 45 + 30
print(f"{t7} + 45 + 30 = {result}")  # 23:45

print("\n=== ПРИМЕРЫ ИЗ ЖИЗНИ ===")
now = Time(9, 0)
print(f"Текущее время: {now}")
print(f"Через 3 часа (180 мин): {now + 180}")
print(f"Через 1.5 часа (90 мин): {now + 90}")
print(f"Через сутки (1440 мин): {now + 1440}")
print(f"Через 25 часов (1500 мин): {now + 1500}")  # 10:00 (следующий день)
````

### 7. Система "Библиотека"

Цель: Связь объектов, композиция.

```python
# Классы: Book (название, автор, is_available), Reader (имя, список взятых книг),
# Library (список книг).
# Методы Library:
#   add_book(book)
#   lend_book(book_title, reader) — если книга доступна, отдать читателю
#   return_book(book_title, reader)
#   list_available_books()
# Reader может иметь не более 3 книг одновременно.

**🔍 Ключевые концепции**

|Концепция	|Где используется	|Зачем
|-----------|-------------------|----------
|Композиция	|Library содержит списки Book и Reader	|Библиотека "состоит из" книг и читателей
|Агрегация	|Reader имеет список взятых книг	|Читатель "имеет" книги (книги существуют отдельно)
|Связь объектов	|Book ↔ Reader через статус is_available	|Книга знает, выдана ли она
|Инкапсуляция	|MAX_BOOKS, borrowed_books	|Читатель сам следит за лимитом
|Поиск	|find_book_by_title, find_reader_by_name	|Навигация между объектами

````python
# Шаг 1: Класс Book
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True
    
    def __str__(self):
        status = "доступна" if self.is_available else "выдана"
        return f'"{self.title}" ({self.author}) - {status}'

# Проверка
b = Book("1984", "Оруэлл")
print(b)              # "1984" (Оруэлл) - доступна
print(b.title)        # 1984
print(b.is_available) # True

b.is_available = False
print(b)              # "1984" (Оруэлл) - выдана
````

````python
# Шаг 2: Класс Reader
class Reader:
    MAX_BOOKS = 3
    
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []
    
    def borrow_book(self, book):
        if len(self.borrowed_books) < self.MAX_BOOKS and book.is_available:
            self.borrowed_books.append(book)
            book.is_available = False
            return True
        return False
    
    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.is_available = True
            return True
        return False
    
    def __str__(self):
        """Красивый вывод читателя."""
        return f"{self.name} (книг: {len(self.borrowed_books)}/{self.MAX_BOOKS})"


# Проверка
r = Reader("Иван")
b1 = Book("1984", "Оруэлл")
b2 = Book("Война и мир", "Толстой")

print(r)                    # Иван (книг: 0/3)
print(r.borrow_book(b1))    # True
print(r.borrow_book(b2))    # True
print(r)                    # Иван (книг: 2/3)
print(b1)                   # "1984" (Оруэлл) - выдана
print(r.return_book(b1))    # True
print(r)                    # Иван (книг: 1/3)
print(b1)                   # "1984" (Оруэлл) - доступна

````

````python
# Шаг 3: Базовый класс Library
class Library:
    def __init__(self, name):
        self.name = name
        self.books = []
        self.readers = []
    
    def add_book(self, book):
        self.books.append(book)
    
    def add_reader(self, reader):
        self.readers.append(reader)
    
    def __str__(self):
        """Красивый вывод библиотеки."""
        return f"Библиотека '{self.name}': {len(self.books)} книг, {len(self.readers)} читателей"


# Проверка
lib = Library("Школьная")
print(lib)                          # Библиотека 'Школьная': 0 книг, 0 читателей

b = Book("1984", "Оруэлл")
r = Reader("Иван")

lib.add_book(b)
lib.add_reader(r)
print(lib)                          # Библиотека 'Школьная': 1 книг, 1 читателей
print(len(lib.books))               # 1
print(len(lib.readers))             # 1
print(lib.books[0].title)           # 1984
print(lib.readers[0].name)          # Иван
````

````python
class Book:
    """Класс, представляющий книгу в библиотеке."""
    
    def __init__(self, title, author):
        """
        Конструктор книги.
        
        Args:
            title: Название книги
            author: Автор книги
        """
        self.title = title
        self.author = author
        self.is_available = True  # По умолчанию книга доступна
    
    def __str__(self):
        """Красивый вывод книги."""
        status = "доступна" if self.is_available else "выдана"
        return f'"{self.title}" ({self.author}) - {status}'


class Reader:
    """Класс, представляющий читателя библиотеки."""
    
    MAX_BOOKS = 3  # Максимум книг у одного читателя
    
    def __init__(self, name):
        """
        Конструктор читателя.
        
        Args:
            name: Имя читателя
        """
        self.name = name
        self.borrowed_books = []  # Список взятых книг
    
    def can_borrow(self):
        """Проверяет, может ли читатель взять ещё книгу."""
        return len(self.borrowed_books) < self.MAX_BOOKS
    
    def borrow_book(self, book):
        """
        Выдаёт книгу читателю.
        
        Args:
            book: Объект книги
        
        Returns:
            True если удалось выдать, False если нет
        """
        if self.can_borrow() and book.is_available:
            self.borrowed_books.append(book)
            book.is_available = False
            return True
        return False
    
    def return_book(self, book):
        """
        Возвращает книгу от читателя.
        
        Args:
            book: Объект книги
        
        Returns:
            True если книга была у читателя, False если нет
        """
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.is_available = True
            return True
        return False
    
    def get_borrowed_books(self):
        """Возвращает список взятых книг."""
        return self.borrowed_books
    
    def __str__(self):
        """Красивый вывод читателя."""
        books_count = len(self.borrowed_books)
        return f"{self.name} (книг: {books_count}/{self.MAX_BOOKS})"


class Library:
    """Класс, представляющий библиотеку."""
    
    def __init__(self, name):
        """
        Конструктор библиотеки.
        
        Args:
            name: Название библиотеки
        """
        self.name = name
        self.books = []  # Список всех книг в библиотеке
        self.readers = []  # Список всех читателей
    
    def add_book(self, book):
        """
        Добавляет книгу в библиотеку.
        
        Args:
            book: Объект книги
        """
        self.books.append(book)
        print(f"📚 Добавлена книга: {book.title}")
    
    def add_reader(self, reader):
        """
        Добавляет читателя в библиотеку.
        
        Args:
            reader: Объект читателя
        """
        self.readers.append(reader)
        print(f"👤 Добавлен читатель: {reader.name}")
    
    def find_book_by_title(self, title):
        """
        Ищет книгу по названию.
        
        Args:
            title: Название книги
        
        Returns:
            Объект книги или None
        """
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        return None
    
    def find_reader_by_name(self, name):
        """
        Ищет читателя по имени.
        
        Args:
            name: Имя читателя
        
        Returns:
            Объект читателя или None
        """
        for reader in self.readers:
            if reader.name.lower() == name.lower():
                return reader
        return None
    
    def lend_book(self, book_title, reader_name):
        """
        Выдаёт книгу читателю.
        
        Args:
            book_title: Название книги
            reader_name: Имя читателя
        """
        # Находим книгу
        book = self.find_book_by_title(book_title)
        if not book:
            print(f"❌ Книга '{book_title}' не найдена в библиотеке")
            return
        
        # Находим читателя
        reader = self.find_reader_by_name(reader_name)
        if not reader:
            print(f"❌ Читатель '{reader_name}' не найден")
            return
        
        # Проверяем доступность книги
        if not book.is_available:
            print(f"❌ Книга '{book_title}' уже выдана")
            return
        
        # Проверяем, может ли читатель взять ещё книгу
        if not reader.can_borrow():
            print(f"❌ Читатель {reader_name} уже взял {Reader.MAX_BOOKS} книг")
            return
        
        # Выдаём книгу
        if reader.borrow_book(book):
            print(f"✅ Книга '{book_title}' выдана читателю {reader_name}")
        else:
            print(f"❌ Не удалось выдать книгу")
    
    def return_book(self, book_title, reader_name):
        """
        Принимает книгу от читателя.
        
        Args:
            book_title: Название книги
            reader_name: Имя читателя
        """
        # Находим книгу
        book = self.find_book_by_title(book_title)
        if not book:
            print(f"❌ Книга '{book_title}' не найдена в библиотеке")
            return
        
        # Находим читателя
        reader = self.find_reader_by_name(reader_name)
        if not reader:
            print(f"❌ Читатель '{reader_name}' не найден")
            return
        
        # Возвращаем книгу
        if reader.return_book(book):
            print(f"✅ Книга '{book_title}' возвращена читателем {reader_name}")
        else:
            print(f"❌ У читателя {reader_name} нет книги '{book_title}'")
    
    def list_available_books(self):
        """Показывает все доступные книги в библиотеке."""
        available_books = [book for book in self.books if book.is_available]
        
        print(f"\n📚 ДОСТУПНЫЕ КНИГИ В БИБЛИОТЕКЕ '{self.name}':")
        print("=" * 50)
        
        if not available_books:
            print("   Нет доступных книг")
        else:
            for i, book in enumerate(available_books, 1):
                print(f"   {i}. {book}")
        print("=" * 50)
    
    def list_borrowed_books(self):
        """Показывает все выданные книги."""
        borrowed_books = [book for book in self.books if not book.is_available]
        
        print(f"\n📖 ВЫДАННЫЕ КНИГИ:")
        print("=" * 50)
        
        if not borrowed_books:
            print("   Нет выданных книг")
        else:
            for i, book in enumerate(borrowed_books, 1):
                print(f"   {i}. {book}")
        print("=" * 50)
    
    def list_readers(self):
        """Показывает всех читателей и взятые ими книги."""
        print(f"\n👥 ЧИТАТЕЛИ БИБЛИОТЕКИ '{self.name}':")
        print("=" * 50)
        
        if not self.readers:
            print("   Нет читателей")
        else:
            for reader in self.readers:
                print(f"\n   📌 {reader}")
                if reader.borrowed_books:
                    print(f"      Взятые книги:")
                    for book in reader.borrowed_books:
                        print(f"        - {book.title}")
                else:
                    print(f"      Нет взятых книг")
        print("=" * 50)
    
    def library_info(self):
        """Показывает полную информацию о библиотеке."""
        print("\n" + "=" * 60)
        print(f"🏛️  БИБЛИОТЕКА '{self.name}'")
        print("=" * 60)
        print(f"📚 Всего книг: {len(self.books)}")
        print(f"📖 Доступно: {len([b for b in self.books if b.is_available])}")
        print(f"📕 Выдано: {len([b for b in self.books if not b.is_available])}")
        print(f"👥 Читателей: {len(self.readers)}")
        print("=" * 60)


# ========== ПРИМЕР ИСПОЛЬЗОВАНИЯ ==========


# Создаём библиотеку
library = Library("Центральная городская библиотека")

# Создаём книги
book1 = Book("1984", "Джордж Оруэлл")
book2 = Book("Мастер и Маргарита", "Михаил Булгаков")
book3 = Book("Война и мир", "Лев Толстой")
book4 = Book("Преступление и наказание", "Фёдор Достоевский")
book5 = Book("Гарри Поттер", "Дж.К. Роулинг")
book6 = Book("Мёртвые души", "Николай Гоголь")

# Добавляем книги в библиотеку
print("=== ДОБАВЛЕНИЕ КНИГ ===")
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_book(book4)
library.add_book(book5)
library.add_book(book6)

# Создаём читателей
reader1 = Reader("Иван Петров")
reader2 = Reader("Мария Сидорова")
reader3 = Reader("Алексей Иванов")

# Добавляем читателей в библиотеку
print("\n=== ДОБАВЛЕНИЕ ЧИТАТЕЛЕЙ ===")
library.add_reader(reader1)
library.add_reader(reader2)
library.add_reader(reader3)

# Показываем информацию о библиотеке
library.library_info()

# Показываем доступные книги
library.list_available_books()

# Выдаём книги
print("\n=== ВЫДАЧА КНИГ ===")
library.lend_book("1984", "Иван Петров")
library.lend_book("Мастер и Маргарита", "Иван Петров")
library.lend_book("Война и мир", "Иван Петров")
library.lend_book("Преступление и наказание", "Иван Петров")  # Попытка взять 4-ю книгу

library.lend_book("Гарри Поттер", "Мария Сидорова")
library.lend_book("Мёртвые души", "Мария Сидорова")
library.lend_book("Война и мир", "Мария Сидорова")  # Эта книга уже выдана

# Показываем доступные книги после выдачи
library.list_available_books()

# Показываем выданные книги
library.list_borrowed_books()

# Показываем читателей
library.list_readers()

# Возвращаем книги
print("\n=== ВОЗВРАТ КНИГ ===")
library.return_book("1984", "Иван Петров")
library.return_book("Война и мир", "Иван Петров")
library.return_book("Гарри Поттер", "Мария Сидорова")

# Показываем финальное состояние
library.library_info()
library.list_readers()
````

````python
# Шаг 3: Добавляем метод get_average_grade() в Student
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = {}
    
    def add_grade(self, course_name, grade):
        if 1 <= grade <= 5:
            self.grades[course_name] = grade
            return True
        return False
    
    def get_average_grade(self):
        """Возвращает средний балл студента."""
        if not self.grades:
            return 0
        return sum(self.grades.values()) / len(self.grades)
    
    def __str__(self):
        return f"{self.name} (ID: {self.student_id}, ср.балл: {self.get_average_grade():.1f})"

# Проверка
s = Student("Алексей", "S003")
s.add_grade("Математика", 5)
s.add_grade("Физика", 4)
s.add_grade("Программирование", 3)
print(s.get_average_grade())  # 4.0
print(s)  # Алексей (ID: S003, ср.балл: 4.0)
````

````python
# Шаг 6: Добавляем метод get_average_grade() в Course
class Course:
    def __init__(self, name, teacher):
        self.name = name
        self.teacher = teacher
        self.students = []
    
    def add_student(self, student):
        if student not in self.students:
            self.students.append(student)
            return True
        return False
    
    def remove_student(self, student):
        if student in self.students:
            self.students.remove(student)
            return True
        return False
    
    def get_average_grade(self):
        """Средний балл всех студентов на курсе."""
        if not self.students:
            return 0
        total = 0
        count = 0
        for student in self.students:
            if self.name in student.grades:
                total += student.grades[self.name]
                count += 1
        return total / count if count > 0 else 0
    
    def __str__(self):
        return f"Курс: {self.name}, Преподаватель: {self.teacher}, Студентов: {len(self.students)}"

# Проверка
c = Course("Программирование", "Иванова")
s1 = Student("Иван", "S001")
s2 = Student("Мария", "S002")

s1.add_grade("Программирование", 5)
s2.add_grade("Программирование", 4)

c.add_student(s1)
c.add_student(s2)

print(c.get_average_grade())  # 4.5
````

````python
# Шаг 7: Класс University (базовый)
class University:
    def __init__(self, name):
        self.name = name
        self.courses = []  # список объектов Course
    
    def add_course(self, course):
        """Добавляет курс в университет."""
        if course not in self.courses:
            self.courses.append(course)
            return True
        return False
    
    def find_course(self, course_name):
        """Находит курс по названию."""
        for course in self.courses:
            if course.name.lower() == course_name.lower():
                return course
        return None
    
    def __str__(self):
        return f"Университет '{self.name}', Курсов: {len(self.courses)}"

# Проверка
uni = University("МГУ")
c1 = Course("Математика", "Петров")
c2 = Course("Физика", "Сидоров")

uni.add_course(c1)
uni.add_course(c2)
print(len(uni.courses))  # 2
print(uni.find_course("Математика").teacher)  # Петров
print(uni.find_course("История"))  # None
print(uni)  # Университет 'МГУ', Курсов: 2
````

````python
# Шаг 8: Добавляем метод average_grade_for_course() в University
class University:
    def __init__(self, name):
        self.name = name
        self.courses = []
    
    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)
            return True
        return False
    
    def find_course(self, course_name):
        for course in self.courses:
            if course.name.lower() == course_name.lower():
                return course
        return None
    
    def average_grade_for_course(self, course_name):
        """Возвращает средний балл всех студентов на курсе."""
        course = self.find_course(course_name)
        if course:
            return course.get_average_grade()
        return 0
    
    def __str__(self):
        return f"Университет '{self.name}', Курсов: {len(self.courses)}"

# Проверка
uni = University("МГУ")
c = Course("Математика", "Петров")

s1 = Student("Иван", "S001")
s2 = Student("Мария", "S002")

s1.add_grade("Математика", 5)
s2.add_grade("Математика", 4)

c.add_student(s1)
c.add_student(s2)
uni.add_course(c)

print(uni.average_grade_for_course("Математика"))  # 4.5
print(uni.average_grade_for_course("Физика"))      # 0
````

````python
# Шаг 9: Добавляем метод top_students() в University
class University:
    def __init__(self, name):
        self.name = name
        self.courses = []
    
    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)
            return True
        return False
    
    def find_course(self, course_name):
        for course in self.courses:
            if course.name.lower() == course_name.lower():
                return course
        return None
    
    def average_grade_for_course(self, course_name):
        course = self.find_course(course_name)
        if course:
            return course.get_average_grade()
        return 0
    
    def top_students(self, n=3):
        """Возвращает список из n лучших студентов по среднему баллу."""
        # Собираем всех студентов (без дубликатов)
        all_students = {}
        for course in self.courses:
            for student in course.students:
                all_students[student.student_id] = student
        
        # Сортируем по среднему баллу
        sorted_students = sorted(all_students.values(),
                                key=lambda s: s.get_average_grade(),
                                reverse=True)
        return sorted_students[:n]
    
    def __str__(self):
        return f"Университет '{self.name}', Курсов: {len(self.courses)}"

# Проверка
uni = University("МГУ")
c1 = Course("Математика", "Петров")
c2 = Course("Физика", "Сидоров")

s1 = Student("Иван", "S001")
s2 = Student("Мария", "S002")
s3 = Student("Алексей", "S003")
s4 = Student("Елена", "S004")

s1.add_grade("Математика", 5)  # 5.0
s2.add_grade("Математика", 4)  # 4.0
s3.add_grade("Математика", 3)  # 3.0
s4.add_grade("Математика", 5)  # 5.0

c1.add_student(s1)
c1.add_student(s2)
c1.add_student(s3)
c1.add_student(s4)
uni.add_course(c1)

top = uni.top_students(2)
print(f"Лучшие студенты: {[s.name for s in top]}")  # ['Иван', 'Елена'] или ['Елена', 'Иван']

top3 = uni.top_students(3)
print(len(top3))  # 3
````

````python
class Student:
    """Класс студента (существует независимо от курсов)."""
    
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = {}  # {название_курса: оценка}
    
    def add_grade(self, course_name, grade):
        """Добавляет оценку за курс."""
        if 1 <= grade <= 5:
            self.grades[course_name] = grade
            return True
        return False
    
    def get_average_grade(self):
        """Возвращает средний балл студента по всем курсам."""
        if not self.grades:
            return 0
        return sum(self.grades.values()) / len(self.grades)
    
    def __str__(self):
        return f"{self.name} (ID: {self.student_id}, ср.балл: {self.get_average_grade():.1f})"


class Course:
    """Класс курса (существует независимо от университета)."""
    
    def __init__(self, name, teacher):
        self.name = name
        self.teacher = teacher
        self.students = []  # список объектов Student
    
    def add_student(self, student):
        """Добавляет студента на курс."""
        if student not in self.students:
            self.students.append(student)
            return True
        return False
    
    def remove_student(self, student):
        """Удаляет студента с курса."""
        if student in self.students:
            self.students.remove(student)
            return True
        return False
    
    def get_average_grade(self):
        """Средний балл всех студентов на курсе."""
        if not self.students:
            return 0
        total = 0
        count = 0
        for student in self.students:
            if self.name in student.grades:
                total += student.grades[self.name]
                count += 1
        return total / count if count > 0 else 0
    
    def __str__(self):
        return f"Курс: {self.name}, Преподаватель: {self.teacher}, Студентов: {len(self.students)}"


class University:
    """Класс университета (агрегирует курсы)."""
    
    def __init__(self, name):
        self.name = name
        self.courses = []  # список объектов Course
    
    def add_course(self, course):
        """Добавляет курс в университет."""
        if course not in self.courses:
            self.courses.append(course)
            return True
        return False
    
    def find_course(self, course_name):
        """Находит курс по названию."""
        for course in self.courses:
            if course.name.lower() == course_name.lower():
                return course
        return None
    
    def average_grade_for_course(self, course_name):
        """Возвращает средний балл всех студентов на курсе."""
        course = self.find_course(course_name)
        if course:
            return course.get_average_grade()
        return 0
    
    def top_students(self, n=3):
        """Возвращает список из n лучших студентов по среднему баллу."""
        # Собираем всех студентов со всех курсов (без дубликатов)
        all_students = {}
        for course in self.courses:
            for student in course.students:
                all_students[student.student_id] = student
        
        # Сортируем по среднему баллу
        sorted_students = sorted(all_students.values(), 
                                key=lambda s: s.get_average_grade(), 
                                reverse=True)
        return sorted_students[:n]
    
    def __str__(self):
        return f"Университет '{self.name}', Курсов: {len(self.courses)}"


# ========== ПРОВЕРКА ==========

# Создаём студентов (существуют отдельно)
s1 = Student("Иван Петров", "S001")
s2 = Student("Мария Сидорова", "S002")
s3 = Student("Алексей Иванов", "S003")
s4 = Student("Елена Смирнова", "S004")

# Добавляем оценки студентам
s1.add_grade("Математика", 5)
s1.add_grade("Физика", 4)
s1.add_grade("Программирование", 5)

s2.add_grade("Математика", 4)
s2.add_grade("Физика", 5)
s2.add_grade("Программирование", 4)

s3.add_grade("Математика", 3)
s3.add_grade("Физика", 4)
s3.add_grade("Программирование", 3)

s4.add_grade("Математика", 5)
s4.add_grade("Физика", 5)
s4.add_grade("Программирование", 5)

# Создаём курсы (существуют отдельно)
c1 = Course("Математика", "Проф. Петров")
c2 = Course("Физика", "Доц. Сидоров")
c3 = Course("Программирование", "Ст. преп. Иванова")

# Записываем студентов на курсы
c1.add_student(s1)
c1.add_student(s2)
c1.add_student(s3)
c1.add_student(s4)

c2.add_student(s1)
c2.add_student(s2)
c2.add_student(s3)
c2.add_student(s4)

c3.add_student(s1)
c3.add_student(s2)
c3.add_student(s3)
c3.add_student(s4)

# Создаём университет и добавляем курсы
university = University("МГУ")
university.add_course(c1)
university.add_course(c2)
university.add_course(c3)

# Проверяем
print("=" * 50)
print(f"УНИВЕРСИТЕТ: {university}")
print("=" * 50)

print("\n=== СТУДЕНТЫ ===")
for student in [s1, s2, s3, s4]:
    print(f"  {student}")

print("\n=== КУРСЫ ===")
for course in university.courses:
    print(f"  {course}")

print("\n=== СРЕДНИЙ БАЛЛ ПО КУРСАМ ===")
for course in university.courses:
    avg = university.average_grade_for_course(course.name)
    print(f"  {course.name}: {avg:.1f}")

print("\n=== ЛУЧШИЕ СТУДЕНТЫ ===")
top = university.top_students(3)
for i, student in enumerate(top, 1):
    print(f"  {i}. {student}")
````

### 9. Класс MathOperations (только статические методы)

Цель: @staticmethod, когда состояние объекта не нужно.

# Создайте класс MathOperations со статическими методами:
#   factorial(n) — рекурсивно или итеративно
#   is_prime(n) — проверка простого числа
#   gcd(a, b) — наибольший общий делитель (алгоритм Евклида)
# Все методы вызываются без создания экземпляра: MathOperations.factorial(5)

**🔍 Ключевые концепции**

|Концепция	|Код	|Зачем нужна
|-----------|-------|-----------
|`@staticmethod`	|`@staticmethod def factorial(n):`	|Метод не требует self (нет доступа к атрибутам объекта)
|Вызов без экземпляра	|`MathOperations.factorial(5)`	|Не нужно создавать объект
|Алгоритм Евклида	|`while b != 0: a, b = b, a % b`	|Эффективный способ найти НОД
|Проверка простоты	|`int(n ** 0.5) + 1`	|Оптимизация: достаточно проверить до корня

````python
# Шаг 1: Создайте класс со статическим методом factorial
class MathOperations:
    @staticmethod
    def factorial(n):
        """Вычисляет факториал числа n."""
        if n < 0:
            raise ValueError("Факториал не определён для отрицательных чисел")
        if n == 0 or n == 1:
            return 1
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Проверка
print(MathOperations.factorial(5))  # 120
print(MathOperations.factorial(0))  # 1
````

````python
class MathOperations:
    """Класс с математическими операциями (только статические методы)."""
    
    @staticmethod
    def factorial(n):
        """
        Вычисляет факториал числа n.
        
        Args:
            n: Неотрицательное целое число
        
        Returns:
            n! (факториал)
        """
        if n < 0:
            raise ValueError("Факториал определён только для неотрицательных чисел")
        if n == 0 or n == 1:
            return 1
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
    
    @staticmethod
    def is_prime(n):
        """
        Проверяет, является ли число простым.
        
        Args:
            n: Целое число
        
        Returns:
            True если число простое, иначе False
        """
        if n < 2:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True
    
    @staticmethod
    def gcd(a, b):
        """
        Находит наибольший общий делитель (НОД) чисел a и b.
        
        Args:
            a: Первое число
            b: Второе число
        
        Returns:
            НОД(a, b) (алгоритм Евклида)
        """
        a, b = abs(a), abs(b)
        while b != 0:
            a, b = b, a % b
        return a


# ========== ПРОВЕРКА ==========

if __name__ == "__main__":
    print("=" * 50)
    print("MathOperations (статический класс)")
    print("=" * 50)
    
    # Вызываем методы без создания экземпляра
    print("\n=== ФАКТОРИАЛ ===")
    print(f"MathOperations.factorial(5) = {MathOperations.factorial(5)}")   # 120
    print(f"MathOperations.factorial(0) = {MathOperations.factorial(0)}")   # 1
    print(f"MathOperations.factorial(7) = {MathOperations.factorial(7)}")   # 5040
    
    print("\n=== ПРОВЕРКА ПРОСТЫХ ЧИСЕЛ ===")
    print(f"MathOperations.is_prime(2) = {MathOperations.is_prime(2)}")     # True
    print(f"MathOperations.is_prime(7) = {MathOperations.is_prime(7)}")     # True
    print(f"MathOperations.is_prime(10) = {MathOperations.is_prime(10)}")   # False
    print(f"MathOperations.is_prime(1) = {MathOperations.is_prime(1)}")     # False
    
    print("\n=== НАИБОЛЬШИЙ ОБЩИЙ ДЕЛИТЕЛЬ ===")
    print(f"MathOperations.gcd(12, 18) = {MathOperations.gcd(12, 18)}")     # 6
    print(f"MathOperations.gcd(17, 19) = {MathOperations.gcd(17, 19)}")     # 1
    print(f"MathOperations.gcd(0, 5) = {MathOperations.gcd(0, 5)}")         # 5
    print(f"MathOperations.gcd(-12, 18) = {MathOperations.gcd(-12, 18)}")   # 6
````

````python
import random
import time

class Hero:
    """Базовый класс для всех героев."""
    
    def __init__(self, name, health, damage):
        self.name = name
        self._health = health  # инкапсуляция
        self.damage = damage
        self.is_alive = True
    
    @property
    def health(self):
        return self._health
    
    @health.setter
    def health(self, value):
        self._health = max(0, value)
        if self._health <= 0:
            self.is_alive = False
    
    def attack(self, enemy):
        """Обычная атака."""
        enemy.health -= self.damage
        return f"{self.name} атакует {enemy.name} и наносит {self.damage} урона!"
    
    def special_attack(self, enemy):
        """Специальная атака (переопределяется в дочерних классах)."""
        pass
    
    def __str__(self):
        return f"{self.name} (❤️ {self.health})"


class Warrior(Hero):
    """Воин: блок щитом (защита + контратака)."""
    
    def __init__(self, name):
        super().__init__(name, health=150, damage=20)
        self.block_chance = 0.3
    
    def special_attack(self, enemy):
        """Блок щитом: снижает урон от следующей атаки и контратакует."""
        print(f"   🛡️ {self.name} поднимает щит!")
        # Блок будет работать в методе take_damage
        return f"   {self.name} готовится к контратаке!"
    
    def take_damage(self, damage):
        """Обработка урона с учётом блока."""
        if random.random() < self.block_chance:
            reduced_damage = damage // 2
            self.health -= reduced_damage
            return f"   🛡️ Блок! Урон снижен до {reduced_damage}!"
        else:
            self.health -= damage
            return f"   {self.name} получает {damage} урона!"


class Mage(Hero):
    """Маг: огненный шар (высокий урон с шансом промаха)."""
    
    def __init__(self, name):
        super().__init__(name, health=100, damage=15)
        self.fireball_chance = 0.7
        self.fireball_damage = 40
    
    def special_attack(self, enemy):
        """Огненный шар: высокий урон, но может промахнуться."""
        if random.random() < self.fireball_chance:
            enemy.health -= self.fireball_damage
            return f"   🔥 {self.name} запускает огненный шар! Наносит {self.fireball_damage} урона!"
        else:
            return f"   💨 {self.name} промахнулся огненным шаром!"
    
    def attack(self, enemy):
        """Переопределяем атаку мага (более слабая)."""
        enemy.health -= self.damage
        return f"{self.name} атакует посохом и наносит {self.damage} урона!"


class Archer(Hero):
    """Лучник: критический удар (удвоенный урон с шансом)."""
    
    def __init__(self, name):
        super().__init__(name, health=120, damage=18)
        self.critical_chance = 0.25
        self.critical_multiplier = 2.5
    
    def special_attack(self, enemy):
        """Критический удар: двойной урон."""
        # Критический удар применяется к обычной атаке
        return None  # Крит сработает при атаке
    
    def attack(self, enemy):
        """Атака с шансом критического удара."""
        if random.random() < self.critical_chance:
            crit_damage = int(self.damage * self.critical_multiplier)
            enemy.health -= crit_damage
            return f"🎯 {self.name} наносит КРИТИЧЕСКИЙ удар! {crit_damage} урона!"
        else:
            enemy.health -= self.damage
            return f"{self.name} атакует и наносит {self.damage} урона!"


# ========== ИГРА ==========

def battle(hero1, hero2):
    """Проводит битву между двумя героями."""
    print("\n" + "=" * 60)
    print(f"⚔️ БИТВА НАЧИНАЕТСЯ: {hero1.name} vs {hero2.name} ⚔️")
    print("=" * 60)
    
    # Определяем порядок ходов (рандом)
    if random.random() < 0.5:
        first, second = hero1, hero2
    else:
        first, second = hero2, hero1
    
    print(f"\n{first.name} атакует первым!\n")
    
    turn = 0
    
    while hero1.is_alive and hero2.is_alive:
        current = first if turn % 2 == 0 else second
        enemy = second if turn % 2 == 0 else first
        
        print(f"\n--- ХОД {turn + 1} ---")
        print(f"📊 {hero1}  |  {hero2}")
        
        # Обычная атака
        print(f"⚔️ {current.attack(enemy)}")
        
        # Специальная атака
        if isinstance(current, Warrior):
            print(f"   {current.special_attack(enemy)}")
        elif isinstance(current, Mage):
            result = current.special_attack(enemy)
            if result:
                print(result)
        elif isinstance(current, Archer):
            pass  # У лучника крит уже в атаке
        
        # Проверка на смерть
        if enemy.health <= 0:
            print(f"\n💀 {enemy.name} повержен!")
            break
        
        time.sleep(1)  # Пауза для драматизма
        turn += 1
    
    # Определяем победителя
    if hero1.is_alive and not hero2.is_alive:
        winner = hero1
    elif hero2.is_alive and not hero1.is_alive:
        winner = hero2
    else:
        winner = None
    
    print("\n" + "=" * 60)
    if winner:
        print(f"🏆 ПОБЕДИТЕЛЬ: {winner.name} с {winner.health} здоровья! 🏆")
    else:
        print("🤝 НИЧЬЯ! 🤝")
    print("=" * 60)
    
    return winner


# ========== ПРОВЕРКА ==========

if __name__ == "__main__":
    print("=" * 50)
    print("ИГРА 'БИТВА ГЕРОЕВ'")
    print("=" * 50)
    
    # Создаём героев
    warrior = Warrior("Александр")
    mage = Mage("Мерлин")
    archer = Archer("Робин Гуд")
    
    print("\n=== ГЕРОИ ===")
    print(f"🏃 {warrior}")
    print(f"🧙 {mage}")
    print(f"🏹 {archer}")
    
    # Выбор героев
    print("\n=== ВЫБЕРИТЕ ГЕРОЕВ ДЛЯ БИТВЫ ===")
    print("1. Воин (высокое здоровье, блок щитом)")
    print("2. Маг (огненный шар, высокий урон)")
    print("3. Лучник (критические удары)")
    
    try:
        choice1 = int(input("\nПервый герой (1-3): "))
        choice2 = int(input("Второй герой (1-3): "))
        
        heroes = {1: warrior, 2: mage, 3: archer}
        hero_a = heroes[choice1]
        hero_b = heroes[choice2]
        
        # Битва
        battle(hero_a, hero_b)
        
    except ValueError:
        print("Неверный ввод! Проводим демо-битву...")
        print("\n=== ДЕМО-БИТВА: ВОИН vs МАГ ===\n")
        battle(warrior, mage)
````

````python
import json
from datetime import datetime, date
from typing import List, Optional

class Task:
    """Класс задачи."""
    
    def __init__(self, description: str, deadline: str):
        """
        Создаёт задачу.
        
        Args:
            description: Описание задачи
            deadline: Дедлайн в формате "ГГГГ-ММ-ДД"
        """
        self.description = description
        self.deadline = datetime.strptime(deadline, "%Y-%m-%d").date()
        self.completed = False
    
    def mark_done(self):
        """Отмечает задачу как выполненную."""
        self.completed = True
    
    def is_overdue(self) -> bool:
        """Проверяет, просрочена ли задача."""
        return not self.completed and self.deadline < date.today()
    
    def to_dict(self) -> dict:
        """Преобразует задачу в словарь для JSON."""
        return {
            "description": self.description,
            "deadline": self.deadline.strftime("%Y-%m-%d"),
            "completed": self.completed
        }
    
    @classmethod
    def from_dict(cls, data: dict) -> 'Task':
        """Создаёт задачу из словаря."""
        task = cls(data["description"], data["deadline"])
        task.completed = data["completed"]
        return task
    
    def __str__(self) -> str:
        status = "✅" if self.completed else "❌" if self.is_overdue() else "⏳"
        return f"{status} {self.description} (до: {self.deadline})"


class Category:
    """Класс категории задач."""
    
    def __init__(self, name: str):
        self.name = name
        self.tasks: List[Task] = []
    
    def add_task(self, task: Task):
        """Добавляет задачу в категорию."""
        self.tasks.append(task)
    
    def get_tasks(self, include_completed: bool = True) -> List[Task]:
        """Возвращает задачи категории."""
        if include_completed:
            return self.tasks
        return [t for t in self.tasks if not t.completed]
    
    def find_task(self, description: str) -> Optional[Task]:
        """Находит задачу по описанию."""
        for task in self.tasks:
            if task.description.lower() == description.lower():
                return task
        return None
    
    def to_dict(self) -> dict:
        """Преобразует категорию в словарь."""
        return {
            "name": self.name,
            "tasks": [task.to_dict() for task in self.tasks]
        }
    
    @classmethod
    def from_dict(cls, data: dict) -> 'Category':
        """Создаёт категорию из словаря."""
        category = cls(data["name"])
        for task_data in data["tasks"]:
            category.add_task(Task.from_dict(task_data))
        return category
    
    def __str__(self) -> str:
        total = len(self.tasks)
        completed = len([t for t in self.tasks if t.completed])
        overdue = len([t for t in self.tasks if t.is_overdue()])
        return f"📁 {self.name}: {total} задач (✅ {completed}, ❌ {overdue})"


class TaskManager:
    """Класс менеджера задач."""
    
    def __init__(self):
        self.categories: List[Category] = []
    
    def add_category(self, category: Category):
        """Добавляет категорию."""
        self.categories.append(category)
    
    def get_category(self, name: str) -> Optional[Category]:
        """Находит категорию по имени."""
        for category in self.categories:
            if category.name.lower() == name.lower():
                return category
        return None
    
    def add_task_to_category(self, category_name: str, task: Task):
        """Добавляет задачу в категорию."""
        category = self.get_category(category_name)
        if category:
            category.add_task(task)
        else:
            raise ValueError(f"Категория '{category_name}' не найдена")
    
    def get_overdue_tasks(self) -> List[Task]:
        """Возвращает все просроченные задачи."""
        overdue = []
        for category in self.categories:
            for task in category.tasks:
                if task.is_overdue():
                    overdue.append(task)
        return overdue
    
    def mark_done(self, task_description: str) -> bool:
        """
        Отмечает задачу как выполненную.
        
        Returns:
            True если задача найдена и отмечена, иначе False
        """
        for category in self.categories:
            task = category.find_task(task_description)
            if task:
                task.mark_done()
                return True
        return False
    
    def save_to_file(self, filename: str):
        """Сохраняет все данные в JSON файл."""
        data = {
            "categories": [category.to_dict() for category in self.categories]
        }
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=2)
        print(f"💾 Данные сохранены в {filename}")
    
    def load_from_file(self, filename: str):
        """Загружает данные из JSON файла."""
        try:
            with open(filename, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            self.categories = []
            for cat_data in data["categories"]:
                self.categories.append(Category.from_dict(cat_data))
            
            print(f"📂 Данные загружены из {filename}")
        except FileNotFoundError:
            print(f"❌ Файл {filename} не найден")
        except json.JSONDecodeError:
            print(f"❌ Ошибка чтения файла {filename}")
    
    def show_all(self):
        """Показывает все категории и задачи."""
        print("\n" + "=" * 50)
        print("📋 ВСЕ ЗАДАЧИ")
        print("=" * 50)
        
        if not self.categories:
            print("   Нет категорий")
        else:
            for category in self.categories:
                print(f"\n{category}")
                if category.tasks:
                    for task in category.tasks:
                        print(f"   {task}")
                else:
                    print(f"   (нет задач)")
        
        overdue = self.get_overdue_tasks()
        if overdue:
            print(f"\n⚠️ ПРОСРОЧЕННЫЕ ЗАДАЧИ ({len(overdue)}):")
            for task in overdue:
                print(f"   ⚠️ {task.description} (до: {task.deadline})")
    
    def __str__(self) -> str:
        total_categories = len(self.categories)
        total_tasks = sum(len(c.tasks) for c in self.categories)
        return f"TaskManager: {total_categories} категорий, {total_tasks} задач"


# ========== ПРИМЕР ИСПОЛЬЗОВАНИЯ ==========

if __name__ == "__main__":
    # Создаём менеджер
    tm = TaskManager()
    
    # Создаём категории
    work = Category("Работа")
    personal = Category("Личное")
    study = Category("Учёба")
    
    tm.add_category(work)
    tm.add_category(personal)
    tm.add_category(study)
    
    # Создаём задачи
    task1 = Task("Закончить отчёт", "2024-12-20")
    task2 = Task("Позвонить клиенту", "2024-12-15")
    task3 = Task("Купить продукты", "2024-12-18")
    task4 = Task("Выучить Python", "2024-12-25")
    
    # Добавляем задачи в категории
    tm.add_task_to_category("Работа", task1)
    tm.add_task_to_category("Работа", task2)
    tm.add_task_to_category("Личное", task3)
    tm.add_task_to_category("Учёба", task4)
    
    # Помечаем задачу как выполненную
    tm.mark_done("Позвонить клиенту")
    
    # Показываем все задачи
    tm.show_all()
    
    # Просроченные задачи
    print("\n" + "=" * 50)
    print("⚠️ ПРОСРОЧЕННЫЕ ЗАДАЧИ")
    print("=" * 50)
    overdue = tm.get_overdue_tasks()
    for task in overdue:
        print(f"   {task}")
    
    # Сохраняем в файл
    tm.save_to_file("tasks.json")
    
    # Создаём новый менеджер и загружаем
    tm2 = TaskManager()
    tm2.load_from_file("tasks.json")
    tm2.show_all()
````
