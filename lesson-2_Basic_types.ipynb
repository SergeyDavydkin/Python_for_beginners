# Задача 5.
По номеру последней квартиры на первом этаже и номеру квартиры бабушки - числам N, K,а так же по количеству этажей в доме - С, необходимо определить номер подъезда и номер этажа, где находится квартира бабушки. Первый подъезд описывается числом 1(не 0!)

```python
# Ввод данных
# N, C, K = map(int, input().split())

N = int(input("Количество квартир на этаже (последняя квартира на 1 этаже 1 подъезда): "))
C = int(input("Количество этажей в доме: "))
K = int(input("Номер квартиры бабушки: "))

# решение 1
print('Решение 1')
print('=='*20)

# Квартир в одном подъезде
apartments = N * C

# Находим подъезд
entrance = (K - 1) // apartments + 1

# Находим номер квартиры внутри подъезда
apartment_in_entrance = (K - 1) % apartments + 1

# Находим этаж внутри подъезда
floor = (apartment_in_entrance + N - 1) // N

# Вывод
print(f"Подъезд: {entrance}")
print(f"Этаж: {floor}")
print()

# решение 2
print('Решение 2')
print('=='*20)
entrance = (K - 1) // (N * C) + 1
floor = ((K - 1) % (N * C)) // N + 1
# Вывод
print(f"Подъезд: {entrance}")
print(f"Этаж: {floor}")
print()

# решение 3
print('Решение 3')
print('=='*20)
# Более явная запись с использованием округления вверх
floor_total = (K + N - 1) // N  # вместо K // N + bool(K % N)
entrance = (floor_total + C - 1) // C  # вместо floor // C + bool(floor % C)
floor = floor_total - C * (entrance - 1)
# Вывод
print(f"Подъезд: {entrance}")
print(f"Этаж: {floor}")
print()


# решение 4
print('Решение 4')
print('=='*20)
# Эквивалент ceil(k/n) без float
floor = (K + N - 1) // n

# Эквивалент ceil(floor/c) без float
gate = (floor + C - 1) // C

result_floor = floor - C * (gate - 1)
# Вывод
print(f"Подъезд: {entrance}")
print(f"Этаж: {result_floor}")
print()


# решение 5
print('Решение 5')
print('=='*20)
# Лямбда-функция для перевода номера в позицию
find_position = lambda k, n, c: (
    (k - 1) // (n * c) + 1,          # подъезд
    ((k - 1) % (n * c)) // n + 1     # этаж
)

entrance, floor = find_position(K, N, C)
# Вывод
print(f"Подъезд: {entrance}")
print(f"Этаж: {floor}")
print()

# 3, 5, 14      -> 1 5
# 3, 5, 23      -> 2 3
# 8, 10, 240    -> 3 10
# 3, 4, 14      -> 2 1
# 3, 5, 58      -> 4 5
```

**1. Простой подход с циклом**

```python
def find_posit(n, c, k):
    entrance = (k - 1) // (n * c) + 1
    floor = ((k - 1) % (n * c)) // n + 1
    return entrance, floor

# Тестовые данные: (N, C, K, expected_entrance, expected_floor)
test_cases = [
    (3, 5, 14, 1, 5),
    (3, 5, 23, 2, 3),
    (8, 10, 240, 3, 10),
    (3, 2, 14, 3, 1),
    (3, 5, 58, 4, 5)
]

print("Тестирование функции find_posit:")
print("=" * 40)

all_passed = True
for i, (n, c, k, exp_entr, exp_floor) in enumerate(test_cases, 1):
    actual_entr, actual_floor = find_posit(n, c, k)
    passed = (actual_entr == exp_entr) and (actual_floor == exp_floor)
    
    print(f"Тест {i}: N={n}, C={c}, K={k}")
    print(f"  Ожидалось: подъезд {exp_entr}, этаж {exp_floor}")
    print(f"  Получено:  подъезд {actual_entr}, этаж {actual_floor}")
    print(f"  Статус: {'✓ ПРОЙДЕН' if passed else '✗ ОШИБКА'}")
    print()
    
    if not passed:
        all_passed = False

print("=" * 40)
print(f"Итог: {'Все тесты пройдены!' if all_passed else 'Есть ошибки!'}")
```

**3. С параметризованными тестами (pytest стиль)**

```python
def find_posit(n, c, k):
    entrance = (k - 1) // (n * c) + 1
    floor = ((k - 1) % (n * c)) // n + 1
    return entrance, floor

# Если установлен pytest, можно использовать декоратор:
# @pytest.mark.parametrize("n,c,k,expected", [...])

def run_parametrized_tests():
    test_data = [
        (3, 5, 14, (1, 5)),
        (3, 5, 23, (2, 3)),
        (8, 10, 240, (3, 10)),
        (3, 2, 14, (3, 1)),
        (3, 5, 58, (4, 5))
    ]
    
    results = []
    for n, c, k, expected in test_data:
        result = find_posit(n, c, k)
        status = "PASS" if result == expected else "FAIL"
        results.append({
            "input": (n, c, k),
            "expected": expected,
            "actual": result,
            "status": status
        })
    
    # Красивый вывод
    print("\nПараметризованные тесты:")
    print("-" * 60)
    for i, r in enumerate(results, 1):
        print(f"Тест {i}: N={r['input'][0]}, C={r['input'][1]}, K={r['input'][2]}")
        print(f"  Ожидалось: {r['expected']}")
        print(f"  Получено:  {r['actual']}")
        print(f"  Статус:    {r['status']}\n")
    
    passed = sum(1 for r in results if r['status'] == "PASS")
    print(f"Итого: {passed}/{len(results)} тестов пройдено")

run_parametrized_tests()
```

**4. С генерацией случайных тестов**

```python
import random

def find_posit(n, c, k):
    entrance = (k - 1) // (n * c) + 1
    floor = ((k - 1) % (n * c)) // n + 1
    return entrance, floor

def alternative_solution(n, c, k):
    """Альтернативная реализация для проверки"""
    total_floors = (k + n - 1) // n
    entrance = (total_floors + c - 1) // c
    floor = total_floors - c * (entrance - 1)
    return entrance, floor

def run_random_tests(num_tests=1000):
    print(f"Запуск {num_tests} случайных тестов...")
    print("=" * 50)
    
    failed_tests = []
    
    for i in range(num_tests):
        # Генерируем случайные параметры
        n = random.randint(1, 20)  # квартир на этаже
        c = random.randint(1, 20)  # этажей в подъезде
        max_k = n * c * random.randint(1, 10)  # максимальный номер квартиры
        k = random.randint(1, max_k)
        
        # Вычисляем двумя способами
        result1 = find_posit(n, c, k)
        result2 = alternative_solution(n, c, k)
        
        if result1 != result2:
            failed_tests.append({
                'n': n, 'c': c, 'k': k,
                'result1': result1,
                'result2': result2
            })
    
    # Вывод результатов
    if not failed_tests:
        print("✓ Все тесты пройдены! Результаты совпадают.")
    else:
        print(f"✗ Найдены расхождения в {len(failed_tests)} тестах:")
        for i, test in enumerate(failed_tests[:5], 1):  # покажем первые 5
            print(f"\nРасхождение {i}:")
            print(f"  N={test['n']}, C={test['c']}, K={test['k']}")
            print(f"  Ваша функция: {test['result1']}")
            print(f"  Альтернативная: {test['result2']}")
        
        if len(failed_tests) > 5:
            print(f"\n... и еще {len(failed_tests) - 5} расхождений")

run_random_tests(500)
```

# Задача 7
Является ли введенное число A степенью 2 или нет. Если является выведите 0, вином случае выведите любое число. Циклами и условиями пользоваться нельзя!

```python
n = int(input('Введите челое число: '))
result = n & (n - 1)

print(result)
```

```python
# Учет случая A=0:
A = int(input())
# Добавляем проверку на 0 через булеву логику
result = (A & (A-1)) | (A == 0)
print(result)
```

**Объяснение финального решения:**

1. `A & (A-1)` — основная проверка (0 для степеней двойки при A>0)

2. `not A` — обработка случая A=0 (возвращает 1 для 0, 0 для других)

3. `|` (ИЛИ) — объединяет результаты:

    - Для A=0: 0 | 1 = 1 (не 0)

    - Для степени 2 при A>0: 0 | 0 = 0

    - Для не степени 2: не-ноль | 0 = не-ноль

```python
# Основа: свойство степеней двойки
# Реализация: битовые операции
# Уточнение: обработка A=0 через логическое NOT
# Расширение: компактная запись

# Улучшенная версия (без явного сравнения):
A = int(input())
# Используем свойство: для A=0, выражение -A & A дает 0
result = (A & (A-1)) | (not A)
print(result)
```

```python
# Альтернативный подход через математику:
A = int(input())
# Если A степень двойки, то log2(A) целое число
import math
# Но math.log может давать погрешности для больших чисел
log_val = math.log2(A) if A > 0 else -1
result = 0 if A > 0 and log_val.is_integer() else 1
print(result)
```

```python
def power_of_two(n):
    return (n & (n - 1)) | (not n)

print(power_of_two(0))  # -> нет
print(power_of_two(1))  # -> да (0)
print(power_of_two(2))  # -> да (0)
print(power_of_two(6))  # -> нет
print(power_of_two(32))  # -> да (0)
print(power_of_two(1024))  # -> да (0)
```
