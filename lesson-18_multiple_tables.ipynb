## Этап 3: Работа с несколькими таблицами

```python
# Подготовка данных
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Настройка отображения
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_rows', 20)

print("="*70)
print("ЭТАП 3: РАБОТА С НЕСКОЛЬКИМИ ТАБЛИЦАМИ")
print("="*70)

# Создадим несколько связанных таблиц для демонстрации
np.random.seed(42)

# Таблица 1: Клиенты
customers = pd.DataFrame({
    'customer_id': range(1, 11),
    'name': [f'Клиент_{i}' for i in range(1, 11)],
    'city': np.random.choice(['Москва', 'СПб', 'Казань', 'Новосибирск', 'Екатеринбург'], 10),
    'age': np.random.randint(25, 60, 10),
    'total_purchases': np.random.randint(1, 20, 10)
})

# Таблица 2: Заказы
orders = pd.DataFrame({
    'order_id': range(101, 121),
    'customer_id': np.random.choice(range(1, 11), 20),
    'order_date': pd.date_range('2024-01-01', periods=20, freq='D'),
    'amount': np.random.uniform(100, 5000, 20).round(2),
    'status': np.random.choice(['completed', 'pending', 'cancelled'], 20, p=[0.7, 0.2, 0.1])
})

# Таблица 3: Товары
products = pd.DataFrame({
    'product_id': range(1, 9),
    'product_name': ['Ноутбук', 'Телефон', 'Планшет', 'Наушники', 'Мышь', 'Клавиатура', 'Монитор', 'Принтер'],
    'category': ['Электроника', 'Электроника', 'Электроника', 'Аксессуары', 'Аксессуары', 'Аксессуары', 'Электроника', 'Офис'],
    'price': np.random.uniform(500, 50000, 8).round(0)
})

# Таблица 4: Детали заказов (связь многие ко многим)
order_details = pd.DataFrame({
    'order_id': np.random.choice(range(101, 121), 35),
    'product_id': np.random.choice(range(1, 9), 35),
    'quantity': np.random.randint(1, 5, 35),
    'discount': np.random.uniform(0, 0.3, 35).round(2)
})

# Таблица 5: Бонусы клиентов (дополнительная таблица)
customer_bonus = pd.DataFrame({
    'customer_id': range(1, 11),
    'bonus_points': np.random.randint(0, 5000, 10),
    'status': np.random.choice(['gold', 'silver', 'bronze'], 10, p=[0.2, 0.3, 0.5])
})

# Таблица 6: Новые клиенты (для демонстрации append/concat)
new_customers = pd.DataFrame({
    'customer_id': range(11, 16),
    'name': [f'Новый_клиент_{i}' for i in range(11, 16)],
    'city': ['Москва', 'СПб', 'Казань', 'Новосибирск', 'Сочи'],
    'age': np.random.randint(20, 55, 5),
    'total_purchases': [0, 0, 0, 0, 0]
})

# Таблица 7: Заказы 2023 года (для демонстрации concat)
orders_2023 = pd.DataFrame({
    'order_id': range(1, 11),
    'customer_id': np.random.choice(range(1, 11), 10),
    'order_date': pd.date_range('2023-01-01', periods=10, freq='ME'),
    'amount': np.random.uniform(100, 4500, 10).round(2),
    'status': np.random.choice(['completed', 'pending', 'cancelled'], 10, p=[0.7, 0.2, 0.1])
})

print("\nИсходные таблицы созданы:")
print(f"Клиенты: {customers.shape}")
print(f"Заказы: {orders.shape}")
print(f"Товары: {products.shape}")
print(f"Детали заказов: {order_details.shape}")
print(f"Бонусы клиентов: {customer_bonus.shape}")
print(f"Новые клиенты: {new_customers.shape}")
print(f"Заказы 2023: {orders_2023.shape}")
```

### 1. МЕТОД APPEND()

```python
print("\n" + "="*70)
print("1. МЕТОД APPEND()")
print("="*70)

# ========== 1. Метод append() ==========
print("\n--- 1.1 Базовое использование append() ---")
print("\n✅ ВАЖНО: append() удален в pandas 2.0+, используйте pd.concat()")

# ========== 1. Добавление строк (вместо append) ==========
print("\n--- 1.1 Добавление одной строки с помощью pd.concat() ---")

# Вместо df.append() используем pd.concat()
new_row = pd.DataFrame({
    'customer_id': [16],
    'name': ['Иванов Иван'],
    'city': ['Москва'],
    'age': [35],
    'total_purchases': [0]
})

# Добавляем одну строку
customers_appended = pd.concat([customers, new_row], ignore_index=True)
print("Добавление одной строки:")
print(f"Исходный размер: {len(customers)}")
print(f"Размер после добавления: {len(customers_appended)}")
print("Последние 3 строки:")
print(customers_appended.tail(3))

# Добавление нескольких строк
new_rows = pd.DataFrame({
    'customer_id': [17, 18],
    'name': ['Петров Петр', 'Сидорова Анна'],
    'city': ['СПб', 'Казань'],
    'age': [28, 42],
    'total_purchases': [0, 0]
})

customers_appended = pd.concat([customers, new_rows], ignore_index=True)
print(f"\nДобавление нескольких строк: {len(customers_appended)} строк")

# ========== 1.2 Добавление строк с разными столбцами ==========
print("\n--- 1.2 Добавление строк с разными столбцами ---")

incomplete_row = pd.DataFrame({
    'customer_id': [19],
    'name': ['Неполный'],
    'city': ['Москва']
    # Нет age и total_purchases
})

customers_with_nan = pd.concat([customers, incomplete_row], ignore_index=True)
print("Добавление строки с недостающими столбцами:")
print(customers_with_nan.tail(2))
print("NaN заполняются автоматически")
```

### 2 Функция pd.concat()

```python
# ========== 2. Функция pd.concat() ==========
print("\n--- 2.1 Вертикальное объединение (axis=0) ---")

# Объединение по строкам
all_customers = pd.concat([customers, new_customers], ignore_index=True)
print("Объединение клиентов + новые клиенты:")
print(f"Размер: {all_customers.shape}")
print(all_customers.tail(5))

# Объединение нескольких таблиц с метками
all_orders = pd.concat([orders_2023, orders], ignore_index=True, keys=['2023', '2024'])
print("\nОбъединение заказов 2023 и 2024:")
print(f"Размер: {all_orders.shape}")
print("Первые 3 и последние 3 строки:")
print(pd.concat([all_orders.head(3), all_orders.tail(3)]))

# ========== 2.2 Горизонтальное объединение (axis=1) ==========
print("\n--- 2.2 Горизонтальное объединение (axis=1) ---")

# Создаем таблицы с общим индексом
customer_basic = customers[['customer_id', 'name', 'city']].copy()
customer_stats = customers[['customer_id', 'total_purchases']].copy()

# Устанавливаем индексы для горизонтального объединения
customer_basic.set_index('customer_id', inplace=True)
customer_stats.set_index('customer_id', inplace=True)

# Горизонтальное объединение
customer_full = pd.concat([customer_basic, customer_stats], axis=1)
print("Горизонтальное объединение с индексами:")
print(customer_full.head())

# ========== 2.3 Разные типы соединения при concat ==========
print("\n--- 2.3 Типы соединения при concat ---")

# Создаем таблицы с разными индексами
df1 = pd.DataFrame({'A': [1, 2, 3]}, index=[0, 1, 2])
df2 = pd.DataFrame({'B': [4, 5, 6]}, index=[2, 3, 4])

print("Таблица 1:")
print(df1)
print("\nТаблица 2:")
print(df2)

print("\nInner join (только пересекающиеся индексы):")
print(pd.concat([df1, df2], axis=1, join='inner'))

print("\nOuter join (все индексы):")
print(pd.concat([df1, df2], axis=1, join='outer'))
```

### 3. Функции pd.merge() и df.join()

```python
# ========== 3. Функции pd.merge() и df.join() ==========
print("\n--- 3.1 pd.merge() - основные типы соединений ---")

# Восстанавливаем индексы для демонстрации
customers_reset = customers.copy()
orders_reset = orders.copy()

# Inner join (по умолчанию)
inner_join = pd.merge(customers_reset, orders_reset, on='customer_id', how='inner')
print("INNER JOIN (только совпадающие записи):")
print(f"Клиентов: {len(customers_reset)}, Заказов: {len(orders_reset)}")
print(f"Результат inner join: {len(inner_join)} строк")
print(inner_join.head())

# Left join
left_join = pd.merge(customers_reset, orders_reset, on='customer_id', how='left')
print("\nLEFT JOIN (все клиенты + их заказы):")
print(f"Результат left join: {len(left_join)} строк")
print("Клиенты без заказов:")
print(left_join[left_join['order_id'].isna()][['customer_id', 'name']].drop_duplicates())

# Right join
right_join = pd.merge(customers_reset, orders_reset, on='customer_id', how='right')
print("\nRIGHT JOIN (все заказы + информация о клиентах):")
print(f"Результат right join: {len(right_join)} строк")

# Outer join
outer_join = pd.merge(customers_reset, orders_reset, on='customer_id', how='outer')
print("\nOUTER JOIN (все клиенты и все заказы):")
print(f"Результат outer join: {len(outer_join)} строк")

# ========== 3.2 Merge с разными именами столбцов ==========
print("\n--- 3.2 Merge с разными именами столбцов ---")

# Переименовываем столбец для демонстрации
customers_renamed = customers_reset.rename(columns={'customer_id': 'cust_id'})
print("Merge с разными именами столбцов:")
merged_diff = pd.merge(customers_renamed, orders_reset, left_on='cust_id', right_on='customer_id', how='inner')
print(merged_diff[['cust_id', 'name', 'order_id', 'amount']].head())

# ========== 3.3 Merge по нескольким столбцам ==========
print("\n--- 3.3 Merge по нескольким столбцам ---")

# Создаем дополнительную таблицу с составным ключом
customer_segments = pd.DataFrame({
    'city': ['Москва', 'СПб', 'Казань', 'Новосибирск', 'Екатеринбург'],
    'age_group': ['25-35', '36-50', '25-35', '36-50', '51+'],
    'segment': ['Premium', 'Standard', 'Premium', 'Standard', 'Basic']
})

# Добавляем age_group в customers
customers_with_group = customers_reset.copy()
customers_with_group['age_group'] = pd.cut(customers_with_group['age'], 
                                            bins=[0, 35, 50, 100], 
                                            labels=['25-35', '36-50', '51+'])

print("Merge по двум столбцам (city и age_group):")
merged_multi = pd.merge(customers_with_group, customer_segments, on=['city', 'age_group'], how='left')
print(merged_multi[['name', 'city', 'age', 'age_group', 'segment']].head())

# ========== 3.4 Метод .join() ==========
print("\n--- 3.4 Метод .join() ---")

# .join() используется для объединения по индексу
customers_indexed = customers_reset.set_index('customer_id')
bonus_indexed = customer_bonus.set_index('customer_id')

print("Использование .join():")
customers_with_bonus = customers_indexed.join(bonus_indexed, how='left', lsuffix='_cust', rsuffix='_bonus')
print(customers_with_bonus[['name', 'bonus_points', 'status']].head())
```

### 4. Связь один ко многим

```python
# ========== 4. Связь один ко многим ==========
print("\n--- 4.1 Пример один ко многим ---")

# Один клиент -> много заказов
customer_orders = pd.merge(customers_reset, orders_reset, on='customer_id', how='inner')
print("Связь 'один клиент - много заказов':")
customer_order_counts = customer_orders.groupby('customer_id').size()
for cust_id in [1, 5, 9]:
    print(f"Клиент {cust_id}: {customer_order_counts.get(cust_id, 0)} заказов")

# Демонстрация с детальным выводом
print("\nДетали заказов клиента 1:")
customer_1_orders = customer_orders[customer_orders['customer_id'] == 1]
print(customer_1_orders[['order_id', 'order_date', 'amount', 'status']])

# ========== 4.2 Агрегация данных при связи один ко многим ==========
print("\n--- 4.2 Агрегация при связи один ко многим ---")

# Агрегируем заказы для каждого клиента
customer_stats = customer_orders.groupby('customer_id').agg({
    'order_id': 'count',
    'amount': ['sum', 'mean', 'max']
}).round(2)

customer_stats.columns = ['total_orders', 'total_amount', 'avg_amount', 'max_amount']
print("Агрегированная статистика по клиентам:")
print(customer_stats)

# Добавляем агрегированные данные обратно к клиентам
customers_with_stats = pd.merge(customers_reset, customer_stats, on='customer_id', how='left')
customers_with_stats.fillna(0, inplace=True)
print("\nКлиенты с агрегированной статистикой:")
print(customers_with_stats[['name', 'total_orders', 'total_amount', 'avg_amount']])
```

### 6. reset_index()
